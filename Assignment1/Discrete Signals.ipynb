```python
import matplotlib.pyplot as plt
import numpy as np
def timescale(x,c):
    r=np.empty([len(x)])
    for i in range(len(x)):
        if i%c==0:
            r[i]=x[i]
        else:
            r[i]=0
    return r
def timeshift(x,s):
    t=np.empty([len(x)])
    for i in range(len(x)):
        if i<s:
            t[i]=x[i-s+len(x)]
        elif (i-s)>=len(x):
            t[i]=x[i-s-len(x)]
        else:
            t[i]=x[i-s]
    return t
def nishan():
    n=np.arange(-50,49,1)
    x=np.empty([10,len(n)])
    te=np.empty([len(n)])
    imp=np.empty([len(n)])
    col=['c','m','b','g','r','y','k','c','m','b','g']
    for i in range(len(n)):
        if i==int(len(n)/2+1):
            imp[i]=1
        else:
            imp[i]=0
    x[0]=imp
    x[1]=3*imp
    x[2]=3*(timescale(imp,2))
    x[3]=3*(timeshift(timescale(imp,2),5))
    xlab=['x1[n]=[n]','x2[n]=3[n]','x3[n]=3[2n]','x4[n]=3[2n-5]','x5[n]=3[n^2+3n+2]','x6[n]=x1[n]+x2[n]+x3[n]','x7[n]=[n]','x8[n]=[n-k] (Impulse Train)','x9[n]=[n]']
    for i in range(len(n)):
        w=i-50
        if (w**2+3*w+2)==0:
            x[4,i]=3*imp[50]
        else:
            x[4,i]=0
    x[5]=x[0]+x[1]+x[2]
    x[6]=imp
    for i in range(len(n)):
        x[7,i]=0
    for i in n:
        te=timeshift(imp,i)
        x[7]+=te
    x[8]=imp
    plt.figure(1,figsize=(15,4))
    fig=2
    plt.scatter(n,imp,color=col[-1])
    plt.bar(n,imp,color=col[-1])
    plt.xlabel('n')
    plt.ylabel('Amplitude')
    plt.title('Figure for impulse signal [n]')
    plt.grid('True')
    for i in range(9):
        plt.figure(fig,figsize=(15,4))
        titl='Figure for '+xlab[i]
        plt.scatter(n,x[i],color=col[i])
        plt.bar(n,x[i],color=col[i])
        plt.xlabel('n')
        plt.ylabel('Amplitude')
        plt.title(titl)
        plt.grid('True')
        fig+=1
nishan()

    
```
