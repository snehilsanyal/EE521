```python
import numpy as np 
import matplotlib.pyplot as plt
# defining function to generate unit step function 
def unit_step (c,n):
    unit= []
    for sample in n:
        if sample<c:
            unit.append(0)
        else:
            unit.append(1)
    return(unit)
c = 0 #delay or advancement of the signal
u= 5#upper limit of signal 
l = -5# lower limt of signal 
n = np.arange(l,u,1)
unit= unit_step(c,n)
plt.stem(n,unit,use_line_collection=True)
plt.xlabel('n')
plt.ylabel('amplitude')
plt.title('unit step signal')
plt.show()
```

```python
import numpy as np 
import matplotlib.pyplot as plt
# defining function to generate unit step function 
def unit_ramp (n):
    ramp= []
    for sample in n:
        if sample<0:
            ramp.append(0)
        else:
            ramp.append(sample)
    return ramp
u= 5#upper limit of signal 
l = -5# lower limt of signal 
n = np.arange(l,u,1)
r= unit_ramp(n)
plt.stem(n,r,use_line_collection=True)
plt.xlabel('n')
plt.ylabel('amplitude')
plt.title('unit ramp signal')
plt.show()
```

```python
#Periodic square wave using impulse
import numpy as np
import matplotlib.pyplot as plt 
from scipy import signal 
sr  = 20 # sampling rate
t= 1 # time interval 
n = 10 # number of pulse
f = 1/(t/(n/2))# pulse frequency
n = np.linspace(-1,t,sr,endpoint=False)
plt.stem(n,signal.square(2*np.pi*f*n),use_line_collection=True)
plt.grid()
plt.title(' x[n]=square pulse')
plt.xlabel('n')
plt.ylabel('amplitude')
plt.show()
```

```python
import numpy as np 
import matplotlib.pyplot as plt
def unit_step (c,n):
    unit= []
    for sample in n:
        if sample<c:
            unit.append(0)
        else:
            unit.append(1)
    return(unit)
c = 0 #delay or advancement of the signal
u= 5#upper limit of signal 
l = -5# lower limt of signal 
n = np.arange(l,u,1)
unit= unit_step(c,n)
plt.stem(n,unit,use_line_collection=True)
plt.xlabel('n')
plt.ylabel('amplitude')
plt.title('step input signal')
plt.show()

def impulse(a,n):
    delta=[]
    for sample in n:
        if sample== a:
            delta.append(3)
        else:
            delta.append(0)
    return delta
a =3
u = 10
l= -10
n = np.arange(l,u,1)
x2= impulse(a,n)
plt.stem(n,x2,use_line_collection=True)
plt.xlabel('n')
plt.ylabel('amplitude')
plt.title('system response=y[n]=2delta[n-3]')
plt.show()
Y = np.convolve(x2,unit)
n = np.arange(0,19)
print('output of the system:',Y)
```
