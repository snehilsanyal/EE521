```python
import matplotlib.pyplot as plt
import numpy as np
t=np.arange(0,3,0.01)
A=[-5,5]
B=[-5,-2,-0.5,-0.25,0,1,2]
x=np.empty([len(t)])
w=1
col=['r','c','k','g','b','y','m']
for i in A:
    c=0
    for j in B:
        x=i*np.exp(j*t)
        plt.figure(w)
        plt.plot(t,x,col[c])
        titl="Plot when A="+str(i)+" and B="+str(j)
        plt.title(titl)
        plt.xlabel('time (t)')
        plt.ylabel('Amplitude')
        plt.grid('True')
        c+=1
        w+=1
```
