```python
import matplotlib.pyplot as plt
import numpy as np
def convo(x,h,hlen):
    t=np.empty([len(x)])
    hnk=np.empty([len(x)])
    conv=np.full(len(x),0)
    for s in range(-50+hlen,51):
        for i in range(len(h)):
            if i<s:
                t[i]=x[i-s+len(h)]
            elif (i-s)>=len(x):
                t[i]=x[i-s-len(h)]
            else:
                t[i]=x[i-s]
        hnk=t[::-1]
        for i in range(len(h)):
            conv[i]+=x[i]*hnk[i]
    return conv
def nishan():
    n=np.arange(-50,49,1)
    x=[2,4,5,2,7]
    h=[8,-5,4]
    xapp=np.empty([len(n)])
    happ=np.empty([len(n)])
    con=np.empty([len(n)])
    for i in n:
        if i>=0 and i<len(x):
            xapp[i+50]=x[i]
        else:
            xapp[i+50]=0
        if i>=0 and i<len(h):
            happ[i+50]=h[i]
        else:
            happ[i+50]=0
    con=convo(xapp,happ,len(h))
    #print(con)    
    plt.figure(1,figsize=(15,4))
    plt.scatter(n,xapp,color='b')
    plt.bar(n,xapp,color='g')
    plt.xlabel('n')
    plt.ylabel('Amplitude')
    plt.title('Figure for impulse signal x[n]')
    plt.grid('True')
    plt.figure(2,figsize=(15,4))
    plt.scatter(n,happ,color='b')
    plt.bar(n,happ,color='g')
    plt.xlabel('n')
    plt.ylabel('Amplitude')
    plt.title('Figure for impulse signal h[n]')
    plt.grid('True')
    plt.figure(3,figsize=(15,4))
    plt.scatter(n,con,color='b')
    plt.bar(n,con,color='g')
    plt.xlabel('n')
    plt.ylabel('Amplitude')
    plt.title('Figure for impulse signal y[n]')
    plt.grid('True')
nishan()
```
