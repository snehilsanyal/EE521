```python
import matplotlib.pyplot as plt
import numpy as np
t=np.arange(0,5,0.01)
x1=5*np.sin(2*np.pi*t)
x2=2*np.sin(2*np.pi*t/3)
x3=4*np.cos(np.pi*t/3)
x4=3*np.cos(2*np.pi*t+np.pi/3)
x5=x1+x2
x6=x3+x4
x7=x1+x3
x8=x1[::-1]
x9=x3[::-1]
plt.figure(1)
plt.plot(t,x1,'c')
plt.title('x1(t)=5sin(2t)')
plt.xlabel('Time(t)')
plt.ylabel('Amplitude')
plt.grid('True')
plt.figure(2)
plt.plot(t,x2,'k')
plt.title('x2(t)=2sin(2t/3)')
plt.xlabel('Time(t)')
plt.ylabel('Amplitude')
plt.grid('True')
plt.figure(3)
plt.plot(t,x3,'m')
plt.title('x3(t)=4cos(t/3)')
plt.xlabel('Time(t)')
plt.ylabel('Amplitude')
plt.grid('True')
plt.figure(4)
plt.plot(t,x4,'r')
plt.title('x4(t)=3cos(2t+/3)')
plt.xlabel('Time(t)')
plt.ylabel('Amplitude')
plt.grid('True')
plt.figure
plt.figure(5)
plt.plot(t,x5,'g')
plt.title('x5(t)=x1(t)+x2(t)')
plt.xlabel('Time(t)')
plt.ylabel('Amplitude')
plt.grid('True')
plt.figure
plt.figure(6)
plt.plot(t,x6,'y')
plt.title('x6(t)=x3(t)+x4(t)')
plt.xlabel('Time(t)')
plt.ylabel('Amplitude')
plt.grid('True')
plt.figure
plt.figure(7)
plt.plot(t,x7,'r')
plt.title('x7(t)=x1(t)+x3(t)')
plt.xlabel('Time(t)')
plt.ylabel('Amplitude')
plt.grid('True')
plt.figure
plt.figure(8)
plt.plot(t,x8,'b')
plt.title('x8(t)=x1(-t)')
plt.xlabel('Time(t)')
plt.ylabel('Amplitude')
plt.grid('True')
plt.figure
plt.figure(9)
plt.plot(t,x9,'k')
plt.title('x9(t)=x3(-t)')
plt.xlabel('Time(t)')
plt.ylabel('Amplitude')
plt.grid('True')
plt.figure
```
