```python
import matplotlib.pyplot as plt
import numpy as np
import cmath
t=np.arange(0,10,0.01)
A=[0.1,0.5,1,2]
b=[-0.25,-0.5,-1,0.5,1]
w=[2*np.pi,np.pi*1/6,5*np.pi/3]
wlabel=['w=2π','w=π/6','w=5π/3']
col=['r','g','b','y']
y=np.empty([len(A),len(b),len(w),len(t)])
for i in range(len(A)):
    for j in range(len(b)):
        for k in range(len(w)):
            y[i,j,k]=A[i]*np.exp(b[j]*t)*np.cos(w[k]*t)
for i in range(len(A)):
    for j in range(len(b)):
        plt.figure(len(b)*i+j+1)
        for k in range(len(w)):
            plt.plot(t,y[i,j,k],col[k],label=wlabel[k])
        plt.xlabel('Time (t)')
        plt.ylabel('Amplitude')
        titl='Plot of y(t)=Ae^(bt)cos(wt) when A='+str(A[i])+' and b='+str(b[j])
        plt.title(titl)
        plt.grid('True')
        plt.legend()
fig_num=len(b)*len(A)
A2=[1,1,2,2]
w2=[0.14*np.pi,0.14*np.pi,0.34*np.pi,0.75*np.pi]
theta=[0,np.pi,0.5*np.pi,np.pi/3]
y2=np.empty([len(A2),2,len(t)])
w2label=['w=0.14π','w=0.14π','w=0.34π','w=0.75π']
thetalabel=['Θ=0','Θ=π','Θ=0.5π','Θ=π/3']
for i in range(len(A2)):
    y2[i,0]=A2[i]*(np.cos(w2[i]*t+theta[i]))
    y2[i,1]=A2[i]*(np.sin(w2[i]*t+theta[i]))
for i in range(len(A2)):
    plt.figure(fig_num+i+1)
    plt.plot(t,y2[i,0],col[0],label='Real Part')
    plt.plot(t,y2[i,1],col[1],label='Imaginary Part')
    titl='Plot of y(t)=Ae^j(wt+Θ) when A='+str(A2[i])+' and '+w2label[i]+' and '+thetalabel[i]
    plt.title(titl)
    plt.xlabel('Time (t)')
    plt.ylabel('Amplitude')
    plt.legend()
    plt.grid('True')
            
```
