```python
import matplotlib.pyplot as plt
import numpy as np
n=10
n2=100
tau=2
u=np.full((n),1)
rect=np.full((n2),2/tau)
r=np.empty([n])
p=np.empty([n])
#rect=np.empty([n])
tri=np.empty([n])
t1=np.arange(0,10,1)
t2=np.arange(-1,1,0.02)
sinc=np.empty([n2])
for t in range (n):
    r[t]=t
    p[t]=(t**2)/2
for t in range(n2):
    if t==n2/2:
        sinc[t]=1
    else:
        sinc[t]=np.sin(np.pi*(t-n2/2)/10)/(np.pi*(t-n2/2)/10)
A=5
B=0.5
C=2
t3=np.arange(t1[0]+C,t1[-1]+1+C,1*B)
t4=np.arange(t2[0]+C,t2[-1]+0.02+C,0.02*B)
w1=int(len(u)/B)
u2=np.full(w1,A)
w2=int(len(r)/B)
r2=np.empty([w2])
w3=int(len(p)/B)
p2=np.empty([w3])
w4=len(t4)
rect2=A*np.full((w4),(t4[-1]-t4[0])/tau)
w5=int(len(t4))
sinc2=np.empty([w5])
for t in range(w2):
    r2[t]=A*t
    p2[t]=A*(t**2)/2
for t in range(w5):
    if t==w5/2:
        sinc2[t]=A*1
    else:
        sinc2[t]=A*np.sin(np.pi*(t-w5/2)/10)/(np.pi*(t-w5/2)/10)
plt.figure(1)
plt.plot(t1,u,'r',t3,u2,'b')
plt.title('Unit Step Function vs Unit step Ax(Bt+C)')
plt.xlabel('Time (t)')
plt.xlim(-2,15)
plt.ylim(-2,10)
plt.grid('True')
plt.ylabel('Amlpitude')
plt.figure(2)
plt.plot(t1,r,'c',t3,r2,'k')
plt.title('Unit Ramp Function vs Unit Ramp Ax(Bt+C)')
plt.grid('True')
plt.xlabel('Time (t)')
plt.ylabel('Amlpitude')
plt.figure(3)
plt.plot(t2,rect,'b',t4,rect2,'p')
plt.title('Rectangle Function')
plt.xlabel('Time (t)')
plt.ylabel('Amlpitude')
plt.xlim(-2,4)
plt.ylim(-3,10)
plt.grid('True')
plt.figure(4)
plt.plot(t2,sinc,'c',t4,sinc2,'k')
plt.title('Sinc Function vs Ax(Bt+C)')
plt.xlabel('Time (t)')
plt.ylabel('Amlpitude')
plt.grid('True')
plt.figure(5)
plt.plot(t1,p,'b',t3,p2,'k')
plt.title('Polynomial Function t^2/2')
plt.xlabel('Time (t)')
plt.ylabel('Amlpitude')
plt.grid('True')
```
