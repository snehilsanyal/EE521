#Basic signals

```
import matplotlib.pyplot as plt
import numpy as np

#Unit step function
def unit_step(t,A,B,C):
  N=len(t)
  y=np.zeros(N)
  for i in range(0,N-1):
    if B*t[i]>=-C:
      y[i]=A
  return y

#Ramp function
def ramp(t,A,B,C):
  N=len(t)
  y=np.zeros(N)
  for i in range(0,N-1):
    if B*t[i]>=-C:
      y[i]=A*(B*t[i]+C)
  return y

#Parabolic function
def parabolic(t,A,B,C):
  N=len(t)
  y=np.zeros(N)
  for i in range(0,N-1):
    if B*t[i]>=-C:
      y[i]=(A*(B*t[i]+C)**2)/2
  return y

#Triangular pulse
def tri(t,T):
  N=len(t)
  y=np.zeros(N)
  for i in range(0,N-1):
    if t[i]>=-T/2 and t[i]<=0:
      y[i]=t[i]+T/2
    if t[i]<=T/2 and t[i]>0:
      y[i]=-t[i]+T/2
  return y

#Rectangular pulse
def rect(t,T):
  N=len(t)
  y=np.zeros(N)
  for i in range(0,N-1):
    if t[i]<=T/2 and t[i]>=-T/2:
      y[i]=1
  return y

#Sinc function
def sinc(t):
  N=len(t)
  y=np.zeros(N)
  for i in range(0,N-1):
    y[i]=np.sin(np.pi*t[i])/(np.pi*t[i])
  return y


t=np.linspace(-5,5,100)
y=np.zeros([6,100])
y[0,:]=unit_step(t,1,1,0)
y[1,:]=ramp(t,1,1,0)
y[2,:]=parabolic(t,1,1,0)
y[3,:]=tri(t,4)
y[4,:]=rect(t,4)
y[5,:]=sinc(t)


plt.figure(1,figsize=(20,10))
plt.subplot(2,3,1)
plt.plot(t,y[0,:])
plt.title("Unit step function")
plt.grid(True)
plt.subplot(2,3,2)
plt.plot(t,y[1,:])
plt.title("Ramp Function")
plt.grid(True)
plt.subplot(2,3,3)
plt.plot(t,y[2,:])
plt.title("Parabolic function")
plt.grid(True)
plt.subplot(2,3,4)
plt.plot(t,y[3,:])
plt.title("Triangular pulse")
plt.grid(True)
plt.subplot(2,3,5)
plt.plot(t,y[4,:])
plt.title("Rectangular pulse")
plt.grid(True)
plt.subplot(2,3,6)
plt.plot(t,y[5,:])
plt.title("Sinc Function")
plt.grid(True)
```

#Exponentials

```
#Exponential functions

def exponential(t,A,B):
  N=len(t)
  y=np.zeros(N)
  for i in range(0,N-1):
    y[i]=A*np.exp(B*t[i])
  return y

t=np.linspace(-10,10,100)
A=np.array([-5,5])
B=np.array([-5,-2,-0.5,-0.25,0,1,2])
plt.figure(1,figsize=(20,20))
k=1
for i in A:
  for j in B:
    plt.subplot(4,4,k)
    plt.plot(t,exponential(t,i,j))
    plt.title(f"{i}exp({j}t)")
    k+=1
```

```
import numpy as np
import matplotlib.pyplot as plt
def sin_exp(t,A,B,w):
  N=len(t)
  y=np.zeros(N)
  for i in range(0,N-1):
    y[i]=A*np.exp(B*t[i])*np.cos(w*t[i])
  return y

A=np.array([ 0.1, 0.5, 1, 2])
B=np.array([ -0.25, -0.5, -1, 0.5, 1])
w=np.array([2*np.pi, np.pi/6,5*(np.pi/3)])
t=np.linspace(-5,5,100)
plt.figure(1)
plt.figure(figsize=(30,50))
h=1
for i in A:
  for j in B:
    for k in w:
      plt.subplot(12,5,h)
      plt.grid(True)
      plt.plot(t,sin_exp(t,i,j,k))
      plt.title(f"{i}exp({j}t)xcos({k}t)")
      h+=1
```

#Complex exponentials

```
def real(t,A,w,theta):
  N=len(t)
  y=np.zeros(N)
  for i in range(0,N-1):
    y[i]=A*np.cos(w*t[i]+theta)
  return y

def im(t,A,w,theta):
  N=len(t)
  y=np.zeros(N)
  for i in range(0,N-1):
    y[i]=A*np.sin(w*t[i]+theta)
  return y

A=np.array([1,1,2,2])
w=np.array([0.14*np.pi,0.14*np.pi,0.34*np.pi,0.75*np.pi])
theta=np.array([0,np.pi,0.5*np.pi,(np.pi/3)])
t=np.linspace(-10,10,100)
plt.figure(1)
plt.figure(figsize=(30,20))
k=1
for i in range(0,4):
  plt.subplot(4,2,k)
  plt.plot(t,real(t,A[i],w[i],theta[i]),c='g')
  plt.title(f"real part of {A[i]}exp(j{w[i]}t+{theta[i]})")
  k+=1
  plt.subplot(4,2,k)
  plt.plot(t,im(t,A[i],w[i],theta[i]),c='r')
  plt.title(f"imaginary part of {A[i]}exp(j{w[i]}t+{theta[i]})")
  k+=1
```

#**Discrete signals**

```
import numpy as np
import matplotlib.pyplot as plt

def delta(n,A,B,C,D):
  i=np.where(B*n*n+C*n+D==0)
  L=len(n)
  x_n=np.zeros(L)
  x_n[i]=A
  return x_n

n=np.arange(-10,10)
plt.figure(1,figsize=(20,10))
plt.subplot(2,4,1)
plt.stem(n,delta(n,1,0,1,0),use_line_collection=True)
plt.title("unit impulse function")

plt.subplot(2,4,2)
plt.stem(n,delta(n,3,0,1,0),use_line_collection=True)
plt.title("3delta[n]")

plt.subplot(2,4,3)
plt.stem(n,delta(n,3,0,2,0),use_line_collection=True)
plt.title("3delta[2n]")

plt.subplot(2,4,4)
plt.stem(n,delta(n,3,0,2,-5),use_line_collection=True)
plt.title("3delta[2n-5]")

plt.subplot(2,4,5)
plt.stem(n,delta(n,3,1,3,2),use_line_collection=True)
plt.title("3delta[n^2+3n+2]")

plt.subplot(2,4,6)
plt.stem(n,delta(n,3,1,3,2),use_line_collection=True)
plt.title("x1[n]+x2[n]+x3[n]")

L=len(n)
z=np.zeros(L)
for i in n:
  p=np.where(n==i)
  z[p]=1

plt.subplot(2,4,7)
plt.stem(n,z,use_line_collection=True)
plt.title("impulse train")


z=np.zeros(L)
for i in n:
  if i%10<5:
    z[i]=1

plt.subplot(2,4,8)
plt.stem(n,z,use_line_collection=True)
plt.title("Periodic square wave")
```

In discrete domain the counterpart for differenciation is difference.Hence,for eg. the unit step function can be derived by ramp function by taking the difference

```
#ramp function from parbolic function
y=parabolic(n,1,1,-2)
x=parabolic(n,1,1,2)
z=np.zeros(len(n))
for i in range(len(n)):
  z[i]=(x[i]-y[i])/2

plt.stem(n,z,use_line_collection=True)
plt.title("ramp function from parbolic function")
```

```
#unit step function from ramp function
n=np.arange(-10,10)
y=ramp(n,1,1,0)
x=ramp(n,1,1,1)
z=np.zeros(len(n))
for i in range(len(n)):
  z[i]=x[i]-y[i]

plt.stem(n,z,use_line_collection=True)
plt.title("unit step function from ramp function")
```

```
#unit impulse function from unit step function
n=np.arange(-10,10)
y=unit_step(n,1,1,-1)
x=unit_step(n,1,1,0)
z=np.zeros(len(n))
for i in range(len(n)):
  z[i]=x[i]-y[i]

plt.stem(n,z,use_line_collection=True)
plt.title("unit impulse function from ramp function")
```

#**Convolution**

```
#Convolution function
def convolution(x,h):
  M=len(x)
  N=len(h)
  y=np.zeros(M+N-1)
  for n in range(0,M+N-1):
    k=0
    while n-k>=0:
      if n-k<N and k<M:
        y[n]=y[n]+x[k]*h[n-k]
      k=k+1
  return y

print(convolution([2,4,5,2,7],[8,-5,4]))
```
