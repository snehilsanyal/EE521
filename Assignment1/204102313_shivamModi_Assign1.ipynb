```python
x_n=[2,4,5,2,7]
h_n=[8,-5,4]
n=np.linspace(-5,5,11)
```

```python
from scipy import signal
```

```python
def convolution(inputSeq, impulseResponse):
    inputSeq = np.array(inputSeq)
    impulseResponse = np.array(impulseResponse)
    n1=len(inputSeq)
    n2=len(impulseResponse)
    outputSeq=[]
    
    for n in range(n1+n2-1):
        summ=0
        for k in range(n1):
            if n-k>=0 and n-k<=n2-1:
                summ=summ+inputSeq[k]*impulseResponse[n-k]
        outputSeq=np.append(outputSeq,summ)
    return outputSeq



convolution(x_n,h_n)

#functoin defined above
plt.subplot(1,2,1)
print("my own function output is",convolution(x_n,h_n))
plt.stem(convolution(x_n,h_n))
plt.title("convoluted output of x_n,h_n")



#predefined function
plt.subplot(1,2,2)
print("standard function output is",np.convolve(x_n,h_n))
plt.stem(np.convolve(x_n,h_n))
plt.title("standard output of x_n,h_n")
```
