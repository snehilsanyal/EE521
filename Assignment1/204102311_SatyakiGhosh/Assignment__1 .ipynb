• Basic signals
Consider the following signal with basic transformations of the form:
y(t) = Ax(Bt + C) where A is the amplitude scaling factor, B is the
time scaling factor, C is the time shift.
Plot the following signals: x 1 (t) = u(t) , x 2 (t) = r(t) = tu(t) ,
2
x 3 (t) = p(t) = t u(t)
, x 4 (t) = rect(t/τ ), x 5 (t) = tri(t/τ ) , x 6 (t) =
2
sinct = sin πt πt where u(t) is the unit impulse signal, r(t) is the ramp
signal, and p(t) is the parabolic signal, rect(t/τ ) is a rectangular
pulse between −τ /2 to τ /2 , similarly tri(t/τ ) is triangular pulse
from −τ /2 to τ /2 . 

```
import matplotlib.pyplot as plt
import numpy as np
import math

zeroSamples=1000
numSamples=1000

# UNIT STEP 
t=np.linspace(-10,10,zeroSamples+numSamples)
x1=[0 for i in range(zeroSamples)]
for i in range(numSamples):
  x1.append(1)

#UNIT RAMP
x2=[0 for i in range(zeroSamples)]
for i in list(np.linspace(0,10,numSamples)):
  x2.append(i)

#UNIT PARABOLIC
x3=[0 for i in range(zeroSamples)]
for i in list(np.linspace(0,10,numSamples)):
  x3.append((i**2)/2)

#RECTANGULAR PULSE
t2=np.linspace(-2,2,zeroSamples+numSamples)
x4=[0 for i in range(zeroSamples-250)]
for i in range(500):
  x4.append(1)
for i in range(numSamples-250):
  x4.append(0) 

#TRIANGULAR PULSE
t2=np.linspace(-2,2,zeroSamples+numSamples)
x5=[0 for i in range(zeroSamples-250)]
for i in list(np.linspace(0,1,250)):
  x5.append(i)
for i in list(np.linspace(1,0,250)):
  x5.append(i)
for i in range(numSamples-250):
  x5.append(0) 

# SINC FUNTION
t3=np.linspace(-6,6,zeroSamples+numSamples)
#x6=np.sinc(t3)
x6 = np.sin(math.pi * t3) / (math.pi * t3)


#PLOTTING THE BASIC FUNCTIONS
plt.rcParams.update({'font.size': 15})
fig,ax=plt.subplots(2,3,figsize=(30,15))
#plt.figure(figsize=(10,10))

ax[0,0].plot(t,x1,c='r',label='unit step function')
ax[0,0].set_xlabel('time')
ax[0,0].legend()
ax[0,0].grid()
ax[0,1].plot(t,x2,c='g',label='unit ramp function')
ax[0,1].set_xlabel('time')
ax[0,1].legend()
ax[0,1].grid()
ax[0,2].plot(t,x3,c='b',label='unit parabolic function')
ax[0,2].set_xlabel('time')
ax[0,2].legend()
ax[0,2].grid()
ax[1,0].plot(t2,x4,c='y',label='rectangular pulse function (-0.5,0.5)')
ax[1,0].set_xlabel('time')
ax[1,0].legend()
ax[1,0].grid()
ax[1,1].plot(t2,x5,c='c',label='triangular pulse function (-0.5,0.5)')
ax[1,1].set_xlabel('time')
ax[1,1].legend()
ax[1,1].grid()
ax[1,2].plot(t3,x6,c='m',label='sinc function')
ax[1,2].set_xlabel('time')
ax[1,2].legend()
ax[1,2].grid()

fig.suptitle('Basic signals')

plt.show()
```

**• Exponentials**
Consider the signals of the form: x(t) = Ae Bt . Plot x(t) for dif-
ferent values of A and B. Take A = −5 and 5 and B in range
[−5, −2, −0.5, −0.25, 0, 1, 2]. Plot all possible combinations to get a
taste of how exponentials behave.

```
import matplotlib.pyplot as plt
import numpy as np

A= [-5,5]
B= [ -5, -2, -0.5, -0.25, 0, 1, 2]
rnge=1
t=np.linspace(-rnge,rnge,100*rnge)
plt.rcParams.update({'font.size': 15})

for a in A: 
  for b in B: 
    x=a*np.exp(b*t)
    plt.figure(figsize=(10,6))
    plt.plot(t,x,c='r')
    plt.xlabel('time (s)')
    plt.ylabel('signal value')
    plt.title(f'Exponential signal : A={a}, B={b}')
    plt.grid()
    plt.show()

```

**• Complex Exponentials**
Consider the signals of form: y(t) = Ae Bt cos ωt . Plot the sig-
nals, where A = 0.1, 0.5, 1, 2 and b = −0.25, −0.5, −1, 0.5, 1 and
ω = 2π, π 6 , 5π

```
import matplotlib.pyplot as plt
import numpy as np

pie=np.pi
A= [0.1, 0.5, 1, 2]
B= [ -0.25, -0.5, -1, 0.5, 1 ]
omega = [2*pie, pie/6 , 5*pie/3]

rnge=10
t=np.linspace(-rnge,rnge,100*rnge)
plt.rcParams.update({'font.size': 15})

for a in A: 
  for b in B: 
    for w in omega:
      x=a*np.exp(b*t)*np.cos(w*t)
      plt.figure(figsize=(10,4))
      plt.plot(t,x,c='r')
      plt.xlabel('time (s)')
      plt.ylabel('signal value')
      plt.title(f'Continuous Exponentially varying sinusoidal signal: {a}*exp({b}t)*cos({w}t)')
      plt.grid()
      plt.show()
```

A special type of complex exponential is y(t) = Ae st , where s = σ +
jω . Consider, a phases shift introduced to the signal y(t) with phase
θ . The resultant signal becomes, y(t) = Ae j(ωt+θ) . Plot the real and
imaginary parts ( Re(y(t)) and Im(y(t)) ) of the following signals,
with the following parameters (A, ω, θ): (1, 0.14π, 0) , (1, 0.14π, π) ,
(2, 0.34π, 0.5π) , (2, 0.75π, π/3) .

```
import matplotlib.pyplot as plt
import numpy as np


pie=np.pi
params=[(1, 0.14*pie, 0) , (1, 0.14*pie, pie) ,(2, 0.34*pie, 0.5*pie) , (2, 0.75*pie, pie/3)]
rnge=10
t=np.linspace(-rnge,rnge,100*rnge)
plt.rcParams.update({'font.size': 15})

for param in params:
  a=param[0]
  w=param[1]
  theta=param[2]
  real=a*np.cos(w*t + theta)
  imag=a*np.sin(w*t + theta)
  fig,ax=plt.subplots(1,2,figsize=(10,4))
  ax[0].plot(t,real,label='real part')
  ax[0].set_xlabel('time (s)')
  ax[0].grid()
  ax[0].legend()
  ax[1].plot(t,imag,label='imaginary part',c='r')
  ax[1].set_xlabel('time (s)')
  ax[1].grid()
  ax[1].legend()
  fig.suptitle(f'Complex Exponential signal : {a}*exp({w} t+{theta})')
  plt.show()
```

**2. Discrete Signals**
Plot the following signals: x 1 [n] = δ[n] , x 2 [n] = 3δ[n] , x 3 [n] = 3δ[2n] ,
x 4 [n] = 3δ[2n−5] , x 5 [n]
= 3δ[n 2 +3n+2] , x 6 [n] = x 1 [n]+x 2 [n]+x 3 [n] ,
P ∞
x 7 [n] = δ[n] , x 8 [n] = k=−∞ δ[n − k] (Impuse Train), x 9 [n] = δ[n]


```
import matplotlib.pyplot as plt
import numpy as np

rnge=10
n=np.arange(-rnge,rnge+1,1)
plt.rcParams.update({'font.size':12})

# delta [n]
x1=[0]*rnge + [1] + [0]*rnge
plt.figure(figsize=(10,4))
plt.stem(n,x1,use_line_collection=True)
plt.xlabel(' Time (samples)')
plt.grid()
plt.title('Discrete Unit Impulse Function = $\delta$[n] =  x1[n] = x7[n] = x9[n] ')
plt.show()

# 3delta [n]
x2=[0]*rnge + [3] + [0]*rnge
plt.figure(figsize=(10,4))
plt.stem(n,x2,use_line_collection=True)
plt.xlabel(' Time (samples)')
plt.grid()
plt.title(' 3 $\delta$[n]')
plt.show()

# 3delta [2n]
x3=[0]*rnge + [3] + [0]*rnge
plt.figure(figsize=(10,4))
plt.stem(n,x3,use_line_collection=True)
plt.xlabel(' Time (samples)')
plt.grid()
plt.title(' 3 $\delta$[2n]')
plt.show()

# 3delta [2n-5]
x4=[0]*rnge + [0] + [0]*rnge
plt.figure(figsize=(10,4))
plt.stem(n,x4,use_line_collection=True)
plt.xlabel(' Time (samples)')
plt.grid()
plt.title(' 3 $\delta$[2n-5]')
plt.show()

# delta[n2 + 3n + 2]
x5=[]
for i in n:
  if(i**2 + 3*i + 2 == 0):
    x5.append(3)
  else:
    x5.append(0)
plt.figure(figsize=(10,4))
plt.stem(n,x5,use_line_collection=True)
plt.xlabel(' Time (samples)')
plt.grid()
plt.title('3 $\delta$[n\N{SUPERSCRIPT TWO} + 3n + 2]')
plt.show()

# x1 + x2 + x3
x6=np.array(x1) + np.array(x2) + np.array(x3)
plt.figure(figsize=(10,4))
plt.stem(n,x6,use_line_collection=True)
plt.xlabel(' Time (samples)')
plt.grid()
plt.title(' x1[n] + x2[n] + x3[n]')
plt.show()

# x1 + x2 + x3
x8=[1]*21
plt.figure(figsize=(10,4))
plt.stem(n,x8,use_line_collection=True)
plt.xlabel(' Time (samples)')
plt.grid()
plt.title(' $\Sigma$ k=−∞ δ[n − k]')
plt.show()
```

The impulse function, unit step function, ramp function and parabolic
function in continuous time are related by the differentiation operation.
If we differentiate unit step we get impulse and so on. Is there any re-
lation between the discrete counterparts? If yes how will you implement
ramp function from parabolic function, unit step from ramp and impulse
function from unit step in the discrete time domain.

```

import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import MaxNLocator
import math

zeroSamples=10
numSamples=10
rnge=10
n=np.arange(-rnge,rnge+1,1)


# UNIT STEP 
#t=np.linspace(-10,10,zeroSamples+numSamples)
x1=[]
for i in n:
  if(i<0):
    x1.append(0)
  else:
    x1.append(1)


#UNIT RAMP
x2=[0 for i in range(zeroSamples)]
for i in list(np.arange(0,numSamples+1,1)):
  x2.append(i)

#UNIT PARABOLIC
x3=[0 for i in range(zeroSamples)]
for i in list(np.arange(0,numSamples+1,1)):
  x3.append((i**2)/2)


#RAMP FROM PARABOLIC
rampFromPara = np.array(list(np.diff(x3)))
for i in range(np.shape(rampFromPara)[0]):
  if(rampFromPara[i]>0):
    rampFromPara[i]=rampFromPara[i]-0.5

#UNIT STEP FROM RAMP
unitFromRamp = list(np.diff(x2))


#IMPULSE FROM UNIT
impulseFromUnit =  list(np.diff(x1))
impulseFromUnit.pop()
impulseFromUnit=[0]+impulseFromUnit

n=n[:-1]
plt.rcParams.update({'font.size':14})
fig,ax=plt.subplots(1,3,figsize=(30,8))
ax[0].xaxis.set_major_locator(MaxNLocator(integer=True))
ax[0].stem(n,rampFromPara,use_line_collection=True)
ax[0].set_title('Ramp from Parabolic function')
ax[0].set_xlabel('Samples')
ax[0].grid()
ax[1].xaxis.set_major_locator(MaxNLocator(integer=True))
ax[1].stem(n,unitFromRamp,use_line_collection=True)
ax[1].set_title('Unit step from Ramp function')
ax[1].set_xlabel('Samples')
ax[1].grid()
ax[2].xaxis.set_major_locator(MaxNLocator(integer=True))
ax[2].stem(n,impulseFromUnit,use_line_collection=True)
ax[2].set_title('Impulse from Unit Step function')
ax[2].set_xlabel('Samples')
ax[2].grid()

plt.suptitle('Differentiation in discrete domain')
plt.show()

```

**In contrast to differentiation in continuous domain, we have difference in discrete domain**. We take the differences of adjacent samples and do some required adjustments. For eg, we subtract 0.5 from the signal obtained by taing consecutive differences of the parabolic function to obtain the ramp function.

3. **Systems** (Time domain approach)
Every discrete LTI system is characterized by its impulse response. The
output of a discrete LTI system in terms of the impulse response is given
by the equation:
y[n] = x[n] ∗ h[n] =
∞
X
x[k]h[n − k]
(1)
k=−∞
This is the convolution sum. Write a function for performing convolution.

Use this function to calculate the output for the following system: x[n] =
2, 4, 5, 2, 7, h[n] = 8, −5, 4.

```
# Signature of Convolution.py
## Function for computing convolution of 2 sequences
def convolution(x, h):
  x = list(x)
  h = list(h)         #converting to lists
  if(len(x)<len(h)):  #taking the smaller of the two in h
    x,h=h,x

  ## Do something with inputSeq and impulseResponse
  ## Both are numpy arrays or list of numbers
  n=len(x)
  m=len(h)            #finding the lengths of the two lists
  L= n+2*(m-1)
  x=[0]*(m-1)+x+[0]*(m-1)
  h.reverse()         #reversing the 'h' list
  hRev=h+[0]*(n+m-2)  #converting each list to length = n+2(m-1) 

  leftLim=0
  rightLim=n+m-2    # length of output is n+m-1: [0 , n+m-2]
  y=[0]*(n+m-1)
  for i in range(leftLim,rightLim+1):   #filling up the output list one by one
    for j in range(L):                  #iterating over the input list and impulse_response list 
      y[i]=y[i]+(x[j]*hRev[j])
    hRev.pop()                          # hRev shifted rightwards by one after calculating each y[i]
    hRev=[0]+hRev
  return y

x = np.array([2, 4, 5, 2, 7])
h = np.array([8, -5, 4])
print(f'Convolution of Input signal(x):{x} and Impulse response(h):{h}:')
print(convolution(x,h))
```

```
import numpy as np
import matplotlib.pyplot as plt
import cmath

impulseResponse = [0,0,0,2]


zeroSamples=10
numSamples=10
rnge=10
n=np.arange(-rnge,rnge+1,1)


# UNIT STEP 
#t=np.linspace(-10,10,zeroSamples+numSamples)
step=[]
for i in range(rnge+1):
    step.append(1)
stepResponse=convolution(step,impulseResponse)
n=np.arange(0,len(stepResponse),1)
fig,ax=plt.subplots(1,2,figsize=(40,4))
plt.rcParams.update({'font.size':20})
ax[0].stem(np.arange(0,len(step),1),step,use_line_collection=True,label='Unit step')
ax[0].set_xlabel('Samples')
ax[0].grid()
ax[0].legend()
ax[1].stem(n,stepResponse,use_line_collection=True,label='Step response')
ax[1].set_xlabel('Samples')
ax[1].grid()
ax[1].legend()
fig.suptitle('Step response')
plt.show()


#UNIT RAMP
ramp=[]
for i in list(np.arange(0,numSamples+1,1)):
  ramp.append(i)
rampResponse=convolution(ramp,impulseResponse)
n=np.arange(0,len(rampResponse),1)
fig,ax=plt.subplots(1,2,figsize=(40,4))
#plt.rcParams.update({'font.size':12})
ax[0].stem(np.arange(0,len(ramp),1),ramp,use_line_collection=True,label='Unit ramp')
ax[0].set_xlabel('Samples')
ax[0].grid()
ax[0].legend()
ax[1].stem(n,rampResponse,use_line_collection=True,label='Ramp response')
ax[1].set_xlabel('Samples')
ax[1].grid()
ax[1].legend()
fig.suptitle('Ramp response')
plt.show()

#UNIT PARABOLIC
para=[]
for i in list(np.arange(0,numSamples+1,1)):
  para.append((i**2)/2)
paraResponse=convolution(para,impulseResponse)
n=np.arange(0,len(paraResponse),1)
fig,ax=plt.subplots(1,2,figsize=(40,4))
#plt.rcParams.update({'font.size':12})
ax[0].stem(np.arange(0,len(para),1),para,use_line_collection=True,label='Unit Parabolic funtion')
ax[0].set_xlabel('Samples')
ax[0].grid()
ax[0].legend()
ax[1].stem(n,paraResponse,use_line_collection=True,label='Parabolic response')
ax[1].set_xlabel('Samples')
ax[1].grid()
ax[1].legend()
fig.suptitle('Parabolic response')
plt.show()

# COMPLEX EXPONENTIAL
s=complex(2,3)
rnge=10
n=np.arange(0,rnge+1,1)
expInput=np.exp(s*n)
realInput=expInput.real
imagInput=expInput.imag

expResponse=np.array(convolution(expInput,impulseResponse))
realResp=expResponse.real
imagResp=expResponse.imag

#input
fig,ax=plt.subplots(1,2,figsize=(40,4))
#plt.rcParams.update({'font.size':12})
ax[0].stem(np.arange(0,len(realInput),1),realInput,use_line_collection=True,label='Complex exponential input : real part')
ax[0].set_xlabel('Samples')
ax[0].grid()
ax[0].legend()
ax[1].stem(np.arange(0,len(imagInput),1),imagInput,use_line_collection=True,label='Complex exponential input : imaginary')
ax[1].set_xlabel('Samples')
ax[1].grid()
ax[1].legend()
fig.suptitle('Complex exponential input')
plt.show()

#response
fig,ax=plt.subplots(1,2,figsize=(40,4))
#plt.rcParams.update({'font.size':12})
ax[0].stem(np.arange(0,len(realResp),1),realResp,use_line_collection=True,label='Complex exponential response : real part')
ax[0].set_xlabel('Samples')
ax[0].grid()
ax[0].legend()
ax[1].stem(np.arange(0,len(imagResp),1),imagResp,use_line_collection=True,label='Complex exponential response : imaginary')
ax[1].set_xlabel('Samples')
ax[1].grid()
ax[1].legend()
fig.suptitle('Complex exponential response')
plt.show()
```
