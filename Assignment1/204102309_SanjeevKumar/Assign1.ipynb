```
import numpy as np
import matplotlib.pyplot as plt
t = np.linspace(-10,10,100)
#Step signal
def step(t,scale):
    y = [0 for i in range(len(t))]
    for i in range(len(t)):
        if t[i]>=0:
            y[i] = scale
        else:
            y[i] = 0
    return y
plt.subplot(231)
plt.plot(t,step(t,1),c='g')
plt.title('Unit step')
# Ramp signal
def ramp(t,scale):
    y = t*step(t,scale)
    return y
plt.subplot(232)
plt.plot(t,ramp(t,1),c='g')
plt.title('Ramp signal')
# Parabolic Signal
def parabolic(t,scale):
  return (0.5*t**2)*step(t,scale)
plt.subplot(233)
plt.plot(t,parabolic(t,1),c='g')
plt.title('Parabloic signal')
#Rectangular signal for T = 4
def rect(t,scale):
    a = step(t+2,scale)
    b = step(t-2,scale)
    y = list(map(lambda x,y:(x-y),a,b))
    return y
plt.subplot(234)
plt.plot(t,rect(t,1),'g')
plt.title('Rectangular signal')
#Triangular signal for T = 4
def tri(t,scale):
    a = ramp(t+2,scale)
    b = ramp(t,scale)
    c = ramp(t-2,scale)
    y = list(map(lambda x,y,z:(x-2*y+z),a,b,c))
    return y
plt.subplot(235)
plt.plot(t,tri(t,1),'g')
plt.title('Triangular signal')
#sinc signal
def sinc(t):
  return np.sin(np.pi*t)/(np.pi*t)
plt.subplot(236)
plt.plot(t, sinc(t), 'g')
plt.title('Sinc signal')
plt.tight_layout()
plt.show()
```

**Discrete Signals**

```
# IMPULSE SIGNAL
n = np.linspace(-5,5,11)
def x1(n,scale):
    y=[]
    for i in n:
        if i == 0:
            y.append(scale)
        else:
            y.append(0)
    return y
plt.stem(n,x1(n,1))
plt.title('x1[n]')
plt.show()
```

**Differentiation** in discrete domain is carried out by subtractiing the given signal by the same signal delayed by 1 unit at all instants.

```
# Obtaining Impulse signal from step signal
y=[]
for i in range(len(n)):
  y.append(step(n,1)[i]-step(n-1,1)[i])
plt.stem(n,y)
```

**Convolution Function**

```
def conv(x,h):
    m = len(x)
    n = len(h)
    y = [0 for i in range(n+m-1)]
    x = x + [0 for i in range(n)]
    h = h + [0 for i in range(m)]
    for i in range(n+m-1):
        for j in range(m):
            if i-j+1>0:
                y[i] = y[i]+x[j]*h[i-j]
    return y
# Implementation of the defined convolution function
x = [2,4,5,2,7]
h = [8,-5,4]
print(conv(x,h))
```
