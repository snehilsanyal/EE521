# BASIC CONTINUOUS TIME SIGNALS
##   Y  =  Ax(Bt+C)
###   Where  A= 5,B=0.5,C=2

```python
import numpy as np
import matplotlib.pyplot as plt


t = np.linspace(-8, 8, 1000)
A=5
B1=0.5
C=2

def u(t):
    return A * (t >= 0)

def r(t):
    return A*t * (t>=0)

def p(t):
    return A*(t**2)/2 * (t>=0)

def rect(t):
    return A * (abs(t) < 0.5)

def tri(t):
    return (1 - abs(t))*A * (abs(t) < 1)

def sinc(t):
    return np.sin(np.pi*t)*A/np.pi*t

functions = [u,r,p,rect,tri,sinc ]

plt.figure()
for i, function in enumerate(functions, start=1):
    plt.subplot(3, 3, i)
    plt.plot(t, function(B1*t+C), '-b')
plt.tight_layout()
plt.show()
```

# SINUSOIDS

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(-5, 5, 50)

def x1(t):
    return np.sin(np.pi*t*2)*5

def x2(t):
    return np.sin(np.pi*t*2/3)*2

def x3(t):
    return np.cos(np.pi*t/3)*4

def x4(t):
    return np.cos(np.pi*t*2+np.pi/3)*3

def x5(t):
    return x1(t) + x2(t)

def x6(t):
    return x3(t) + x4(t)

def x7(t):
    return x1(t) + x3(t)

def x8(t):
    return x1(-t)

def x9(t):
    return x3(-t)

functions = [x1,x2,x3,x4,x5,x6,x7,x8,x9 ]

plt.figure()
for i, function in enumerate(functions, start=1):
    plt.subplot(3, 3, i)
    plt.plot(t, function(t), '-b')
plt.tight_layout()
plt.show()
```

# DISCRETE SIGNALS
##  IMPULSE FUNCTION

```python
import numpy as np 
import matplotlib.pyplot as plt 

def unit_impulse1(a, n): 
    delta =[] 
    for sample in n: 
        if sample == a: 
            delta.append(1) 
        else: 
            delta.append(0) 
              
    return delta 

def unit_impulse2(a, n): 
    delta =[] 
    for sample in n: 
        if sample == a: 
            delta.append(3) 
        else: 
            delta.append(0) 
              
    return delta 


def unit_impulse3(a, n): 
    delta =[] 
    for sample in n: 
        if sample**2+3*sample+2 == 0: 
            delta.append(3) 
        else: 
            delta.append(0) 
              
    return delta 




a=5
n = np.arange(-10, 10, 1) 
d1 =unit_impulse1(0, n)
d2 =unit_impulse2(0, n)
d3 =unit_impulse2(a, n)
d4 =unit_impulse3(0, n)
plt.subplot(6,1,1)
plt.stem(n, d1,use_line_collection=True)
plt.subplot(6,1,2)
plt.stem(n, d2,use_line_collection=True)
plt.subplot(6,1,3)
plt.stem(n/2, d2,use_line_collection=True)
plt.subplot(6,1,4)
plt.stem(n/2, d3,use_line_collection=True)
plt.subplot(6,1,5)
plt.stem(n, d4,use_line_collection=True)
plt.tight_layout()
```

### IMPULSE TRAIN

```python
import numpy as np 
import matplotlib.pyplot as plt 


def impulsetrain(a, n): 
    delta =[] 
    for sample in n: 
        if sample == a: 
            delta.append(1) 
        else: 
            delta.append(0) 
              
    return delta 


n = np.arange(-10, 10, 1)
for i in range(-1000,1000):
    res = impulsetrain(i, n)
    plt.stem(n,res,use_line_collection=True)
    
```

### DISCRETE SQUARE WAVE USING IMPULSE TRAIN

```python
import numpy as np 
import matplotlib.pyplot as plt 


sqr=[]
n = np.arange(-30, 35, 1)
for i in n:
    if i%10<5:
        sqr.append(1)
        
    else:
        sqr.append(0)
        

plt.stem(n,sqr,use_line_collection=True)
plt.title('Discrete periodic square wave')
plt.show()
```

# DISCRETE BASIC SIGNALS


```python
import numpy as np
import matplotlib.pyplot as plt



def unit_step(a, n): 
    unit =[] 
    for sample in n: 
        if sample<a: 
            unit.append(0) 
        else: 
            unit.append(1) 
    return(unit)

def unit_ramp(n): 
    ramp =[] 
    for sample in n: 
        if sample<0: 
            ramp.append(0) 
        else: 
            ramp.append(sample) 
    return ramp 

def parabola(n): 
    para =[] 
    for sample in n: 
        if sample<0: 
            para.append(0) 
        else: 
            para.append((sample**2)/2) 
    return para 

def rect(n): 
    rec =[] 
    N=4
    for sample in n: 
        if ((sample)<N and (sample)>-N): 
            rec.append(1) 
        else: 
            rec.append(0) 
    return rec

def trit(n): 
    tri =[] 
    N=4
    for sample in n: 
        if (abs(sample))<N : 
            tri.append(N-abs(sample)) 
        else:
            tri.append(0)
    return tri


a=0
n = np.arange(-5, 20, 1) 
unit = unit_step(a, n)
r = unit_ramp(n)
p= parabola(n)
d= rect(n)
t=trit(n)
plt.subplot(3,3,1)
plt.title('step')
plt.stem(n, unit,use_line_collection=True)
plt.subplot(3,3,2)
plt.title('ramp')
plt.stem(n, r,use_line_collection=True) 
plt.subplot(3,3,3)
plt.title('parabola')
plt.stem(n, p,use_line_collection=True) 
plt.subplot(3,3,4)
plt.title('rect')
plt.stem(n, d,use_line_collection=True)   
plt.subplot(3,3,5)
plt.title('tri')
plt.stem(n, t,use_line_collection=True)
plt.tight_layout()
```
