# **Basic signals**

```

def unit_step(t):
    """Unit step signal"""
    return 1 * (t >= 0)

def unit_ramp(t):
  """Unit ramp signal"""
  return t*(t>=0)

def parabolic(t):
  return ((t**2)/2)*(t>0)
def rectangular(t):
    """ rectangular pulse signal"""
    return 1 * (abs(t) < 0.5)

def triangular(t):
    """ triangular signal"""
    return (1 - abs(t)) * (abs(t) < 1)
def sinc_f(t):
  """sinc signal"""
  return np.sin(np.pi*t)/(np.pi*t)

```

```
A=5
signal = [unit_step,unit_ramp,parabolic,rectangular,triangular,sinc_f]
t=np.linspace(-5, 5, 200)

plt.figure()
for i, function in enumerate(signal, start=1):
    plt.plot(t, A*function(t), '-b')
    plt.title(function.__doc__)
    plt.ylabel(r'y(t) =Ax(Bt+c) ')
    plt.xlabel(r't (in s)')
    plt.grid()
    plt.show()
```

# **2. Discrete Signals**

```
def impulse(n):
  return 1*(n==0)
n=np.arange(-10,10,1)
x1=impulse(n)
x2=3*impulse(n)
x3=3*impulse(2*n)
x4=3*impulse(2*n-5)
x5=3*impulse(n**2+3*n+2)
x6=x1+x2+x3
K=6
x7=np.sum(impulse(n-k) for k in np.arange(-K,K,1)) 
n=np.arange(-10,10,1)
dist_fun=[x1,x2,x3,x4,x5,x6,x7]
for i, f in enumerate(dist_fun,start=1):
    plt.subplot(7,1,i)
    plt.stem(f)
    plt.xlabel(r'n')
    plt.grid()
    plt.show()
```

# **Systems (Time domain approach)**

```
#  Function for computing convolution of 2 sequences
def convolution(inputSeq,impulseResponse):
  inputSeq = np.array(inputSeq)
   
  impulseResponse = np.array(impulseResponse)
 
  m=len(inputSeq)
  n=len(impulseResponse)
  L=m+n-1
  inputSeq=np.pad(inputSeq,(0,L-m),'constant')
  impulseResponse=np.pad(impulseResponse,(0,L-n),'constant')
  outputSeq=np.zeros(L)
  for i in range(0,L):
    sum=0
    for j in range(0,i+1):
      sum=sum+impulseResponse[i-j]*inputSeq[j]
    outputSeq[i]=sum
    



## Do something with inputSeq and impulseResponse
## Both are numpy arrays or list of numbers
  return outputSeq
```

```
xn=2,4,-5,2,7
hn=8,-5,4
yn=convolution(xn,hn)
plt.subplot(3,1,1)
plt.stem(xn)
plt.title("input sequence")
plt.subplot(3,1,2)
plt.stem(hn)
plt.title("inpulse response")
plt.subplot(3,1,3)
plt.stem(yn)
plt.title("output sequence")
print("out put sequence=",yn)
```
