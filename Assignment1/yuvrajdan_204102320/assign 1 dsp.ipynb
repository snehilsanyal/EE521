#           Basic signals

 x1(t) = u(t) , x2(t) = r(t) = tu(t) ,x3(t) = p(t) = t^2/2u(t)

```python
import numpy as np
import matplotlib.pyplot as plt

t=np.linspace(-2,2,1000)
u_t=(t>=0)
r_t=t*u_t
p_t=((t**2)*(u_t))/2
signal=[(u_t,"unit step"),(r_t,"ramp signal" ),(p_t,"parabolic signal")]
c=1
fig, axs = plt.subplots(2, 3, figsize=(10,15))
for (i,j) in signal:
    plt.subplot(3,1,c)
    plt.plot(t,i)
    plt.ylim(-.20,1.20)
    plt.title(str(j))
    plt.axhline(0,color='b',lineWidth=0.5)
    plt.axvline(0,color='b',lineWidth=0.5)
    #plt.axis('off')
    c=c+1
```

# x5(t)=triangular pulse

```python
import numpy as np
import matplotlib.pyplot as plt
fig=plt.figure(1)
t=np.linspace(-2,2,1000)
def traingularfunction(t):
    if t>=-1and t<=0:
        return (t+1)
    elif t>=0 and t<=1:
        return (-t+1)
    else:
        return 0
tra=[]
for time in t:
    tra.append(traingularfunction(time))    
    
plt.plot(t,tra)    
plt.ylim(-.20,1.20)
plt.axhline(0,color='b',lineWidth=0.5)
plt.axvline(0,color='b',lineWidth=0.5)
plt.text(-0.20,1,'$1$',size=15)
plt.xlabel('time(t)')
plt.title('Traingular pulse')
#plt.axis('off')       
        
```

# 2. Discrete Signals


```python
import numpy as np
import matplotlib.pyplot as plt
n=np.arange(-10,20)
A=[1,3,3,3]
B=[1,1,2,2]
C=[0,0,0,5]
parameters=list(zip(A,B,C))
fig, axs = plt.subplots(2, 3, figsize=(20,15))
l=np.size(n)
imp=np.zeros(l)
ind=np.where(n==0)
p=1
name=['x1[n] = δ[n]','x2[n] = 3δ[n]','x3[n] = 3δ[2n]','x4[n] = 3δ[2n−5] ']
j=0
for (a,b,c) in parameters:
    if (c/b) in n:
        imp[ind]=a
        plt.subplot(2,2,p)
        plt.stem((n+c)/b,imp,use_line_collection=True)
        plt.title(str(name[j]))
        plt.xlabel('sample number')
        plt.ylabel('Amplitude')
        plt.grid(True)
        j=j+1
        
    else:

        imp=np.zeros(l)
        plt.subplot(2,2,p)
        plt.stem((n+c)/b,imp,use_line_collection=True)
        plt.title(str(name[j]))
        plt.xlabel('sample number')
        plt.ylabel('Amplitude')
        plt.grid(True)
        j=j+1
        
    p=p+1
    
```

# x5[n] = 3δ[n^2+3n+2]
n^2+3n+2=0  this equation has two roots which  are  n=-1 and n=-2 

```python
import numpy as np
import matplotlib.pyplot as plt
n=np.arange(-10,10)
def imp(n1,n2):
    y=[]
    for i in n:
        if i==n1 or i==n2:
            y.append(3)
        else:
            y.append(0)
            
    return y 
Y=imp(-1,-2)
plt.stem(n,Y,use_line_collection=True)
plt.title('x5[n] = 3δ[n^2+3n+2]')
plt.xlabel('sample number')
plt.ylabel('Amplitude')
plt.grid(True)
```

# Ploting step,ramp and prabolic signal in discerte

```python
#plotting step signal,ramp,parabolic
import numpy as np
import matplotlib.pyplot as plt
n=np.arange(-10,10)
def imp():
    y=[]
    for i in n:
        if (i<0):
            y.append(0)
        else:
            y.append(1)
            
    return y 
def imp1():
    y=[]
    for i in n:
        if (i<0):
            y.append(0)
        else:
            y.append(i)
            
    return y 
def imp2():
    y=[]
    for i in n:
        if (i<0):
            y.append(0)
        else:
            y.append((i*i)/2)
            
    return y 

Y=imp()
Y1=imp1()
Y2=imp2()
fig, axs = plt.subplots(2, 3, figsize=(10,15))
plt.subplot(3,1,1)
plt.stem(n,Y,use_line_collection=True)
plt.title('u[n]  step signal')
plt.xlabel('sample number')
plt.ylabel('Amplitude')
plt.grid(True)
plt.subplot(3,1,2)
plt.stem(n,Y1,use_line_collection=True)
plt.title('r[n]  ramp signal')
plt.xlabel('sample number')
plt.ylabel('Amplitude')
plt.grid(True)
plt.subplot(3,1,3)
plt.stem(n,Y2,use_line_collection=True)
plt.title('p[n]  parabolic signal')
plt.xlabel('samples')
plt.ylabel('Amplitude')
plt.grid(True)
```

# Convolution

```python
import numpy as np
import matplotlib.pyplot as plt
def conv(x,h):
    inputSeq=x
    impulseResponse=h
    m=len(inputSeq)
    n=len(impulseResponse)
    L=m+n-1
    OutputSeq=[]
    inputSeq=np.pad(inputSeq,(0,L-m),'constant')
    impulseResponse=np.pad(impulseResponse,(0,L-n),'constant')
    
    for n in range(0,L):
        total_sum=0
        for k in range(0,L):
            if n>=k:
                 total_sum=total_sum+inputSeq[n-k]*impulseResponse[k]
        #OutputSeq.append[total_sum]            
        #print(total_sum)
        OutputSeq.append(total_sum)
        #print(OutputSeq)
    return OutputSeq   
h=[8,-5,4]
x=[2,4,5,2,7]
Y=conv(x,h)
print(Y)
fig, axs = plt.subplots(2, 3, figsize=(15,15))
n=np.arange(0,7)
plt.subplot(3,1,1)
plt.stem(n,Y,use_line_collection=True)
plt.ylabel("Amplitude")
plt.title('output sequence')
plt.subplot(3,1,2)
plt.title('Input sequence')
plt.ylabel("Amplitude")
n1=np.arange(0,5)
plt.stem(n1,x,use_line_collection=True)
plt.subplot(3,1,3)
plt.title('Impulse response')
plt.ylabel("Amplitude")
n2=np.arange(0,3)
plt.stem(n2,h,use_line_collection=True)
```
