# BASIC SIGNALS :

```python
import numpy as np                         ##import libraries
import matplotlib.pyplot as plt


## unit step function

def unit_step(time,amp,phase):
    u = np.zeros(len(time))
    for i in range(len(time)):
        if time[i] >= phase:
            u[i] = amp
    return u

## ramp function 

def ramp_sig(time,slope,phase):
    r = np.zeros(len(time))
    for i in range(len(time)):
        if time[i]>= -phase:
            r[i] =slope*(time[i] + phase)
    return r 

## parabolic function

def parabolic_sig(time,amp,phase) :
    p = np.zeros(len(time))
    for i in range(len(time)):
        if time[i] >=  phase:
            p[i] = 0.5*amp*((time[i])**2 + phase)
    return p
    
t = np.linspace(-5,5,1000)

x1 = unit_step(t,1,0)
plt.subplot(331)
plt.plot(t,x1)
plt.title('unit step')

x2 = ramp_sig(t,4,0)
plt.subplot(332)
plt.plot(t,x2)
plt.title('unit ramp')

x3 = parabolic_sig(t,1,0)
plt.subplot(333)
plt.plot(t,x3)
plt.title('parabolic signal')


## rectangular function
u1 = unit_step(t,6,-3)
u2 = unit_step(t,6,3)
x4 = u1 - u2
plt.subplot(334)
plt.plot(t,x4)
plt.title('rectangular signal')
plt.ylim(0,10)


## triangular function 

r1 = ramp_sig(t,4,3)
r2 = ramp_sig(t,-8,0)
r3 = ramp_sig(t,4,-3)
x5 = r1+r2+r3
plt.subplot(335)
plt.plot(t,x5)
plt.ylim(0,20)
plt.title('triangular signal')
```

```python
## the amplitude scaled and time shifted signals with amplitude scaling factor A = 5 and shift C = 2 sec

M1 = 5*unit_step(t,1,2)
plt.subplot(331)
plt.plot(t,M1)
plt.title('unit step')
M2 = 5*ramp_sig(t,4,-2)
plt.subplot(332)
plt.plot(t,M2)
plt.title('unit ramp')
M3 = parabolic_sig(t,1,2)
plt.subplot(333)
plt.plot(t,M3)
plt.title('parabolic')
```

```python
##plot the time scaled versions of the original signals with time scaling factor as B = 0.5

M1 = unit_step(0.5*t,5,0)
plt.subplot(331)
plt.plot(t,M1)
plt.title('unit step')
M2 = ramp_sig(0.5*t,4,0)
plt.subplot(332)
plt.plot(t,M2)
plt.title('unit ramp')
M3 = parabolic_sig(0.5*t,1,0)
plt.subplot(333)
plt.plot(t,M3)
plt.title('parabolic')
```

```python
##plot the time scaled versions of the original signals with time scaling factor as B = 2

M1 = unit_step(2*t,5,0)
plt.subplot(331)
plt.plot(t,M1)
plt.title('unit step')
M2 = ramp_sig(2*t,4,0)
plt.subplot(332)
plt.plot(t,M2)
plt.title('unit ramp')
M3 = parabolic_sig(2*t,1,0)
plt.subplot(333)
plt.plot(t,M3)
plt.title('parabolic')
```

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0,10,1000)          
f1 = 2 
x1 = 5*np.sin(np.pi*f1*t)                  ## x1(t) = 5 sin 2t

f2 = 2/3
x2 = 2*np.sin(np.pi*f2*t)                 ## x2(t) = 2 sin (2/3)t 

f3 = 1/3
x3 = 5*np.sin(np.pi*f3*t)                ## x3(t) = 4 cos t/3


f4 = 2                                       ## x4(t)  = 3 cos(2t + /3)
p4 = 1/3
x4 = 5*np.sin(np.pi*f4*t + np.pi*p4)

x5 = x1 + x2                                ## x5(t) = x1(t) + x2(t)

x6 = x3+x4                                 ## x6(t) = x3(t) + x4(t) 

x7 = x1 + x3                                ## x7(t) = x1(t) + x3(t)

f8 = 2 
x8 = 5*np.sin(np.pi*f8*(-t))             ## x8(t) = x1(t) 

f9 = 1/3
x9 = 5*np.sin(np.pi*f9*(-t))                  ## x9(t) = x3(t)

plt.subplot(331)
plt.plot(t,x1)
plt.title('5sin2t')
plt.subplot(332)
plt.plot(t,x2)
plt.title('2sin(2t/3)')
plt.subplot(333)
plt.plot(t,x3)
plt.title('4cos t/3')
plt.subplot(334)
plt.plot(t,x4)
plt.title('3 cos(2t + /3)')
plt.subplot(335)
plt.plot(t,x5)
plt.title('5sin2t + 2sin(2t/3) ')
plt.subplot(336)
plt.plot(t,x6)
plt.title('4cos t/3 + 3 cos(2t + /3) ')
plt.subplot(337)
plt.plot(t,x7)
plt.title('5sin2t + 4cos t/3')
plt.subplot(338)
plt.plot(t,x8)
plt.title('5sin2(-t)')
plt.subplot(339)
plt.plot(t,x9)
plt.title('4cos (-t)/3')
```

```python
import numpy as np
import matplotlib.pyplot as plt

def impulse(phase,time,amp):
    delfun = []
    for i in time:
        if i == phase:
            delfun.append(amp)
        else:
            delfun.append(0)
    return delfun

t = np.arange(-10,10,0.5)
x1 = impulse(0,t,1)
plt.subplot(331)
plt.stem(t,x1,use_line_collection = True)
plt.title('x1[n] = [n]')

x2 = impulse(0,t,3)
plt.subplot(332)
plt.stem(t,x2,use_line_collection = True)
plt.title('x2[n] = 3[n]')

x3 = impulse(0,2*t,3)
plt.subplot(333)
plt.stem(t,x3,use_line_collection = True)
plt.title('x3[n] = 3[2n]')

x4 = impulse(5,2*t,3)
plt.subplot(334)
plt.stem(t,x4,use_line_collection = True)
plt.title('x3[n] = 3[2n-5]')

x5 = impulse(0,t**2 + 3*t + 2,3)
plt.subplot(335)
plt.stem(t,x5,use_line_collection = True)
plt.title('x3[n] = 3[n^2+3n+2]')

from operator import add
x = list(map(add,x1,x2))
x6 = list(map(add,x,x3))
plt.subplot(336)
plt.stem(t,x6,use_line_collection = True)
plt.title('x6[n] = x1+x2+x3')
```

```python
## x8[n] = impulse train

def impulse(phase1,phase2,time,amp):
    delfun = []
    for i in time:
        if i >= phase1 and i <= phase2:
            delfun.append(amp)
        else:
            delfun.append(0)
    return delfun

p1 = int(input("enter the value of phase 1 : "))
p2 = int(input("enter the value of phase 2 : "))
x8 = impulse(p1,p2,t,1)
plt.stem(t,x8,use_line_collection = True)
plt.title('impulse train')
```

```python
## unit step function

def unit_step(time,amp,phase):
    u = np.zeros(len(time))
    for i in range(len(time)):
        if time[i] >= phase:
            u[i] = amp
    return u

t = np.linspace(-5,5,50)

x1 = unit_step(t,1,0)
plt.subplot(331)
plt.plot(t,x1)
plt.title('unit step')
x2 = unit_step(t,1,0)
plt.subplot(334)
plt.stem(t,x2,use_line_collection = True)
plt.title('discrete unit step')


## ramp function 

def ramp_sig(time,slope,phase):
    r = np.zeros(len(time))
    for i in range(len(time)):
        if time[i]>= -phase:
            r[i] =slope*(time[i] + phase)
    return r 

x3 = ramp_sig(t,4,0)
plt.subplot(332)
plt.plot(t,x3)
plt.title('unit ramp')
x4 = ramp_sig(t,4,0)
plt.subplot(335)
plt.stem(t,x4,use_line_collection = True)
plt.title('discrete ramp step')


## parabolic function

def parabolic_sig(time,amp,phase) :
    p = np.zeros(len(time))
    for i in range(len(time)):
        if time[i] >=  phase:
            p[i] = 0.5*amp*((time[i])**2 + phase)
    return p

x5 = parabolic_sig(t,1,0)
plt.subplot(333)
plt.plot(t,x5)
plt.title('parabolic signal')
x6 = parabolic_sig(t,1,0)
plt.subplot(336)
plt.stem(t,x6,use_line_collection = True)
plt.title('discrete parabolic signal')
```

```python
import numpy as np

list_num_1 = []                                      ## for user input list 
length1 = int(input("length of the input list:"))
for i in range(length1):
    data = int(input())
    list_num_1.append(data)
print("the input list is :", list_num_1)


list_num_2 = []                                      ## for user impulse response list
length2 = int(input("length of the impulse response list:"))
for i in range(length2):
    data = int(input())
    list_num_2.append(data)
print("the input list is :", list_num_2)

def convolution(list_num_1,list_num_2):          ## defining a convolution function
    N = length1 + length2 - 1
    out = np.zeros(N)
    u = N - length1
    v = N - length2
    list_num_1 = np.pad(list_num_1,(0,u),'constant')
    list_num_2 = np.pad(list_num_2,(0,v),'constant')
    for v in range(N):
        for i in range(N):
            if v>= i:
                out[v] = out[v] + list_num_1[v-i]*list_num_2[i]
    return out            
print('the convolved list is : ', convolution(list_num_1,list_num_2))
```
