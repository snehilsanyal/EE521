* **Basic Signals:** Consider the following signal with basic transformations of the form: y(t) = Ax(Bt+C) where A is the amplitude scaling factor, B is the time scaling factor, C is the time shift. Plot the following signals: (Refer Assigment 1 question pdf)

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# for function x1:

t = np.linspace(-10, 10, 100)

def x1(t, a):
    u = [0 for i in range(len(t))]
    for i in range(len(t)):
        u[i] = (a if t[i]>=0 else 0)
    return u

plt.plot(t, x1(t, 1), 'blue')
plt.xlabel('t')
plt.ylabel('u(t)')
plt.title('plot for function x1')
plt.grid()
plt.show
```

```python
# for function x2:

t = np.linspace(-10, 10, 100)

def x2(t, a):
    out = t*x1(t, a)
    return out

plt.plot(t, x2(t, 1), 'blue')
plt.xlabel('t')
plt.ylabel('r(t)')  #x2 = r(t)
plt.title('plot for function x2')
plt.grid()
plt.show
```

```python
# for function x3:

t = np.linspace(-10, 10, 100)

def x3(t, a):
    out = (t**2)*0.5*x1(t, a)
    return out

plt.plot(t, x3(t, 1), 'blue')
plt.xlabel('t')
plt.ylabel('p(t)')  #x3 = p(t) = (t**2)*u(t)*0.5
plt.title('plot for function x2')
plt.grid()
plt.show
```

```python
# for function x4:

t = np.linspace(-10, 10, 1000)

def x4(t):
    return 1 * (abs(t) < 0.5)

plt.plot(t, x4(t), 'blue')
plt.xlabel('t')
plt.ylabel('x4(t)')  
plt.title('plot for function x4')
plt.grid()
plt.show
```

```python
# for function x5:

t = np.linspace(-10, 10, 1000)

def x5(t):
    return (1 - abs(t)) * (abs(t) < 1)

plt.plot(t, x5(t), 'blue')
plt.xlabel('t')
plt.ylabel('x5(t)')  
plt.title('plot for function x5')
plt.grid()
plt.show
```

```python
# for function x6:

t = np.linspace(-10, 10, 1000)

def x6(t):
    return 1 * (t/2 >= 0) - 1 * (t/2 < 0)

plt.plot(t, x6(t), 'blue')
plt.xlabel('t')
plt.ylabel('x6(t)')  
plt.title('plot for function x6')
plt.grid()
plt.show
```

```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
t = np.linspace(-10, 10, 1000)

def sin(a, w, t, phi):
    return a*np.sin(w*t + phi)
```

```python
# for the sinusoidal function x1(t)

def x1(t):
    return sin(5, 2*np.pi, t, 0)

plt.plot(t, x1(t), 'red')
plt.xlabel('t')
plt.ylabel('x1(t)')
plt.title('plot for the sinusoidal function x1(t)')
plt.grid()
plt.show
```

* **Complex Exponentials:**

**Part 1:**

* To plot signal for values of

* A = 0.1, 0.5, 1, 2
* B = -0.25, -0.5, -1, 0.5, 1
* w = 2 * pi, pi/6, 5 * pi/3

```python
t = np.linspace(-10, 10, 100)

def complex_exp(A, B, t, w):
    return A*np.e**(B*t)*np.cos(w*t)
```

```python
# for w = 2*pi

A = [0.1, 0.5, 1, 2]
B = [-0.25, -0.5, -1, 0.5, 1]
w = 2*np.pi

plt.figure()
for i in range(len(A)):
    for j in range(len(B)):
            plt.plot(t, complex_exp(A[i], B[j], t, w), 'purple')
            title = "A="+str(A[i]), "B="+str(B[j]), "w="+str(w)
            plt.grid()
            plt.title(title)
            plt.show()
```

```python
# for w = pi/6

A = [0.1, 0.5, 1, 2]
B = [-0.25, -0.5, -1, 0.5, 1]
w = np.pi/6

plt.figure()
for i in range(len(A)):
    for j in range(len(B)):
            plt.plot(t, complex_exp(A[i], B[j], t, w), 'purple')
            title = "A="+str(A[i]), "B="+str(B[j]), "w="+str(w)
            plt.grid()
            plt.title(title)
            plt.show()
```

```python
# for w = 5*pi/3

A = [0.1, 0.5, 1, 2]
B = [-0.25, -0.5, -1, 0.5, 1]
w = 5*np.pi/3

plt.figure()
for i in range(len(A)):
    for j in range(len(B)):
            plt.plot(t, complex_exp(A[i], B[j], t, w), 'purple')
            title = "A="+str(A[i]), "B="+str(B[j]), "w="+str(w)
            plt.grid()
            plt.title(title)
            plt.show()
```

```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
# for the function x1[n]

n = np.arange(-10, 11, 1)

def x1(n, a):
    d = []
    for i in n:
        if i == a: 
            d.append(1)
        else:
            d.append(0)
            
    return d

plt.stem(n, x1(n, 0), use_line_collection = True)
plt.xlabel('n')
plt.ylabel('d(n)')
plt.title('plot for function x1')
plt.grid()
plt.show
```

```python
# for the function x2[n]


x2 = np.array(x1(n, 3))

plt.stem(n, x2, use_line_collection = True)
plt.xlabel('n')
plt.ylabel('x2')
plt.title('plot for function x2')
plt.grid()
plt.show
```

```python
# for unit step signal (dicrete)

n = np.arange(-10, 11, 1)
def unit_step():
    d = []
    for i in n:
        if(i>=0):
            d.append(1)
        else:
            d.append(0)
            
    return d

plt.stem(n, unit_step(), use_line_collection = True)
plt.xlabel('n')
plt.ylabel('amplitude')
plt.title('plot for unit step dicrete signal')
plt.grid()
plt.show
```

```python
# for ramp signal (dicrete)

n = np.arange(-10, 11, 1)
def ramp():
    d = []
    for i in n:
        if(i>=0):
            d.append(i)
        else:
            d.append(0)
            
    return d

plt.stem(n, ramp(), use_line_collection = True)
plt.xlabel('n')
plt.ylabel('amplitude')
plt.title('plot for ramp dicrete signal')
plt.grid()
plt.show
```

```python
# for parabolic signal (dicrete)

n = np.arange(-10, 11, 1)
def para():
    d = []
    for i in n:
        if(i>=0):
            d.append((i**2)*0.5)
        else:
            d.append(0)
            
    return d

plt.stem(n, para(), use_line_collection = True)
plt.xlabel('n')
plt.ylabel('amplitude')
plt.title('plot for parabolic dicrete signal')
plt.grid()
plt.show
```
