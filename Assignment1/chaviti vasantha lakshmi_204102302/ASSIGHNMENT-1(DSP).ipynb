# **BASIC SIGNALS

```python
def u(t):
    "unit step signal-->u(t)"
    return 1*(t >= 0)
def r(t):
    "ramp signal-->r(t)"
    return t*(1*(t >= 0))
def p(t):
    "parabolic signal-->p(t)"
    return ((t**2)/2)*1*(t >= 0)
def rec(t):
    "rectangular pulse-->rec(t/T)"
    return (t>=(-T/2))^(t>=(T/2))
def tri(t):
    "triangular pulse-->tri(t/T)"
    return (1 -(abs(t)/T))*(abs(t)<T)
def sinc(t):
    "sinc function-->sinc(t)"
    return  np.sin(np.pi*t)/(np.pi*t)
T=1;functions = [u,r,p,rec,tri,sinc]
t = np.linspace(-2,2,1000)
fig = plt.figure(10,figsize=(14,8))
for i,function in enumerate(functions,start=1):
    plt.subplot(3,2,i)
    plt.plot(t,function(t),'blue')
    plt.axhline(0,color='black',linewidth = 0.5)
    plt.axvline(0,color='black',linewidth = 0.5)
    plt.ylim((-0.30,2.20))
    plt.title(function.__doc__,color='red',size=20)
plt.tight_layout()
plt.show()
```

```python
import numpy as np 
import matplotlib.pyplot as plt
fig = plt.figure(1,figsize=(18,6))
def unit_step(a, n): 
    unit =[] 
    for sample in n: 
        if sample<a: 
            unit.append(0) 
        else: 
            unit.append(1) 
    return(unit) 
a=0;n = np.arange(-10,11, 1) 
unit = unit_step(a, n)   
plt.subplot(2,2,1)
plt.stem(n, unit,use_line_collection=True)
plt.title('Unit step signal--> u[n]',color='red',size=20)
def unit_ramp(n): 
    ramp =[] 
    for sample in n: 
        if sample<0: 
            ramp.append(0) 
        else: 
            ramp.append(sample) 
    return ramp 
n = np.arange(-10,11,1) 
r = unit_ramp(n) 
plt.subplot(2,2,2)
plt.stem(n, r,use_line_collection=True)  
plt.title('Ramp signal-->r[n]',color='red',size=20) 
```

```python
fig = plt.figure(1,figsize=(18,8))
def parabolic(n): 
    parabolic =[] 
    for sample in n: 
        if sample<0: 
            parabolic.append(0) 
        else: 
            parabolic.append(sample**2) 
    return parabolic
n = np.arange(-1,1,0.1) 
x = parabolic(n) 
plt.subplot(2,2,3)
plt.stem(n, x,use_line_collection=True) 
plt.xlabel('n') 
plt.xticks(np.arange(-1,1,0.2)) 
plt.ylabel('x[n]') 
plt.title('parabolic Signal',color='red',size=20)
```

```python
fig = plt.figure(1,figsize=(10,3))
def rect(n): 
    rect =[] 
    N=4
    for sample in n: 
        if ((sample)<N and (sample)>-N): 
            rect.append(1) 
        else: 
            rect.append(0) 
    return rect
n = np.arange(-20,20,1) 
x = rect(n) 
plt.stem(n, x,use_line_collection=True) 
plt.title('rectangle Signal',color='red',size=20)
```

```python
def i1(a, n):
    "impulse(n)"
    delta1 =[] 
    for sample in n: 
        if sample == a: 
            delta1.append(1) 
        else: 
            delta1.append(0) 
    return delta1 
def i2(a, n):
    "3*impulse(n)"
    delta2 =[] 
    for sample in n: 
        if sample == a: 
            delta2.append(3) 
        else: 
            delta2.append(0) 
    return delta2 
def i3(a, n): 
    "3*impulse(2n)"
    delta3 =[] 
    for sample in n: 
        if sample == a: 
            delta3.append(3) 
        else: 
            delta3.append(0) 
    return delta3 
def i4(a, n):
    "3*impulse(2n-5)"
    delta4 =[] 
    for sample in n: 
        if (2*sample-2.5) == a: 
            delta4.append(3) 
        else: 
            delta4.append(0) 
    return delta4
def i5(a, n): 
    "3*impulse(n**2+3n+2)"
    delta5 =[] 
    for sample in n: 
        if sample == -2: 
            delta5.append(3)
        elif sample == -3:
            delta5.append(3)
        else: 
            delta5.append(0) 
    return delta5
a=0;functions = [i1,i2,i3,i4,i5]
n = np.arange(-10,10,1)
fig = plt.figure(10,figsize=(10,14))
for i,f in enumerate(functions,start=1):
    plt.subplot(5,1,i)
    plt.stem(n,f(a,n),use_line_collection=True)
    plt.title(f.__doc__,color='red',size=20)
plt.tight_layout()
```

```python
fig = plt.figure(1,figsize=(18,6))
def impulsetrain(n): 
    parabolic =[] 
    for sample in n:  
            parabolic.append(1) 
    return parabolic
n = np.arange(-20,20,1) 
x = impulsetrain(n) 
plt.subplot(2,2,3)
plt.stem(n,x,use_line_collection=True)  
plt.title('Impulse train',color='red',size=20)
```

# **SYSTEMS

```python
import numpy as np
import matplotlib.pyplot as plt
fig = plt.figure(1,figsize=(10,7))
def convolution(inputseq,impulseresponse):
        lengthA=np.size(inputseq)
        lengthB=np.size(impulseresponse)
        C = np.zeros(lengthA + lengthB -1)
        for m in np.arange(lengthA):
            for n in np.arange(lengthB):
                C[m+n] = C[m+n] + inputseq[m]*impulseresponse[n]
        return C 
a=[2,4,5,2,7]
b=[8,-5,4]
n = np.arange(-10,10,1)
p=convolution(a,b)
print("output sequence=",p)
plt.subplot(3,1,1)
plt.stem(p,use_line_collection=True)
plt.title('output',color='red',size=20)
plt.subplot(3,1,2)
plt.stem(a,use_line_collection=True)
plt.title('input',color='red',size=20)
plt.subplot(3,1,3)
plt.stem(b,use_line_collection=True)
plt.title('impulse sequence',color='red',size=20)
```
