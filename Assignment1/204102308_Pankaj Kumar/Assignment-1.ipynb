# Basic Signal

```python
import matplotlib.pyplot as plt
import numpy as np
n=10
n2=100
tau=2
u=np.full((n),1)
rect=np.full((n2),2/tau)
r=np.empty([n])
p=np.empty([n])
tri=np.empty([n])
t1=np.arange(0,10,1)
t2=np.arange(-1,1,0.02)
sinc=np.empty([n2])
for t in range (n):
    r[t]=t
    p[t]=(t**2)/2
for t in range(n2):
    if t==n2/2:
        sinc[t]=1
    else:
        sinc[t]=np.sin(np.pi*(t-n2/2)/10)/(np.pi*(t-n2/2)/10)
A=5
B=0.5
C=2
t3=np.arange(t1[0]+C,t1[-1]+1+C,1*B)
t4=np.arange(t2[0]+C,t2[-1]+0.02+C,0.02*B)
w1=int(len(u)/B)
u2=np.full(w1,A)
w2=int(len(r)/B)
r2=np.empty([w2])
w3=int(len(p)/B)
p2=np.empty([w3])
w4=len(t4)
rect2=A*np.full((w4),(t4[-1]-t4[0])/tau)
w5=int(len(t4))
sinc2=np.empty([w5])
for t in range(w2):
    r2[t]=A*t
    p2[t]=A*(t**2)/2
for t in range(w5):
    if t==w5/2:
        sinc2[t]=A*1
    else:
        sinc2[t]=A*np.sin(np.pi*(t-w5/2)/10)/(np.pi*(t-w5/2)/10)
plt.figure(1)
plt.plot(t1,u,'r',t3,u2,'b')
plt.title('Unit Step Function vs Unit step Ax(Bt+C)')
plt.xlabel('Time (t)')
plt.xlim(-2,15)
plt.ylim(-2,10)
plt.grid('True')
plt.ylabel('Amlpitude')
plt.figure(2)
plt.plot(t1,r,'c',t3,r2,'k')
plt.title('Unit Ramp Function vs Unit Ramp Ax(Bt+C)')
plt.grid('True')
plt.xlabel('Time (t)')
plt.ylabel('Amlpitude')
plt.figure(3)
plt.plot(t2,rect,'b',t4,rect2,'p')
plt.title('Rectangle Function')
plt.xlabel('Time (t)')
plt.ylabel('Amlpitude')
plt.xlim(-5,15)
plt.ylim(-5,15)
plt.grid('True')
plt.figure(4)
plt.plot(t2,sinc,'c',t4,sinc2,'k')
plt.title('Sinc Function vs Ax(Bt+C)')
plt.xlabel('Time (t)')
plt.ylabel('Amlpitude')
plt.grid('True')
plt.figure(5)
plt.plot(t1,p,'b',t3,p2,'k')
plt.title('Polynomial Function t^2/2')
plt.xlabel('Time (t)')
plt.ylabel('Amlpitude')
plt.grid('True')
```

# Exponential Signal

```python
import numpy as np
import matplotlib.pyplot as plt
t=np.arange(0,3,0.01)
A=[-5,5]
B=[-5,-2,-0.5,-0.25,0,1,2]
x=np.empty([len(t)])
w=1
col=['r','c','k','g','b','y','m']
for i in A:
    c=0
    for j in B:
        x=i*np.exp(j*t)
        plt.figure(w)
        plt.plot(t,x,col[c])
        titl="Plot for A="+str(i)+" and B="+str(j)
        plt.title(titl)
        plt.xlabel('time (t)')
        plt.ylabel('Amplitude')
        plt.grid('True')
        c+=1
        w+=1
```

# sinusoids

```python
import matplotlib.pyplot as plt
import numpy as np
t=np.arange(0,5,0.01)
x1=5*np.sin(2*np.pi*t)
x2=2*np.sin(2*np.pi*t/3)
x3=4*np.cos(np.pi*t/3)
x4=3*np.cos(2*np.pi*t+np.pi/3)
x5=x1+x2
x6=x3+x4
x7=x1+x3
x8=x1[::-1]
x9=x3[::-1]
plt.figure(1)
plt.plot(t,x1,'c')
plt.title('x1(t)=5sin(2πt)')
plt.xlabel('Time(t)')
plt.ylabel('Amplitude')
plt.grid('True')
plt.figure(2)
plt.plot(t,x2,'k')
plt.title('x2(t)=2sin(2πt/3)')
plt.xlabel('Time(t)')
plt.ylabel('Amplitude')
plt.grid('True')
plt.figure(3)
plt.plot(t,x3,'m')
plt.title('x3(t)=4cos(πt/3)')
plt.xlabel('Time(t)')
plt.ylabel('Amplitude')
plt.grid('True')
plt.figure(4)
plt.plot(t,x4,'r')
plt.title('x4(t)=3cos(2πt+π/3)')
plt.xlabel('Time(t)')
plt.ylabel('Amplitude')
plt.grid('True')
plt.figure
plt.figure(5)
plt.plot(t,x5,'g')
plt.title('x5(t)=x1(t)+x2(t)')
plt.xlabel('Time(t)')
plt.ylabel('Amplitude')
plt.grid('True')
plt.figure
plt.figure(6)
plt.plot(t,x6,'y')
plt.title('x6(t)=x3(t)+x4(t)')
plt.xlabel('Time(t)')
plt.ylabel('Amplitude')
plt.grid('True')
plt.figure
plt.figure(7)
plt.plot(t,x7,'r')
plt.title('x7(t)=x1(t)+x3(t)')
plt.xlabel('Time(t)')
plt.ylabel('Amplitude')
plt.grid('True')
plt.figure
plt.figure(8)
plt.plot(t,x8,'b')
plt.title('x8(t)=x1(-t)')
plt.xlabel('Time(t)')
plt.ylabel('Amplitude')
plt.grid('True')
plt.figure
plt.figure(9)
plt.plot(t,x9,'k')
plt.title('x9(t)=x3(-t)')
plt.xlabel('Time(t)')
plt.ylabel('Amplitude')
plt.grid('True')
```

# 2. Discrete signals

# X1[n] = d[n]

```python
ndef unit_impulse(a, n): 
    delta =[] 
    for sample in n: 
        if sample == a: 
            delta.append(1) 
        else: 
            delta.append(0) 
              
    return delta 
  
a = 0 
UL = 5
LL = -5
n = np.arange(LL, UL, 1) 
d = unit_impulse(a, n) 
plt.stem(n, d) 
plt.xlabel('n') 
plt.xticks(np.arange(LL, UL, 1)) 
plt.ylim(0, 2) 
plt.ylabel('d[n]') 
plt.title('Unit Impulse d[n]')
```

# X2[n] = 3d[n]

```python
def unit_impulse(a, n): 
    delta =[] 
    for sample in n: 
        if sample == a: 
            delta.append(3) 
        else: 
            delta.append(0) 
              
    return delta 
  
a = 0
UL = 5
LL = -5
n = np.arange(LL, UL, 1) 
d = unit_impulse(a, n) 
plt.stem(n, d) 
plt.xlabel('n') 
plt.xticks(np.arange(LL, UL, 1)) 
plt.ylim(0, 5) 
plt.ylabel('3d[n]') 
plt.title('Unit Impulse 3d[n]')
```
