# 1 . Analog filter 

Que 1-
• Consider a RC circuit with input voltage Vi = 10V applied to the
series RC and output Vo taken from the ends of C with R = 47kΩ
and C = 47nF. Find the transfer function of the circuit and plot
the frequency and magnitude spectrum. Which kind of filter is this?
Prove your claim by putting in the frequency as 100Hz and 10000Hz
and finding the values of Vo. Find the cutoff frequency and verify that
the behaviour is justified.


```python
import numpy as np
import matplotlib.pyplot as plt
import cmath

t = np.arange(1,100,1)
v_in=10*np.sin(2*np.pi*t)
R=47000  # R = 47kohm
C=47*10**(-9)  # c = 47nF
f=np.arange(1,100,1)
omega=2*np.pi*f
# transfer function of RC filter 
TF=1/(1+ 1j*omega*R*C)  

mag=abs(TF)
phase=np.angle(TF)

Xc=1/(2*np.pi*f*C)
Z=(R**2 +Xc**2)**0.5
v_out=(v_in*Xc)/Z # output voltage taken end from capacitor
 
output=[mag,phase]
titles=['magnitude spectrum of RC filter','phase spectrum  of  RC filter ']
fig =plt.figure(figsize=(15,5))
y_axis=['magnitude',' angle']
for i in range(2):
    plt.subplot(1,2,i+1)
    plt.plot(f,output[i])
    plt.title(titles[i])
    plt.ylabel(y_axis[i])
    plt.grid()
```

```python
'''finding the cutoff frequence of RC filter '''
vmax = 10
# let cut off frequency is fo then fc =1/2piRC
fc = 1/(2*np.pi*R*C)
# let given central freq fo =100Hz
fo = 100
Xc=1/(2*np.pi*fo*C)
Z=(R**2 +Xc**2)**0.5
v_out1=(vmax*Xc)/Z
# let given frequency fo =10000Hz
fo=10000
Xc=1/(2*np.pi*fo*C)
Z=(R**2 +Xc**2)**0.5
v_out2=(vmax*Xc)/Z
print('cut off fequency of RC filter is : ' ,fc,'Hz')
print('output voltage when fo = 100Hz is: ' ,v_out1,'volts')
print('output voltage when fo =10000Hz is : ' ,v_out2,'volts')
print('So this RC circuit is a low pass filter ')
```

Take any digital image of your choice and apply mean and median
filtering. What are the differences between the two?

```python
'''Mean filter '''
import cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
      
# Read the image
imgpath  = 'C:/Users/lena.jpg'
img = cv2.imread(imgpath,0)

m, n = img.shape
# Develop Averaging filter(3, 3) mask
ker = np.ones([3, 3], dtype = int)
ker = ker / 9
   
# Convolve the 3X3 mask over the image 
filtered_img1= np.zeros([m, n])
  
for i in range(1, m-1):
    for j in range(1, n-1):
        out = img[i-1, j-1]*ker[0, 0]+img[i-1, j]*ker[0, 1]+img[i-1, j + 1]*ker[0, 2]+img[i, j-1]*ker[1, 0]+ img[i, j]*ker[1, 1]+img[i, j + 1]*ker[1, 2]+img[i + 1, j-1]*ker[2, 0]+img[i + 1, j]*ker[2, 1]+img[i + 1, j + 1]*ker[2, 2]
         
        filtered_img1[i, j]= out
          
filtered_img1= filtered_img1.astype(np.uint8)
output  = [img,filtered_img1]
titles = ['Input image','Averaging filtered image ']
for l in range(2):
    plt.subplot(1,2,l+1)
    plt.imshow(output[l],cmap = 'gray')
    plt.title(titles[l])
    plt.xticks([])
    plt.yticks([])
plt.show()
'''Median filter '''
# Obtain the number of rows and columns 
# of the image
m, n = img.shape
   
# Traverse the image. For every 3X3 area, 
# find the median of the pixels and
# replace the ceter pixel by the median
filtered_img2= np.zeros([m, n])
  
for i in range(1, m-1):
    
    for j in range(1, n-1):
        
        out = [img[i-1, j-1], img[i-1, j],img[i-1, j + 1],img[i, j-1],img[i, j],img[i, j + 1],img[i + 1, j-1],img[i + 1, j],
               img[i + 1, j + 1]]
          
        out = sorted(out)
        filtered_img2[i, j]= out[4]
        
filtered_img2 = filtered_img2.astype(np.uint8)
output  = [img,filtered_img2]
titles = ['Input  image','Median filtered image ']
for l in range(2):
    plt.subplot(1,2,l+1)
    plt.imshow(output[l],cmap = 'gray')
    plt.title(titles[l])
    plt.xticks([])
    plt.yticks([])
plt.show()
```
