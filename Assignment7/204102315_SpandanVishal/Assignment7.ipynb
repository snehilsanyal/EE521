```python
#Importing relevant libraries
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

#**Analog Filters**

#Low Pass Filter
#Transfer Function:$\frac{1}{1+sRC}$

```python
R=47000
C=47*10**(-9)
f=np.arange(1,100,1)
w=2*np.pi*f
H=1/(1+1j*w*R*C)
mag=abs(H)
phase=np.angle(H)
P=[mag,phase]
label=['Magnitude Plot','Phase Plot']

plt.figure(1,figsize=(15,5))
for i in range(2):
  plt.subplot(1,2,i+1)
  plt.plot(f,P[i])
  plt.title(label[i])
  plt.grid(True)
  plt.xlabel("Frequency(Hz)")
```

```python
V=10/(np.sqrt(2))
f=100
Xc=1/(2*np.pi*f*C)
Z=np.sqrt(R**2+Xc**2)
Vo=(V*Xc)/Z
print("The output voltage for f=100Hz",Vo)


f=10000
Xc=1/(2*np.pi*f*C)
Z=np.sqrt(R**2+Xc**2)
Vo=(V*Xc)/Z
print("The output voltage for f=10000Hz",Vo)

print("The cutoff frequency is",1/(2*np.pi*R*C),"Hz")

print("We can conclude from the above magnitudes of transfer function for low and High Frequency that the filter is low pass")
```

#High Pass Filter
#Transfer Function:$\frac{sRC}{1+sRC}$

```python
R=47000
C=47*10**(-9)
f=np.arange(1,100,1)
w=2*np.pi*f
H=(1j*w*R*C)/(1+1j*w*R*C)
mag=abs(H)
phase=np.angle(H)
P=[mag,phase]
label=['Magnitude Plot','Phase Plot']

plt.figure(1,figsize=(15,5))
for i in range(2):
  plt.subplot(1,2,i+1)
  plt.plot(f,P[i])
  plt.title(label[i])
  plt.grid(True)
  plt.xlabel("Frequency(Hz)")
```

#Band Pass Filter
#Transfer Function:$\frac{sRC_1}{(1+sRC_1)(1+sRC_2)}$

```python
R=20000
C1=7.95*10**(-9)
C2=2.65*10**(-10)
f=np.arange(1000,40000,1)
w=2*np.pi*f
H=(1j*w*R*C1)/((1+1j*w*R*C1)*(1+1j*w*R*C2))
mag=abs(H)
phase=np.angle(H)
P=[mag,phase]
label=['Magnitude Plot','Phase Plot']

plt.figure(1,figsize=(15,5))
for i in range(2):
  plt.subplot(1,2,i+1)
  plt.plot(f,P[i])
  plt.title(label[i])
  plt.grid(True)
  plt.xlabel("Frequency(Hz)")
```

What kind of filter is this? Compute the transfer function to plot the frequency and amplitude plots to verify your claim. Consider the

rate to be 1 Hz. Find and plot the filter output.

What is this filter called? Compute the transfer function to plot the

filter? Improve the above filter by the differential equation:

#Moving Average Filter
#$y[n] = \frac{1}{3}(x[n] + x[n − 1] + x[n − 2])$

```python
w = np.linspace(0,2,100)

H = 1/3*(1 + np.exp(-1j*w) + np.exp(-1j*2*w) )
mag = abs(H)
phase = np.angle(H)

P=[mag,phase]
label=['Magnitude Plot','Phase Plot']

plt.figure(1,figsize=(15,5))
for i in range(2):
  plt.subplot(1,2,i+1)
  plt.plot(w,P[i])
  plt.title(label[i])
  plt.grid(True)
  plt.xlabel("Frequency(Hz)")
```

#Hanning Filter
#$y[n] = \frac{1}{4}(x[n]) + \frac{1}{2}(x[n − 1]) + \frac{1}{4}(x[n − 2])$

```python
w = np.linspace(0,2,100)

H = 0.25 + 0.5*np.exp(-1j*w) + 0.25*np.exp(-1j*2*w)
mag = abs(H)
phase = np.angle(H)

P=[mag,phase]
label=['Magnitude Plot','Phase Plot']

plt.figure(1,figsize=(15,5))
for i in range(2):
  plt.subplot(1,2,i+1)
  plt.plot(w,P[i])
  plt.title(label[i])
  plt.grid(True)
  plt.xlabel("Frequency(Hz)")
```

#Applying Mean And Median filters to image

```python
img =  cv2.imread('/content/image.jpg',0)

plt.figure(1,figsize=(15,5))
plt.subplot(1,3,1)
plt.title("Original image")
plt.imshow(img)

#Rows and columns of image
m, n = img.shape
   
# Mean filter(3, 3) mask
mask = np.ones([3, 3], dtype = int)
mask = mask / 9
   
img_new = np.zeros([m, n])
  
for i in range(1, m-1):
    for j in range(1, n-1):
        temp = img[i-1, j-1]*mask[0, 0]+img[i-1, j]*mask[0, 1]+
        img[i-1, j + 1]*mask[0, 2]+img[i, j-1]*mask[1, 0]+ 
        img[i, j]*mask[1, 1]+img[i, j + 1]*mask[1, 2]+
        img[i + 1, j-1]*mask[2, 0]+img[i + 1, j]*mask[2, 1]+
        img[i + 1, j + 1]*mask[2, 2]
        img_new[i, j]= temp
          
img_new = img_new.astype(np.uint8)
plt.subplot(1,3,2)
plt.imshow(img_new)
plt.title("Image after mean filtering")

img_new1 = np.zeros([m, n])
  
for i in range(1, m-1):
    for j in range(1, n-1):
        temp = [img[i-1, j-1],
               img[i-1, j],
               img[i-1, j + 1],
               img[i, j-1],
               img[i, j],
               img[i, j + 1],
               img[i + 1, j-1],
               img[i + 1, j],
               img[i + 1, j + 1]]
          
        temp = sorted(temp)
        img_new1[i, j]= temp[4]
  
img_new1 = img_new1.astype(np.uint8)
plt.subplot(1,3,3)
plt.imshow(img_new1)
plt.title("Image after median filtering")
```

#Filtering Speckle noise from image

```python
from skimage.util import random_noise
img = cv2.imread("image2.jpg",0)
img1 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(20,15))

plt.subplot(131)
plt.title('Original')
plt.imshow(img1)
plt.xticks([])
plt.yticks([])


# speckle noise
uniform = np.random.uniform(0, 1.4,img1.size)
uniform = uniform.reshape(img1.shape[0],img1.shape[1],img1.shape[2]).astype('uint8')
noise = img1 + img1 * uniform
plt.subplot(132)
plt.title('Image with speckle noise')
plt.imshow(noise)
plt.xticks([])
plt.yticks([])

import scipy.signal.signaltools
filtered_img = scipy.signal.signaltools.wiener(noise)
plt.subplot(133)
plt.imshow(filtered_img)
plt.title('Filtered image')
plt.xticks([])
plt.yticks([])
```

#Filtering Salt and Pepper noise from image

```python
import random
def add_noise(img):
  
    # Getting the dimensions of the image
    row , col = img.shape

    number_of_pixels = random.randint(300, 10000)
    for i in range(number_of_pixels):
        
        y=random.randint(0, row - 1)   # Pick a random y coordinate
        x=random.randint(0, col - 1)   # Pick a random x coordinate
        img[y][x] = 255
        # Color that pixel to white
    number_of_pixels = random.randint(300 , 10000)
    for i in range(number_of_pixels):
        y=random.randint(0, row - 1)
        x=random.randint(0, col - 1)
        img[y][x] = 0
          
    return img
  

img1=cv2.imread('image.jpg',1)
b=np.mean(img1,-1)
noisy_img=add_noise(b)
m,n=noisy_img.shape
filtered_img = np.zeros([m, n])
  

#Median Filtering
for i in range(1, m-1):
    for j in range(1, n-1):
        temp = [noisy_img[i-1, j-1],
               noisy_img[i-1, j],
               noisy_img[i-1, j + 1],
               noisy_img[i, j-1],
               noisy_img[i, j],
               noisy_img[i, j + 1],
               noisy_img[i + 1, j-1],
               noisy_img[i + 1, j],
               noisy_img[i + 1, j + 1]]
          
        temp = sorted(temp)
        img_new1[i, j]= temp[4]


filtered_img = img_new1.astype(np.uint8)

plt.figure(1,figsize=(15,5))
plt.subplot(1,3,1)
plt.imshow(cv2.imread('image.jpg',0))
plt.title("Original Image")
plt.xticks([])
plt.yticks([])
plt.subplot(1,3,2)
plt.imshow(noisy_img)
plt.title("Image with salt and pepper noise")
plt.xticks([])
plt.yticks([])
plt.subplot(1,3,3)
plt.imshow(filtered_img)
plt.title("Filtered image using Median filtering")
plt.xticks([])
plt.yticks([])
```

#Analog Filters in digital form

```python
R = 47000
C = 47*(10**(-9))

w = np.linspace(0,np.pi,100)
T = 1

z=np.exp(1j*w)
s = (2/T)*((z - 1)/(z + 1))  


LP1_digital = 1/(s*R*C + 1)
HP_digital = (s*R*C)/(s*R*C + 1)
LP2_digital = (1/(s*R*C + 1))**2

C1=7.95*10**(-9)
C2=2.65*10**(-10)
BP_digital = (s*C1)/((1+s*R*C1)*(1+s*R*C2))

fig = plt.figure(1,figsize=(25,15))



P=[abs(LP1_digital),np.angle(LP1_digital),abs(HP_digital),np.angle(HP_digital),abs(LP2_digital),
   np.angle(LP2_digital),abs(BP_digital),np.angle(BP_digital)]
label=['Magnitude Plot of Digital LP filter','Phase Plot of Digital LP Filter',
       'Magnitude Plot of Digital HP filter','Phase Plot of Digital HP Filter',
       'Magnitude Plot of Digital LP filter-2nd order','Phase Plot of Digital LP Filter-2nd order',
       'Magnitude Plot of Digital Band Pass filter','Phase Plot of Digital Band Pass Filter']

plt.figure(1,figsize=(25,15))
for i in range(8):
  plt.subplot(4,2,i+1)
  plt.plot(w,P[i])
  plt.title(label[i])
  plt.grid(True)
```
