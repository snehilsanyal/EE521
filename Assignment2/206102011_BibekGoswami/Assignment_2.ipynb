```python
#Uniform(a,b)
def uniform_pdf(x,a,b):
    y = []
    for i in range(len(x)):
        if x[i]>=a and x[i]<=b:
            y.append(1/(b-a))
        else:
            y.append(0)
    mean = median = (a+b)/2
    mode = (a+b)/3
    variance = np.power(b-a,2)/12
    return y, mean, median, mode,variance

def uniform_cdf(x,a,b):
    y = []
    for i in range(len(x)):
        if x[i]>=a and x[i]<=b:
            y.append((x[i]-a)/(b-a))
        else:
            y.append(0)
    return y

x = np.linspace(-10,10,100)
```

```python
a = [4,-2,-5]
b = [6,4,3]

fig = plt.figure(figsize=(8,15))

for i in range(len(a)):
    y,cal_mean,cal_mode,cal_median,cal_var=uniform_pdf(x,a[i],b[i])
    plt.subplot(len(a),1,i+1)
    plt.plot(x,y)
    plt.scatter(cal_mean, 0,c='r')
    plt.scatter(cal_median,0,c='g')
    plt.scatter(cal_mode,0,c='b')
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.title("PDF of Uniform distribution for a="+str(a[i])+" & b="+str(b[i]))

plt.show()
```

```python
# Exponential(lambda)
def exponential_pdf(x,lam):
    y = []
    for i in range(len(x)):
        if x[i]>=0:
            y.append(lam * np.exp(-1*lam*x[i]))
        else:
            y.append(0)
    mean = 1/lam
    median = np.log(2)/lam
    mode = 0
    variance = 1/(np.power(lam,2))
    return y,mean,median,mode,variance

def exponential_cdf(x,lam):
    y = []
    for i in range(len(x)):
        if x[i]>=0:
            y.append(1-np.exp(-1*lam*x[i]))
        else:
            y.append(0)
    return y

x = np.linspace(0,10,100)
```

```python
# Gamma(k,theta)
def gamma_dist_pdf(x,k,theta):
    y =[]
    for i in range(len(x)):
        y.append((np.power(x[i],k-1)*np.exp(-1*x[i]/theta))/(np.power(theta,k)*np.math.factorial(k-1)))
    mean = k*theta
    median = None
    mode = (k-1)*theta
    variance = k*np.power(theta,2)
    return y,mean,median,mode,variance

def gamma_dist_cdf(x,k,theta):
    y = []
    for i in range(len(x)):
        tmp = 0
        for j in range(k-1):
            tmp = (1/np.math.factorial(j)) * np.power((x[i]/theta),j) * np.exp(-1/theta * x[i])
        y.append(tmp)
    return y

x = np.linspace(0,20,100)
```

```python
# Beta(alpha,beta)
def beta_dist(x,al,bt):
    y = []
    for i in range(len(x)):
        y.append((np.math.factorial(al+bt)/(np.math.factorial(al)*np.math.factorial(bt)))  * np.power(x[i],al-1) * np.power(1-x[i],bt-1))
    mean = al/(al+bt)
    median = (al-(1/3))/(al+bt-(2/3))
    mode = (al-1)/(al+bt-2)
    variance = (al*bt)/(np.power(al+bt,2)*(al+bt+1))
    return y,mean,median,mode,variance

x = np.linspace(-10,10,100)

# Could not find the CDF due to incomplete form
```

```python
# Gaussian(mu,sigma)
def gaussian_dist_pdf(x,mu,sigma):
    y = []
    for i in range(len(x)):
        y.append((1/(math.sqrt(2*np.pi) * sigma)) * np.exp(-1*(((x[i]-mu)/sigma)**2)))
    mean = median = mode = mu
    variance = np.power(sigma,2)
    return y,mean,median,mode,variance

def phi_x(t):
    s=0
    for i in range(-10,t):
        s += np.exp(-1/2 * np.power(t,2))
    return ((1/math.sqrt(2*np.pi)) * s)

def gaussian_dist_cdf(x,mu,sigma):
    y = []
    for i in range(len(x)):
        y.append((x[i]-mu)/sigma)
    return y

x = np.linspace(-10,10,100)
```

```python
# Chi-squared(k)
def chi_sq(x,k):
    y = []
    for i in range(len(x)):
        if(x[i]>0):
            y.append((np.power(x[i],(k/2)-1) * np.exp(-1/2*x[i]))/(np.power(2,k/2)*(math.sqrt(np.pi) * (np.math.factorial(k-2) / np.power(2, (k-1)/2)))))
        else:
            y.append(0)
    mean = k
    median = k*np.power((1-(2/(9*k))),3)
    mode = max(k-2,0)
    variance = 2*k
    return y,mean,median,mode,variance

# Could not find the CDF due to incomplete form

x = np.linspace(0,20,100)
```

```python
#bernoulli distribution(p,k)
def bernoulli_dist(prob,k):
    y = []
    for i in range(len(k)):
        y.append(np.power(prob,k[i])*np.power(1-prob,1-k[i]))
    mean = prob
    variance = prob*(1-prob)
    return y,mean,variance
```

```python
p = [0.1,0.5,0.8]
x = np.array([0,1])

fig = plt.figure(figsize=(8,15))

for i in range(len(p)):
    y_t,mu_t,var_t = bernoulli_dist(1-p[i],x)
    plt.subplot(len(p),1,i+1)
    plt.stem(x,y_t)
    plt.scatter(mu_t,0,c='r')
    plt.scatter(var_t,0,c='g')
    plt.xlabel('Success')
    plt.ylabel('PMF')
    plt.title("Bernoulli Distribution PMF with p(failure)="+str(p[i]))

plt.show()
```

```python
#Binomial distribution(n,p,k)
def binomial_pmf(n,p,k):
    y = []
    for i in range(len(k)):
        y.append(((np.math.factorial(n))/(np.math.factorial(k[i])*np.math.factorial(n-k[i])))*np.power(p,k[i])*np.power(1-p,n-k[i]))
    mean = n*p
    variance = n*p*(1-p)
    return y,mean,variance
```

```python
p = 0.4
x = np.arange(0,21)

fig = plt.figure()

y_t,mu_t,var_t = binomial_pmf(20,1-p,x)
plt.stem(x,y_t)
plt.scatter(mu_t,0,c='r')
plt.scatter(var_t,0,c='g')
plt.xlabel('Success')
plt.ylabel('PMF')
plt.title("Binomial Distribution PMF with p(failure)="+str(p))

plt.show()
```

```python
#Geometric distribution(p,k)
def geometric_pmf(p,k):
    y = []
    for i in range(len(k)):
        y.append(np.power(1-p,k[i]-1)*p)
    mean = (1-p)/p
    variance = (1-p)/np.power(p,2)
    return y,mean,variance
```

```python
p = 0.5
x = np.arange(0,10)

fig = plt.figure()

y_t,mu_t,var_t = geometric_pmf(p,x)
plt.stem(x,y_t)
plt.scatter(mu_t,0,c='r')
plt.scatter(var_t,0,c='g')
plt.xlabel('Success')
plt.ylabel('PMF')
plt.title("Geometric Distribution PMF with p(failure)="+str(p))

plt.show()
```

```python
#Poisson distribution(lambda,k)
def poisson_pmf(lam,k):
    y = []
    for i in range(len(k)):
        y.append(((np.power(lam,k[i])*np.exp(-1*k[i]))/np.math.factorial(k[i])))
    mean = lam
    variance = lam
    return y,mean,variance
```

```python
def delay(x,k):
    y = []
    for i in range(k):
        y.append(0)
    for i in range(k,len(x)):
        y.append(x[i-k])
    return y

def advance(x,k):
    y=[]
    for i in range(len(x)-k):
        y.append(x[i+k])
    for i in range(len(x)-k,len(x)):
        y.append(0)
    return y

def auto_correlation(seq):
    y=[]
    for i in range(len(seq)):
        seq_adv = advance(seq,i)
        sum = 0
        for j in range(len(seq)):
            sum += seq[j] * seq_adv[j]
        y.append(sum)
    y_r = []
    for i in range(len(y)-1):
        y_r.append(y[len(y)-i-1])
    return y_r+y
```

```python
x1 = np.array([1,2,3,2,1,2,3,4,3,2,1])
n = np.arange(-1*len(x1)+1,len(x1))

start = timeit.timeit()
y = auto_correlation(x1)
end = timeit.timeit()
print("Scratch: "+str(abs(end-start)))

start = timeit.timeit()
ys = np.correlate(x1,x1,'full')
end = timeit.timeit()
print("Numpy: "+str(abs(end-start)))

fig = plt.figure(figsize=(8,15))

plt.subplot(2,1,1)
plt.stem(n,y)
plt.scatter(0,0,c='g')
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.title('Autocorrelation function from scratch')
plt.subplot(2,1,2)
plt.stem(n,ys)
plt.scatter(0,0,c='g')
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.title('Autocorrelation function using numpy')

plt.show()
```

```python
#adding gaussian noise to image
def gauss_noise(image):
    r,c,ch = image.shape
    mean = 0
    sigma=1
    gauss = np.random.normal(mean,sigma,(r,c,ch))
    gauss = gauss.reshape(r,c,ch)
    img_n = image + gauss
    return np.array(img_n)

#adding salt & pepper noise
def salt_pepper_noise(image,amount):
    r,c,ch = image.shape
    img_n=np.copy(image)
    n_s = np.ceil(amount * image.size * 0.5)
    n_p = np.ceil(amount * image.size * 0.5)
    # Salt part
    coords = [np.random.randint(0, i - 1, int(n_s))
              for i in image.shape]
    img_n[coords] = 1
    # Pepper part
    coords = [np.random.randint(0, i - 1, int(n_p))
              for i in image.shape]
    img_n[coords] = 0
    
    return np.array(img_n)

#adding speckle noise
def speckle_noise(image):
    r,c,ch = image.shape
    gauss = np.random.randn(r,c,ch)
    gauss = gauss.reshape(r,c,ch)
    img_n = image + (image*gauss)
    return np.array(img_n)
```

```python
img = Image.open("jp1.jpg")

fig = plt.figure(figsize=(8,15))

plt.subplot(4,1,1)
plt.imshow(img)
plt.title('Original Image')

img_g = gauss_noise(np.array(img))
img_g = Image.fromarray(img_g.astype('uint8'), 'RGB')

plt.subplot(4,1,2)
plt.imshow(img_g)
plt.title('Gaussian Noise')

img_s = speckle_noise(np.array(img))
img_s = Image.fromarray(img_s.astype('uint8'), 'RGB')

plt.subplot(4,1,3)
plt.imshow(img_s)
plt.title('Speckle Noise')

img_sp = salt_pepper_noise(np.array(img),0.3)
img_sp = Image.fromarray(img_sp.astype('uint8'), 'RGB')

plt.subplot(4,1,4)
plt.imshow(img_sp)
plt.title('Salt & Pepper Noise')

plt.show()
```
