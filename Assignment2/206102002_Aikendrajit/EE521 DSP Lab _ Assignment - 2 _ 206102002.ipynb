```python
import numpy as np
import math
from scipy import stats
from collections import Counter
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def mean(sample_input):
    return sum(sample_input) / len(sample_input)

def median(sample_input):
    n = len(sample_input)
    index = n // 2
    # for odd no. of observations
    if n % 2:
        return sorted(sample_input)[index]
    # for even no. of observations
    return sum(sorted(sample_input)[index - 1:index + 1]) / 2

def mode(sample_input):
    c = Counter(sample_input)
    return [k for k, v in c.items() if v == c.most_common(1)[0][1]]

def variance(sample_input):
    m = sum(sample_input) / len(sample_input)
    var_v = sum((xi - m) ** 2 for xi in sample_input) / len(sample_input)
    return var_v
```

```python
def uniformpdf(t, in1, in2):
    p = ((t<=in2) & (t>=in1))
    q = (1/(in2 - in1))
    res = q*(p == 1)
    return res

n = np.linspace(-5,15,1000)
plt.figure(figsize=(10,5))
print("mean for the given samples are:", mean([2, 4]), mean([-2, 8]), mean([3, 8]))
print("median for the given samples are:", median([2, 4]), median([-2, 8]), median([3, 8]))
print("mode for the given samples are:", mode([2, 4]), mode([-2, 8]), mode([3, 8]))
print("variance for the given samples are:", variance([2, 4]), variance([-2, 8]), variance([3, 8]))

plt.plot(n,uniformpdf(n,2,4),label="For in1=2, in2=4")
#plt.plot(mean([-2, 8]), marker = "*", label="mean For in1=2, in2=4")
#plt.plot(median([3, -3.2]), marker = "*", label="median For in1=2, in2=4")
#plt.plot(mode([3, -3.2]), marker = "*", label="mode For in1=2, in2=4")
#plt.plot(variance([3, -3.2]), marker = "*", label="variance For in1=2, in2=4")
plt.plot(n,uniformpdf(n,-2, 8),label="For in1=-2, in2=8")
plt.plot(n,uniformpdf(n, 3, 8),label="For in1=3, in2=8")
plt.title("Pro. Den. Func. Of a Uniform Distribution")
plt.xlabel("in", Fontsize = 15)
plt.ylabel("pdf", Fontsize = 15)
plt.legend()
plt.grid()
plt.show()
```

```python
def exponentialpdf(x, lamda):
    d = []
    for i in x:
        if i >= 0:
            d.append(lamda * (np.exp(-lamda*i)))
        else:
            d.append(0)
    return d

def exponentialcdf(x, lamda):
    d = []
    for i in x:
        if i >= 0:
            d.append(1 - np.exp(-lamda*i))
        else:
            d.append(0)
    return d
```

```python
x = np.linspace(-5,15,1000)

plt.figure(figsize=(10,5))
plt.plot(x, exponentialpdf(x, 2),label="For Lambda = 2")
plt.plot(x, exponentialpdf(x,.3),label="For Lambda = .3")
plt.plot(x, exponentialpdf(x, 1.2),label="For Lambda = 1.2")
plt.title("Prob. Den. Func. Of an exponential Distribution")
plt.xlabel("in", Fontsize = 15)
plt.ylabel("pdf", Fontsize = 15)
plt.legend()
plt.grid()
plt.show()
```
