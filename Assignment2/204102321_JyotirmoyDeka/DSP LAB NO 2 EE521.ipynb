```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def uniform(x,a,b):
    v = [1/(b-a) if a <= i and i <= b else 0 for i in x]
    mean = 0.5*(a+b)
    median = 0.5*(a+b)
    variance = ((b-a)**2)/12
    return x,v,mean,median,variance
```

```python
x = np.arange(-100,100)
for j in [(-30,50),(-10,10),(40,90)]:
    a , b = j[0], j[1]
    mean = 0.5*(a+b)
    median = 0.5*(a+b)
    variance = ((b-a)**2)/12
    x,v,mean,median,variance = uniform(x,a,b)
    y = np.cumsum(v)
    print("the mean for a = "+str(a) +" and b =" + str(b) + " is: ", mean)
    print("the median for a = "+str(a) +" and b =" + str(b) + " is: ", median)
    print("the variance for a = "+str(a) +" and b =" + str(b) + " is: ", variance)
    plt.figure(figsize=(15,5))
    plt.subplot(2,2,1)
    plt.title("uniform pdf")
    plt.plot(x,v,label=r'a=%d,b=%d' % (a,b))
    plt.legend()
    plt.subplot(2,2,2)
    plt.title("uniform CDF")
    plt.plot(x,y,label=r'a=%d,b=%d' % (a,b))
    plt.legend()
```

```python
def exponential(x,l):
    y = l* np.exp(-l*x)
    mean = 1/l
    mode = 0
    median = np.log(2)/l
    variance = 1/(l**2)
    return x, y,mean,mode,median,variance
for i in [1, 10, 5]:
    t = np.arange(0, 20, 0.1)
    x, y,mean,mode,median,variance = exponential(t,i)
    print("the mode for lambda = "+str(i) +" is: ", mode)
    print("the mean for lambda = "+str(i) +" is: ", mean)
    print("the median for lambda = "+str(i) +" is: ", median)
    print("the variance for lambda = "+str(i) +" is: ", variance)
    plt.figure(figsize=(5,5))
    plt.subplot(2,2,1)
    plt.title("exponential PDF")
    plt.plot(t, y, label=r'$\lambda=%d$' % (i))
    plt.legend()
    plt.subplot(2,2,2)
    plt.title("exponential CDF")
    plt.plot(t,np.cumsum(y), label=r'$\lambda=%d$' % (i))
    plt.legend()
```

```python
def gamma_function(n):
    cal = 1
    for i in range(2, n):
        cal *= i
    return cal
def gamma(x, a, b):
    c = (b ** a) / gamma_function(a)
    y = c * (x ** (a - 1)) * np.exp(-b * x)
    mean = a/b
    variance = (a/b)**2
    mode = (a-1)/b
    return x, y,mean,variance,mode

def beta(x, a, b):
    gamma = gamma_function(a + b) /(gamma_function(a) * gamma_function(b))
    y = gamma * (x ** (a - 1)) * ((1 - x) ** (b - 1))
    mean = a/(a+b)
    median = (a-(1/3))/(a+b - (2/3))
    variance = a*b/(((a+b)**2)*(a+b+1))
    mode = (a-1)/(a+b-2)
    return x, y,mean,median,variance,mode 

def chi_squared(x, k):
    c = 1 / (2 ** (k/2)) * gamma_function(k//2)
    y = c * (x ** (k/2 - 1)) * np.exp(-x /2)
    mean = k
    var = 2*k
    median =k*pow((1-2/(9*k)),2)
    return x, y,mean,var,median
```

# The distribution here is bernoulli distribution

```python
def bernoulli(p, k):       ## bernoulli distribution
    return p if k else 1 - p

samples = 20
p1 = 0.5
x = np.arange(samples)
y = []
for i in range(samples):
    j = bernoulli(p1, k=bool(random.getrandbits(1)))
    y.append(j)

mean= np.mean(y)
variance = (np.std(y))**2
print("the mean for p = 0.1 is :",mean)
print("the variance for p = 0.1  is :",variance)
plt.scatter(x, y, label=r'$\mu=%.2f,\ \sigma=%.2f$' % (mean,variance))
plt.title("bernoulli for p = 0.5")
plt.legend()
```

```python
samples = 20
p = 0.8
p2 = 1 -p
x = np.arange(samples)
y = []
for i in range(samples):
    j = bernoulli(p2, k=bool(random.getrandbits(1)))
    y.append(j)

mean= np.mean(y)
variance = (np.std(y))**2
print("the mean for p = 0.8 is :",mean)
print("the variance for p = 0.8  is :",variance)
plt.scatter(x, y, label=r'$\mu=%.2f,\ \sigma=%.2f$' % (mean,variance))
plt.title("bernoulli for p = 0.8")
plt.legend()
```

# Here, the distribution is limited to n trials(from that of Bernoulli distribution) which is BINOMIAL DISTRIBUTION.

```python
import operator as op
from functools import reduce

def const(n, r):
    r = min(r, n-r)
    numerator = reduce(op.mul, range(n, n-r, -1), 1)
    denominator = reduce(op.mul, range(1, r+1), 1)
    return numerator / denominator

def binomial(n, p):
    q = 1 - p
    y = [const(n, k) * (p ** k) * (q ** (n-k)) for k in range(n)]
    mean = np.mean(y)
    variance = (np.std(y))**2
    return y,mean,variance

for ls in [(0.4, 20)]:
    p, samples = ls[0], ls[1]
    x = np.arange(samples)
    y,mean,variance = binomial(samples, p)
    print("the mean is ",mean)
    print("the variance is ",variance)
    plt.title("bionomial pmf")
    plt.scatter(x, y, label=r'$\mu=%.3f,\ \sigma=%.3f$' % (mean,variance))
plt.legend()
```

# Here,trials of bernoulli experiment (n) is very large, so the Binomial distribution gets approximated to a Poisson distribution.

```python
import operator as op
from functools import reduce

def const(n, r):
    r = min(r, n-r)
    numerator = reduce(op.mul, range(n, n-r, -1), 1)
    denominator = reduce(op.mul, range(1, r+1), 1)
    return numerator / denominator

def poisson(n, p):
    q = 1 - p
    y = [((const(n, k)*(p ** k) *np.exp (-p))/p)for k in range(n)]
    mean = np.mean(y)
    variance = np.sqrt(mean)
    return y,mean,variance

for ls in [(0.4, 20)]:
    p, samples = ls[0], ls[1]
    x = np.arange(samples)
    y,mean,variance = poisson(samples, p)
    print("the mean is ",mean)
    print("the variance is ",variance)
    plt.title("poisson pmf")
    plt.scatter(x, y, label=r'$\mu=%.3f,\ \sigma=%.3f$' % (mean,variance))
plt.legend()
```

```python
def acf(data):
    mean = np.mean(data)
    v = np.std(data)
    var = v**2
    length = len(data)

    acf_array = []
    for t in np.arange(0,length):
        temp = np.mean((data[:(length-t)] - mean)*(data[t:] - mean))/var
        acf_array.append(temp)

    acf_array = np.array(acf_array)
    return acf_array
a1 = acf(x1)
a2 = acf(x2)
a3 = acf(x3)
a4 = acf(x4)
a5 = acf(x5)
a6 = acf(x6)
```
