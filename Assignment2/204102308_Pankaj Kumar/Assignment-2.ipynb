# Autocorrelation

```python
x1 = [1,2,3,2,1,2,3,4,3,2,1]
x2 = [1,2,1,2,1,2,1,2]
x3 = [1,2,3,2,4,6,4,8,12]
x4 = [2,4,2,4,2,4,2,4]
x5 = [1,2,-3,2,-1,2,-3,4,-3,2,1]
x6 = [2*x1[i]+1 for i in x1]

def autocorr(a1,b1):
    acf = np.correlate(a1,b1,mode ='full')
    return acf
def soln(list):
    n1 =range(len(list))
    n2 =range(-len(list)+1,len(list))
    plt.subplot(121)
    plt.stem(n1,list,'r',use_line_collection=True)
    plt.grid('True')
    plt.title('list')
    plt.subplot(122)
    R1 =autocorr(list,list)
    plt.stem(n2,R1,'b',use_line_collection=True)
    plt.title('ACF')
    plt.grid('True')
    plt.show()
    print('ACF at lag =2:',R1[len(list)-1+2])
    print('power of signal=',R1[len(list)-1])
    return
soln(x1)
soln(x2)
soln(x3)
soln(x4)
soln(x5)
soln(x6)
```

# Crosscorrelation

```python
import time
def conv(x,h):
    m = len(x)
    n = len(h)
    y = [0 for i in range(n+m-1)]
    x = x + [0 for i in range(n)]
    h = h + [0 for i in range(m)]
    for i in range(n+m-1):
        for j in range(m):
            if i-j+1 >0:
                y[i] = y[i]+x[j]*h[i-j]
    return y
def autocorr2(a,b,lag):
    b = list(reversed((b)))
    acf = conv(a,b)
    if len(a)<len(b):
        return acf[len(a)-1+lag]
    if len(a)>=len(b):
        return acf[len(b)-1+lag]

def soln2(l1,l2):
    print('cross correlation at lag=2 samples b/w x1 =%s and x2= %s' %(l1,l2))
    t1 =time.time()
    out = autocorr(l1,l2)
    if len(l1)<len(l2):
        print('By inbuilt function',out[len(l1)-1+2])
    if len(l1)>=len(l2):
        print('By inbuilt function',out[len(l2)-1+2])
    t2 = time.time()
    print('Time taken :',1000*(t2-t1))
    print('By user defined function :',autocorr2(l1,l2, 2))
    t3 = time.time()
    print('Time taken :',1000*(t3-t2))
    print('\n\n')
    return None
soln2(x1,x2)
soln2(x1,x3)
soln2(x1,x4)
soln2(x1,x5)
soln2(x1,x6)
soln2(x2,x3)
soln2(x1,x4)
soln2(x1,x5)
soln2(x1,x6)
soln2(x3,x4)
soln2(x3,x5)
soln2(x3,x6)
soln2(x4,x5)
soln2(x4,x6)
soln2(x5,x6)
```
