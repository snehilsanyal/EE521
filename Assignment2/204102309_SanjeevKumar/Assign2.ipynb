```
import numpy as np
from math import factorial
import scipy.stats as sp
import matplotlib.pyplot as plt
t = np.linspace(-8,8,100)
#Uniform distribution
def uniformpdf(a,b):
    return 1/(b-a)  * (t>=a) * (t<=b)
def uniformcdf(a,b):
  y=[]
  
  for i in t:
    if i<a:
      y.append(0)
    elif i<=b and i>=a:
      y.append((i-a)/(b-a))
    else:
      y.append(1)
  return y
plt.plot(t,uniformpdf(-4,5),'r',label = 'a= %d b=%d' %(-4,4))
plt.plot(t,uniformpdf(-1,5),'b',label = 'a= %d b=%d' %(-1,5))
plt.plot(t,uniformpdf(-2,3),'g',label = 'a= %d b=%d' %(-2,3))
plt.title('Uniform distribution pdf')
plt.legend()
plt.show()
plt.plot(t,uniformcdf(-4,5),'r',label = 'a= %d b=%d' %(-4,4))
plt.plot(t,uniformcdf(-1,5),'b',label = 'a= %d b=%d' %(-1,5))
plt.plot(t,uniformcdf(-2,3),'g',label = 'a= %d b=%d' %(-2,3))
plt.title('Uniform distribution cdf')
plt.legend()
plt.show()
```

```
#Exponential distribution
def expopdf(k):
    return k * np.e**(-k*t) * (t>0)
def expocdf(k):
    return (1-np.e**(-k*t)) * (t>0)
plt.plot(t,expopdf(0.5),'r',label = 'lambda = %d' %0.5)
plt.plot(t,expopdf(1),'b',label = 'lambda = %d' %1)
plt.plot(t,expopdf(2),'g',label = 'lambda = %d' %2)
plt.title('Exponential distribution pdf')
plt.legend()
plt.show()
plt.plot(t,expocdf(0.5),'r',label = 'lambda = %d' %0.5)
plt.plot(t,expocdf(1),'b',label = 'lambda = %d' %1)
plt.plot(t,expocdf(2),'g',label = 'lambda = %d' %2)
plt.title('Exponential distribution cdf')
plt.legend()
plt.show()
```

```
#Gamma distribution
def gammapdf(a,b):
    return ((1/b)**a * t**(a-1) * np.e**(-(1/b)*t))/factorial((a-1)) * (t>=0)
plt.plot(t,gammapdf(1,2),'r',label = 'a= %d b=%d' %(1,2))
plt.plot(t,gammapdf(2,2),'b',label = 'a= %d b=%d' %(2,2))
plt.plot(t,gammapdf(9,0.5),'g',label = 'a= %d b=%d' %(9,0.5))
plt.title('Gamma distribution pdf')
plt.legend()
plt.show()
# I am importing the inbulit functions now due to complexity of the further distributions
def gammacdf(a,b):
  return sp.gamma.cdf(t,a,0,b)
plt.plot(t,gammacdf(1,2),'r',label = 'a= %d b=%d' %(1,2))
plt.plot(t,gammacdf(2,2),'b',label = 'a= %d b=%d' %(2,2))
plt.plot(t,gammacdf(9,0.5),'g',label = 'a= %d b=%d' %(9,0.5))
plt.title('Gamma distribution cdf')
plt.legend()
plt.show()
```

```
#Beta Distribution
t = np.linspace(0,1,100)
def betapdf(a,b):
    return sp.beta.pdf(t,a,b)
plt.plot(t,betapdf(5,1),'r',label = 'a= %d b=%d' %(5,1))
plt.plot(t,betapdf(2,2),'b',label = 'a= %d b=%d' %(2,2))
plt.plot(t,betapdf(1,3),'g',label = 'a= %d b=%d' %(1,3))
plt.title('beta distribution pdf')
plt.legend()
plt.show()

def betacdf(a,b):
  return sp.beta.cdf(t,a,b)
plt.plot(t,betacdf(5,1),'r',label = 'a= %d b=%d' %(5,1))
plt.plot(t,betacdf(2,2),'b',label = 'a= %d b=%d' %(2,2))
plt.plot(t,betacdf(1,3),'g',label = 'a= %d b=%d' %(1,3))
plt.title('beta distribution cdf')
plt.legend()
plt.show()
```

```
#Gaussian Distribution
t = np.linspace(-5,5,100)
def Gaussianpdf(a,b):
    return sp.norm.pdf(t,a,b)
plt.plot(t,Gaussianpdf(0,0.3),'r',label = 'a= %d b=%f' %(0,0.3))
plt.plot(t,Gaussianpdf(-1,0.2),'b',label = 'a= %d b=%f' %(-1,0.2))
plt.plot(t,Gaussianpdf(1,1),'g',label = 'a= %d b=%d' %(1,1))
plt.title('Gaussian distribution pdf')
plt.legend()
plt.show()

def Gaussiancdf(a,b):
  return sp.norm.cdf(t,a,b)
plt.plot(t,Gaussiancdf(0,0.3),'r',label = 'a= %d b=%f' %(0,0.3))
plt.plot(t,Gaussiancdf(-1,0.2),'b',label = 'a= %d b=%f' %(-1,0.2))
plt.plot(t,Gaussiancdf(1,1),'g',label = 'a= %d b=%d' %(1,1))
plt.title('Gaussian distribution cdf')
plt.legend()
plt.show()
#Standard Normal Distribution
plt.plot(t,sp.norm.pdf(t,0,1),'k')
plt.title('Standard Normal Distribution')
```

```
#Chi-Square Distribution
t = np.linspace(0,10,100)
def Chi2pdf(k): # k is called Degree of Freedom
    return sp.chi2.pdf(t,k)
plt.plot(t,Chi2pdf(2),'r',label = 'k = %d' %2)
plt.plot(t,Chi2pdf(3),'b',label = 'k = %d' %3)
plt.plot(t,Chi2pdf(4),'g',label = 'k = %d' %4)
plt.title('Chi-Square distribution pdf, k= Dof')
plt.legend()
plt.show()

def Chi2cdf(k): 
  return sp.chi2.cdf(t,k)
plt.plot(t,Chi2cdf(2),'r',label = 'k = %d' %2)
plt.plot(t,Chi2cdf(3),'b',label = 'k = %d' %3)
plt.plot(t,Chi2cdf(4),'g',label = 'k = %d' %4)
plt.title('Chi-Square distribution cdf')
plt.legend()
plt.show()
```

```
# Mean,  Median and Mode for Exponential distribution
t = np.linspace(0,10,100)
plt.subplot(211)
plt.plot(t,expopdf(0.5),'r',label = 'lambda = %d' %0.5)
plt.plot(2,0,'ro',label = 'mean') # mean = 1/lambda
 # Median is the value where cdf value of the given distribution is 0.5
def median(cdf):
  for i in cdf:
      if 0.49<i<0.51:
          return t[list(cdf).index(i)]
plt.plot(median(expocdf(0.5)),0,'bo' , label ='median')
def mode(pdf):
  g = list(pdf)
  return (t[g.index(max(g))])     # Mode is the value corresponding to the maximum value in the pdf
plt.plot(mode(expopdf(0.5)),0,'go' , label ='mode')
plt.title('Exponential distribution')
plt.legend()
plt.show()
```

```
n = np.linspace(0,50,51)
def poissonD(mu,k):
    pmf = []
    for i in k:
        pmf.append(np.e**(-mu)* (mu**i)/np.math.factorial(i))
    return pmf
plt.plot(n,poissonD(5,n))
plt.title('PMF of X(Prob(no. of bits)')
plt.xlabel('no. of bits')
plt.ylabel('Prob')
plt.show()
```

```
# CROSS CORRELATIONS
import time
def conv(x,h):
    m = len(x)
    n = len(h)
    y = [0 for i in range(n+m-1)]
    x = x + [0 for i in range(n)]
    h = h + [0 for i in range(m)]
    for i in range(n+m-1):
        for j in range(m):
            if i-j+1>0:
                y[i] = y[i]+x[j]*h[i-j]
    return y
def autocorr(a,b,lag):
    b = list(reversed((b)))
    acf = conv(a,b)
    if len(a)<len(b):
        return acf[len(a)-1+lag]
    if len(a)>=len(b):
        return acf[len(b)-1+lag]

def soln2(l1,l2):  
  print('Cross correlation at lag = 2 samples between x1 = %s and x2 = %s' %(l1,l2))
  t1 = time.time()
  out = inbltacf(l1,l2)
  if len(l1)<len(l2):
    print('By inbuilt fn' , out[len(l1)-1+2])
  if len(l1)>=len(l2):
    print('By inbuilt fn' , out[len(l2)-1+2])
  t2 = time.time()
  print('Time taken :' , 1000*(t2 - t1))
  print('by self defined fn :',autocorr(l1,l2,2))
  t3 = time.time()
  print('Time taken :' , 1000*(t3-t2))
  print('\n\n')
  return None
soln2(x1,x2)
soln2(x1,x3)
soln2(x1,x4)
soln2(x1,x5)
soln2(x1,x6)
soln2(x2,x3)
soln2(x2,x4)
soln2(x2,x5)
soln2(x2,x6)
soln2(x3,x4)
soln2(x3,x5)
soln2(x3,x6)
soln2(x4,x5)
soln2(x4,x6)
soln2(x5,x6)
```

**IMAGE PROCESSING**

```
import cv2
import numpy as np
import random
import matplotlib.pyplot as plt
image = cv2.imread('/content/sample.jpeg',1)
image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.imshow(image)
plt.title('Original image')
plt.show()
```

```
#pepper and salt noise
rows,columns,t = image.shape
p = 0.05 # Spraying 5 percent of proabibilty of the image pixels
out = np.zeros(image.shape,np.uint8)
for i in range(rows):
    for j in range(columns):
        r = random.random()
        if r<p/2:
            out[i][j] = [0,0,0]   # pepper sprinkle
        elif r<p:
            out[i][j] = [255,255,255]   # salt sprinkle
        else:
            out[i][j] = image[i][j]
plt.imshow(out)
plt.show()
```

```
# GAUSSIAN NOISE addition
out = np.zeros(image.shape,np.uint8)
for i in range(rows):
    for j in range(columns):

        for k in range(t):
            r = np.random.normal(0, 20) # mean be 0 and variance be 20

            out[i][j][k] = image[i][j][k] + r

plt.imshow(out)
plt.show()
```

Speckle noise , Salt and pepper noise can be filtered by mean and median filter where as Gaussian can be filtered by DoG filters.
