```python
import matplotlib.pyplot as plt  
import numpy as np  
import random

def uniform(x,a,b):
    pdf =[]
    for t in x:
        if t>=a and t<=b:
            pdf.append(1)
        else:
            pdf.append(0)
            
    cdf = np.cumsum(pdf)
    
    return pdf,cdf


t1=np.linspace(-3,10,100)
lst = [uniform(t1,-2,2),uniform(t1,3,8),uniform(t1,4,8)]
p=['pdf','cdf']
fig,ax=plt.subplots(2,3,figsize=(8,6))
c=1
for i in range(len(lst)):
    for j in range(0,2): 
            plt.subplot(2,3,c)
            plt.plot(t1,lst[i][j])
            plt.title(str(p[j])+'plot')
            c=c+1
       
```

### BETA DISTRIBUTION

```python
import matplotlib.pyplot as plt  
import numpy as np  
import random
import math

def beta(x,alpha,bet):
    pdf =[]
    ga=math.gamma(alpha)
    gb=math.gamma(bet)
    gc=math.gamma(alpha+bet)
    den=(ga*gb)/gc
    for t in x:
        r=(t**(alpha-1))*((1-t)**(bet-1))/den
        pdf.append(r)
        
            
            
    cdf = np.cumsum(pdf)
    
    return pdf,cdf


t1=np.linspace(0,1,100)
lst = [beta(t1,1,3),beta(t1,5,1),beta(t1,2,5)]
p=['pdf','cdf']
c0=['black','blue','red']
fig,ax=plt.subplots(2,3,figsize=(8,6))
c=1
for i in range(len(lst)):
    for j in range(0,2): 
            plt.subplot(2,3,c)
            plt.plot(t1,lst[i][j])
           
            plt.title(str(p[j])+'plot')
            c=c+1
```

PART 1 - Consider a transmitter that can transmit messages only in the form of 2 bits 0 and 1. The receiver receives a single message sent by the transmitter say X. Since, we do not know before hand which bit was sent X follows a certain distribution. Which distribution is this? Find the PMF and plot it. Also calculate the mean and variance of X. Assume probability for receiving 0(failure) is p = 0.1, 0.5, 0.8.

```python
import numpy as np 
import matplotlib.pyplot as plt 

def impulse(a, n,p): 
    delta =[] 
    for sample in n: 
        if sample == 0: 
            delta.append(p) 
        elif sample==1: 
            delta.append(1-p) 
            
        else:
            delta.append(0)
              
    return delta 

n=np.arange(-3,7,1)
prob=[0.1,0.5,0.8]
fig,ax=plt.subplots(3,1,figsize=(20,10))
for j,e in enumerate(prob,start=1):
    for i in range(2):
        res = impulse(i,n,e)
        plt.subplot(3,1,j)
        plt.stem(n,res,use_line_collection=True)
```

In the problem discussed above ,we have plotted the PMF for given three different values of P(probability of error) which is equivalent to receiving a zero at the output .So the PMF would be 2 impulses one at zero with amplitude equal to probability of error and other at one with amplitude = 1-probability of error.
This follows a bernoulli distribution .
Mean = p = probability of success
Variance=pq

for 1st case , mean=0.9,variance=0.09
for 2nd case , mean=0.5,variance=0.25
for 1st case , mean=0.2,variance=0.16

PART 2-Consider the same transmitter but this time the receiver receives 20 bits. Each bit is independent of the previous one. The probability of receiving a 0 is p = 0.4. Let X be the number of 1s (successes) that is received at the receiver’s end. What is the distribution followed by X. How is it related to the first distribution? Calculate the mean and variance and plot the PMF.

```python
import matplotlib.pyplot as plt  
import numpy as np  
import random
import math

def nCr(n,r):
    
    return (fact(n)/(fact(r)*fact(n-r)))

def fact(n):
    
    res = 1
    for i in range(2,n+1):
        res = res * i
        
    return res

def binomial(t,n,p):
    pdf =[]
    
    for j in t:
        res = nCr(n,j)
        mul= ((1-p)**j)*(p**(n-j))
        bi=res*mul
        pdf.append(bi)
            
    return pdf


t=np.arange(5,20,1)
res = binomial(t,20,0.4)
plt.stem(t,res)
plt.title('PMF')
```

The PMF of X(success)= Receiving 1, follows binomial distribution that is nCr(1-p)^rp^(n-r). For varying values of X we have plotted a PMF of binomial distribution. Mean = np = 200 into 6=12 Variance= npq = 200 into 60 into 4=4.8

PART 3-Consider an experiment in which bits are transmitted sequentially. Each bit is independent of the other. When the first bit arrives the receiver checks for a 1. If it is 1 the receiver takes a decision. If it is a 0 the receiver waits for the next bit. This goes on till the receiver finds out that the bit was 1 and the experiment is dropped once the bit is 0. Let X denote that the bit received was 1 on the k th time (example, X=2 suggests that the bit sequence was 0, 1 and once 1 was received the experiment was stopped). Find the PMF of X. What is this distribution known as? Calculate the mean and variance and plot the PMF. How is this distribution related to the first distribution

```python
import matplotlib.pyplot as plt  
import numpy as np  
import random
import math

def geometric(n,X,p):
    pdf =[]
    for t in n:
        if (t==X):
            pdf.append(1-p)
            
        else:
            pdf.append(p)
        
        
            
    return pdf


n=np.arange(1,11,1)
res = geometric(n,8,0.3)
plt.stem(n,res)
```

This is GEOMETRIC DISTRIBUTION , as getting first 1 is considered as success and the experiment is stopped. So using this distribution we find the probability of finding first success at Xth position and remaining X-1 are the zeroes that is the failure.
Mean = 1/p =1/0.7=1.42
Variance = 1-p/p^2 =0.6122

PART 4-Consider an experiment in which the receiver receives different bits randomly from many transmitters within a second’s interval. Each second it was observed that each second (at any interval) the mean number of bits that were received was 5. Let X be the number of bits received every second (during an interval). What is the distribution that is followed by X. Is there any relation between this and the second distribution that we discussed? Calculate the mean, variance and plot the PMF.

```python
import matplotlib.pyplot as plt  
import numpy as np  
import random
import math


def fact(n):
    
    res = 1
    for i in range(2,n+1):
        res = res * i
        
    return res


def Poisson(n,lam):
    pdf =[]
    for t in n:
        res=(lam**t)*np.exp(-lam)/fact(t)
        pdf.append(res)
        
            
    return pdf


n=np.arange(1,11,1)
res = Poisson(n,5)
plt.stem(n,res)
```

The above distriution follows Poisson Distribution. Here the given mean = 5 serves as a value for lambda which is a parameter
in poisson distribution.
Mean = lambda =5
Variance=lambda =5

## AUTOCORRELATION FUNCTION

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import time

tic=time.time()

def mean1(x):
    sum=0
    for i in range(len(x)):
        sum = sum + x[i]
    mean = sum/len(x)
    return mean
    
def variance(x):
    sum=0
    n = len(x)
    me=mean(x)
    for i in range(len(x)):
        dev = (x[i] - me) ** 2 
        sum=sum+ dev
    
    var = sum/n
    
    return var  


def autocorr(s1,s2,delay):
    mean=np.mean(s1)
    var=variance(s1)
    xp=s1-mean
    res = range(len(s1))
    corr=[1. if l==0 else np.sum(xp[l:]*xp[:-l])/len(s1)/var for l in res]

    return np.array(corr)
       
toc=time.time()
elapsed  = toc-tic
print("ELASPED TIME-",elapsed)
```

### CROSS CORRELATION FUNCTION

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

tic=time.time()
def mean(x):
    sum=0
    for i in range(len(x)):
        sum = sum + x[i]
    mean = sum/len(x)
    return mean
    
def variance(x):
    sum=0
    n = len(x)
    me=mean(x)
    for i in range(len(x)):
        dev = (x[i] - me) ** 2 
        sum=sum+ dev
    
    var = sum/n
    
    return var  


def crosscorr(s1,s2,delay):
    xp=[]
    yp=[]
    corr = []
    me=mean(s1)
    var=variance(s1)
    for i in s1:
        xp.append(i-me)
    del s2[-delay:]
    s1=np.pad(s2,(delay,0),'constant')
    
    sub = len(s1)-delay
    
    for i in s2:
        yp.append(i-me)
    for l in range(delay,sub):
        corr.append((xp[l]*yp[l])/len(s1)/var)
        
    return corr

toc=time.time()
elapsed  = toc-tic
print('ELASPED TIME-',elapsed)

```

### Generate a random sinusoid and a noisy sinusoid

```python
import numpy as np
import matplotlib.pyplot as plt
from random import gauss
from random import seed
from pandas import Series
from pandas.plotting import autocorrelation_plot
import matplotlib.pyplot as plt

w=np.pi/8
def unit_step(a,b, n): 
    unit =[] 
    for sample in n: 
        if sample<a and sample>b: 
            unit.append(0) 
        else: 
            unit.append(1) 
    return(unit)

n =  np.linspace(-1, 20, 100)
A = unit_step(0,4, n)
theta =unit_step(0,2*np.pi, n)
x=A*np.cos(w*n + theta)
noise = np.random.normal(0, 1, 100)
noisy=x+noise
plt.subplot(2,1,1)
plt.title('Random Sinusoid')
plt.stem(n,x,use_line_collection=True)
plt.subplot(2,1,2)
plt.title('Noisy Sinusoid')
plt.stem(n,noisy,use_line_collection=True)
plt.tight_layout()
```
