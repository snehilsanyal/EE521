# (a) Uniform (a,b)

```python
import numpy as np 
import matplotlib.pyplot as plt
import random
t=np.linspace(-1,6,1000)
def uniform(a,b):
    
    y=[]
    mean=(a+b)/2
    median=(a+b)/2
    std=(a+b)/np.sqrt(12)
    mode=random.randrange(a,b)
    for i in t:
        if i>=a and i<=b:
            y.append(1/(b-a))
        else :
            y.append(0)
    return y,mean,median,std ,mode 

L=[uniform(0,1), uniform(1,3),uniform(0,5)]
mark=['o','^','*','+']
alist=['mean','median','std']
c=1
fig, axs = plt.subplots(2, 3, figsize=(15,10))
for i in range(len(L)):
    plt.subplot(3,2,c)
    plt.plot(t,L[i][0])
    plt.title('PDF')
    for j in range(0,3):
        plt.plot(L[i][j+1],marker=mark[j],ms=15,label=alist[j])
    c=c+1
    plt.subplot(3,2,c)
    plt.plot(t,np.cumsum(L[i][0]))
    plt.title('CDF')
    
   
    c=c+1
```

# (b) Exponential (λ)

```python
import numpy as np 
import matplotlib.pyplot as plt
t=np.linspace(0,10,1000)
fig, axs = plt.subplots(2, 3, figsize=(15,10))
def expo(lamdas):
    c=1
    for lamda in lamdas:
        y=lamda*np.exp(-lamda*t)
        mean=lamda
        median=np.log(2)/lamda
        mode=0
        alist=[mean,median,mode]
        name=['mean','median','std']
        mark=['o','^','*']
        plt.subplot(3,2,c)
        plt.plot(t,y)
        plt.title('pdf for lambda='+str(lamda))
        for j in range(len(alist)):
            plt.plot(alist[j],marker=mark[j],ms=15,label=name[j])
        plt.legend(loc='upper right')
        c=c+1
        plt.subplot(3,2,c)
        cdf=np.cumsum(y)
        plt.plot(t,cdf)
        plt.title('cdf for lambda='+str(lamda))
        c=c+1
    
lamdas=[0.5,1,1.5]  
expo(lamdas)
```

# (c) Gamma (γ)

```python
import numpy as np 
import matplotlib.pyplot as plt
import math
fig, axs = plt.subplots(2, 3, figsize=(15,10))
def gamma(parameter):
    c=1
    for (alpha,lamda) in  parameter:
        t=np.linspace(0,10,1000)
        mean=alpha/lamda
        mode=(alpha-1)/lamda
        std=math.sqrt(alpha/lamda**2)
        alist=[mean,mode,std]
        name=['mean','mode','std']
        mark=['o','^','*']
        y=((lamda**alpha)*(t**(alpha-1))*np.exp(-lamda*t))/(math.gamma(alpha))
        plt.subplot(3,2,c)
        plt.plot(t,y)
        plt.title('pdf for alpha='+str(alpha)+',lambda='+str(lamda))
        for j in range(len(alist)):
            plt.plot(alist[j],marker=mark[j],ms=15,label=name[j])
        plt.legend(loc='upper right')
        c=c+1
        plt.subplot(3,2,c)
        cdf=np.cumsum(y)
        plt.plot(t,cdf,c='r')
        plt.title('cdf for alpha='+str(alpha)+',lambda='+str(lamda))
        c=c+1
alist=[(1,2),(5,1),(2,1)]        
gamma(alist)        
        
```

# (d) Beta (β)

```python
import numpy as np 
import matplotlib.pyplot as plt
import math
fig, axs = plt.subplots(2, 3, figsize=(15,10))

def beta(parameters):
    t=np.linspace(0,1,1000)
    c=1
    for (a,b) in parameters:
        y=((t**(a-1))*((1-t)**(b-1)))/(math.gamma(a)*math.gamma(b))/(math.gamma(a+b))
        mean=a/(a+b)
        median= (a-1/3)/(a+b-2/3)
        std=math.sqrt(a*b/(((a+b)**2)*(a+b+1)))
        mode=(a-1)/(a+b-2)
        alist=[mean,median,mode,std]
        name=['mean','median','mode','std']
        mark=['o','^','*','x']
        plt.subplot(3,2,c)
        plt.plot(t,y)
        plt.title('pdf for a='+str(a)+',b='+str(b))
        for j in range(len(alist)):
            plt.plot(alist[j],marker=mark[j],ms=15,label=name[j])
        plt.legend(loc='upper right')
        c=c+1
        plt.subplot(3,2,c)
        cdf=np.cumsum(y)
        plt.plot(t,cdf,c='r')
        plt.title('cdf for a='+str(a)+',lambda='+str(b))
        c=c+1
       
    
para=[(2,5),(5,1),(2,2)]    
beta(para)
        
    
    
```

# (e) Gaussian (µ,σ)

```python
import numpy as np 
import matplotlib.pyplot as plt
import math
fig, axs = plt.subplots(2, 3, figsize=(20,15))

def normal(parameters):
    t=np.linspace(-6,6,1000)
    c=1
    for (mean,sigma) in parameters:
        y=np.exp(-((t-mean)/sigma)**2)/(sigma*math.sqrt(2*np.pi))
        means=mean
        median=mean
        std=sigma
        mode=mean
        alist=[means,median,mode,std]
        name=['mean','median','mode','std']
        mark=['o','^','*','x']
        plt.subplot(3,2,c)
        plt.plot(t,y)
        plt.title('pdf for mean='+str(mean)+',std='+str(sigma))
        plt.axhline(0,color='b',lineWidth=0.5,c='k')
        plt.axvline(0,color='b',lineWidth=0.5,c='k')
        for j in range(len(alist)):
            plt.plot(alist[j],marker=mark[j],ms=15,label=name[j])
        plt.legend(loc='upper right')
        c=c+1
        plt.subplot(3,2,c)
        cdf=np.cumsum(y)
        plt.plot(t,cdf,c='r')
        plt.title('cdf for mean='+str(mean)+',std='+str(sigma))
        plt.axhline(0,color='b',lineWidth=0.5,c='k')
        plt.axvline(0,color='b',lineWidth=0.5,c='k')
        c=c+1
       
    
para=[(0,0.5),(0,1),(-1,1)]    
normal(para)
        
  
```

# (f) Standard Normal (0, 1)

# (g) Chi-squared (k)

```python
import numpy as np 
import matplotlib.pyplot as plt
import math
fig, axs = plt.subplots(2, 3, figsize=(15,10))

def Chi_square(parameters):
    t=np.linspace(0,8,1000)
    c=1
    for k in parameters:
        y=((t**((k/2)-1))*np.exp(-(t/2)))/((2**(k/2))*math.gamma(k/2))
        mean=k
        median=k*(1-2/(9*k))**3
        std=math.sqrt(2*k)
        mode=max(k-2,0)
        alist=[mean,median,mode,std]
        name=['mean','median','mode','std']
        mark=['o','^','*','x']
        plt.subplot(3,2,c)
        plt.plot(t,y)
        plt.title('pdf for K='+str(k))
        plt.ylim(0,k+0.5)
        for j in range(len(alist)):
            plt.plot(alist[j],marker=mark[j],ms=15,label=name[j])
        plt.legend(loc='upper right')
        
        c=c+1
        plt.subplot(3,2,c)
        cdf=np.cumsum(y)
        plt.plot(t,cdf,c='r')
        plt.title('cdf for K='+str(k))
        c=c+1
       
    
para=[2,2.5,3]    
Chi_square(para)
        
  
```

# Discrete distribution

## Bernoulli distribution

```python
import numpy as np 
import matplotlib.pyplot as plt
x=[0,1]
def bernouli(p):
    pmf=[]
    for i in x:
        if (i==0):
            pmf.append((1-p))
        if i==1:
            pmf.append(p)
        
    return pmf 

P=[0.1,0.5,0.8]
c=1
fig, axs = plt.subplots(1,1, figsize=(10,5))
for i in P:
    a=bernouli(i)
    plt.subplot(1,3,c)
    plt.bar(x,a)
    plt.title('PMF OF BERNOULLI  p='+str(i))
    print('mean(p) ='+str(i))
    print('variance(pq) ='+str(i*(1-i)))
    c=c+1
```

# Binomial distribution

```python
import numpy as np 
import matplotlib.pyplot as plt
import math

n=20
r=np.arange(0,n+1)
def binomial(n,p):
    PMF=[]
    for r in range(n+1):
        y=(math.factorial(n)/(math.factorial(n-r)*math.factorial(r)))*(p**r)*((1-p)**(n-r))
        PMF.append(y)
        
    return PMF 

p=0.4
A=binomial(20,p)
plt.stem(r,A)
plt.title('binomial pmf, n=20,p=0.4')
print('mean=np='+str(n*p))
print('variance=npq='+str(n*p*(1-p)))
```

# Geometric distribution

```python
import numpy as np 
import matplotlib.pyplot as plt
import math
def Geometric(p):
        PMF=[]
        x=[]
        for k in range(7):
            if p>=0.01:
                y=((1-p)**(k-1))*p
                PMF.append(y)
                x.append(k)
                
                
        return PMF ,x 
P=0.5       
a=Geometric(P) 
plt.stem(a[1],a[0],use_line_collection=True)
plt.title('pmf of geometric p=0.5')
print('mean='+str(1/P))
print('variance='+str((1-P)/P**2))
```

# Poisson distribution

```python
import numpy as np 
import matplotlib.pyplot as plt
import math
def poisson(lamda):
    PMF=[]
    for k in range(0,25):
        y=((lamda**k)*(np.exp(-lamda)))/math.factorial(k)
        PMF.append(y)
        
    return PMF
lamda=5
out=poisson(lamda)
x=np.arange(0,25)
plt.stem(x,out,use_line_collection=True)
plt.title('poisson distribution')
print('mean='+str(lamda))
print('variance='+str(lamda))
```

# Auto correlation

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import statistics
import random
import math
import pandas as pd
import statsmodels.api as sm
import warnings
warnings.filterwarnings("ignore")

def autocorrelation(x,y,delay):
    n=len(x)
    m=len(y)

    mean1=sum(x)/n
    mean2=sum(y)/m
    if (n-m) !=0 :
        if n<m:
            x = np.pad(x, (abs(n-m),0),'constant')
        
        else:
            y = np.pad(y,(abs(n-m),0),'constant')
    auto_cofficent=[]
    summer1=0
    summer2=0
    n1=len(x)
    sum1=[]
    sum2=[]
    for i in range(n1):
        x_diff = (x[i]-mean1)**2
        y_diff = (y[i]-mean2)**2
        sum1.append(x_diff)
        sum2.append(y_diff)
    k1 = sum(sum1)
    k2 = sum(sum2)
    a = np.sqrt(k1)
    b = np.sqrt(k2)
    summer = a*b
    
    for k in range(0,delay+1):
        summer1=0
        summer2=0
        for i in range(n1-k):
            sum1 =((x[i]-mean1)*(y[i+k]-mean2))
            summer1=summer1+sum1
        
        rk=summer1/summer
        auto_cofficent.append(rk)
    return  auto_cofficent  
        
```

```python
x1 = [1, 2, 3, 2, 1, 2, 3, 4, 3, 2, 1]
x2= [1, 2, 1, 2, 1, 2, 1, 2]
x3 = [1, 2, 3, 2, 4, 6, 4, 8, 12]
x4 = [2, 4, 2, 4, 2, 4, 2, 4]
x5 =  [1, -2, 3, -2, 1, -2, 3, -4, 3, -2, 1] 
x6 =[2*x1[n]+1 for n in range(len(x1))]  

x = [x1,x2,x3,x4,x5,x6]
```

```python
inbulid_function =[sm.tsa.acf(x[i], nlags=2) for i in range(len(x))]
a = [autocorrelation(x[i],x[i],2) for i in range(len(x))]
for i in range(len(a)):
    print(str([i+1])+" auto correlation using inbulit library :", str(inbulid_function[i]))
    print(str([i+1])+" auto correlation using user defined function:", str(a[i]))
```
