```python
'''Noisy sinusoid'''
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
import random

random.seed(1)
noise = [random.gauss(0.0,1.0) for t in range(100) ]
noise=pd.Series(noise)
noise.plot()
plt.title('gausian noise with mean=0 and sigma=1 ')
plt.show()
n=np.linspace(-10,10,100)
signal=[5*np.sin(2*np.pi*t) for t in n]
x=[noise[i]+signal[i] for i in range(100)]
plt.plot(n,x)
plt.plot(n,x)
plt.plot(n,signal)
plt.title('noisy sinusoid')
plt.grid()
plt.show()
```

```python
import numpy as np 
import matplotlib.pyplot as plt
import math
'''GAMMA DISTRIBUTION'''

def gamma(parameter):
    i = 1
    for (alpha,beta) in parameter:
        t = np.linspace(0,10,100)
        y = (beta**alpha)*(t**(alpha-1)*np.exp(-beta*t))/(math.gamma(alpha))
        mean = alpha/beta
        mode = (alpha-1)/beta
        varr = alpha/beta**2
        print("mean is:",mean)
        print("mode is :",mode)
        print("variance is:",varr)
        
        plt.subplot(3,2,i)
        plt.plot(t,y)
        plt.title('pdf')
        plt.plot(mean,marker="^",label = "mean")
        plt.plot(mode,marker="*",label= 'median')
        plt.plot(varr,marker="o",label = "variance")
        i = i+1
        plt.subplot(3,2,i)
        y1 = np.cumsum(y)
        plt.title('cdf')
        plt.plot(t,y1)
        i = i+1
alist= [(1,2),(1,5),(6,5)]

fig, axs  = plt.subplots(2,3,figsize=(10,10))
gamma(alist)
```

```python
''' Gaussian  Distribution '''
import math
import numpy as np
import matplotlib.pyplot as plt
def gauss(parameters):
    x = np.linspace(-10,10,1000)
    i = 1
    for (mean , sigma) in parameters:
        y = np.exp(-((x-mean)/sigma)**2)/(sigma*math.sqrt(2*np.pi))
        plt.subplot(3,2,i)
        plt.plot(x,y)
        plt.title('pdf')
        mean = mean
        mode = mean 
        median = mean
        print('mean =median=mode is:',mean)
        var = sigma**2
        print('variance is:',var)
        plt.plot(mean,marker="^",label = "mean")
        plt.plot(var,marker="o",label = "variance")
        plt.plot(mode,marker="*",label="mode")
        plt.plot(median,marker=">",label="median")
        i = i+1
        plt.subplot(3,2,i)
        y1 = np.cumsum(y)
        plt.plot(x,y1)
        plt.title('cdf')
        i=i+1
para= [(0,0.5),(0,1),(1,2)]
fig, axs  = plt.subplots(2,3,figsize=(10,10))
gauss(para)           
```

```python
'''Beta distribution'''
import numpy as np
import matplotlib.pyplot as plt
import math
def beta(parameters):
    x = np.linspace(0,1,1000)
    i  =1
    for (a,b) in parameters:
        y= ((x**(a-1))*((1-x)**(b-1)))/((math.gamma(a)*math.gamma(b))/math.gamma(a+b))
        mean = a/(a+b)
        print("mean is :",mean)
        median = (a-0.33)/(a+b-0.67)
        print("median is:",median)
        mode = (a-1)/(a+b-2)
        print("mode is:",mode)
        var = a*b/((a+b)**2*(a+b+1))
        print("variance is:",var)
        plt.subplot(3,2,i)
        plt.plot(x,y)
        plt.title('pdf')
        plt.plot(mean,marker="^",label = "mean")
        plt.plot(mode,marker="*",label= 'median')
        plt.plot(var,marker="o",label = "variance")
        plt.plot(median,marker="<",label="median")
        i = i+1
        plt.subplot(3,2,i)
        y1 = np.cumsum(y)
        plt.plot(x,y1)
        plt.title('cdf')
        i = i+1
para= [(3,5),(3,7),(2,7)]
fig, axs  = plt.subplots(2,3,figsize=(10,10))
beta(para)        
```

```python
'''standard normal distribution N~(mu,sigma**2)'''
import math
import numpy as np
import matplotlib.pyplot as plt
def gauss(parameters):
    x = np.linspace(-10,10,1000)
    i = 1
    for (mean , sigma) in parameters:
        y = np.exp(-((x-mean)/sigma)**2)/(sigma*math.sqrt(2*np.pi))
        plt.subplot(3,2,i)
        plt.plot(x,y)
        plt.title('pdf')
        mean = mean
        mode = mean 
        median = mean
        print('mean =median=mode is:',mean)
        var = sigma**2
        print('variance is:',var)
        plt.plot(mean,marker="^",label = "mean")
        plt.plot(var,marker="o",label = "variance")
        plt.plot(mode,marker="*",label="mode")
        plt.plot(median,marker=">",label="median")
        i = i+1
        plt.subplot(3,2,i)
        y1 = np.cumsum(y)
        plt.plot(x,y1)
        plt.title('cdf')
        i=i+1
para= [(0.5,1),(0.5,2),(-2,2)]
fig, axs  = plt.subplots(2,3,figsize=(10,10))
gauss(para)           
```

```python
'''uniform distribution '''
import numpy as np 
import matplotlib.pyplot as plt
import math
def uniform(x,a,b):
    y = [1/(b-a) if a<=i and i<=b else 0 for i in x]
    m = 0.5*(a+b)
    med= 0.5*(a+b)
    var = ((b-a)**2)/12
    return(x,y,m,med,var)
x = np.arange(-50,50)
for l in [(5,7),(5,8),(2,8)]:
    a,b,= l[0],l[1]
    m = 0.5*(a+b)
    print('mean is:',m)
    med= 0.5*(a+b)
    print('median is:',med)
    
    var = ((b-a)**2)/12
    print('variance is:',var)
    x,y,m,med,var = uniform(x,a,b)
    y1 = np.cumsum(y)
    plt.subplot(2,1,1)
    plt.plot(x,y,label = r' a=%d,b=%d' %(a,b))
    plt.grid()
    plt.title('pdf')
    plt.legend()
    plt.plot(m,marker="^")
    plt.plot(med,marker="*0 ")
    plt.plot(var,marker="o")
    plt.show()
    plt.subplot(2,1,2)
    plt.plot(x,y1,label = r' a=%d,b=%d' %(a,b))
    plt.legend()
    plt.title('cdf')
    plt.grid()
    plt.show()
```

```python
'''exponential distribution '''
import numpy as np
import matplotlib.pyplot as plt
import math
def exponential(x,L):
    y=L*np.exp(-L*x)
    mean = 1/L
    median = np.log(2)/L
    variance = 1/pow(L,2)
    return(x,y,mean,median,variance)
for i in [3,6,8]:
    t = np.arange(0,40,0.1)
    x,y,mean,median,variance=exponential(t,i)
    y1=np.cumsum(y)
    plt.subplot(2,1,1)
    plt.plot(x,y)
    plt.grid()
    plt.title('pdf')
    plt.legend()
    plt.plot(m,marker="^")
    plt.plot(med,marker="*")
    plt.plot(var,marker="o")
    plt.show()
    plt.subplot(2,1,2)
    plt.plot(x,y1)
    plt.legend()
    plt.title('cdf')
    plt.grid()
    plt.show()
    
```

```python
'''Chi squared distribution'''
import numpy as np
import matplotlib.pyplot as plt
import math
def chi(parameter):
    x = np.linspace(0,10,1000)
    i =1
    for k in parameter:
        y =((x**((k/2)-1))*np.exp(-(x/2)))/((2*(k/2))*math.gamma(k/2))
        
        plt.subplot(3,2,i)
        plt.plot(x,y)
        plt.title('pdf')
        mean = k
        print("mean is:",mean)
        
        median = k*pow((1-2/(9*k)),2)
        print("median is:",median)
        variance = 2*k
        print("variance is:",variance)
        i = i+1
        plt.subplot(3,2,i)
        y1 = np.cumsum(y)
        plt.plot(x,y1)
        plt.title('cdf')
        i = i+1
fig, axs  = plt.subplots(2,3,figsize=(10,10))
l=[3,4,5]
chi(l)
        
```
