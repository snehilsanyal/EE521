# exponential distribution

```python
def exponential(x,Lam):
    mean = 1/Lam
    median = np.log(2)/Lam
    variance = 1/pow(Lam,2)
    mode=0
    pdf=[]
    for i in x:
        if (i>0):
            pdf.append(Lam * np.exp(-Lam*i))
        else:
            pdf.append(0)
    cdf = np.cumsum(pdf)
    return cdf,pdf,mean,median,variance,mode
x = np.linspace(-10,30,100)
function = [exponential(x,0.5),exponential(x,2),exponential(x,3)]
mark=['*','o','+','^']
label1 =['mean','median','mode','variance']
z=['pdf','cdf']
g=['exponential(x,0.5)','exponential(x,2)','exponential(x,3)']
fig,ax=plt.subplots(2,3,figsize=(15,10))
c=1
for i in range(len(function)):
    for j in range(0,2):
        plt.subplot(3,2,c)
        plt.plot(x,function[i][j],color='blue')
        plt.gca().set_title(str(z[j])+' for ' + str(g[i]),color='red',size=15)
        for k in range(len(mark)):
            plt.plot(function[i][k+2],marker=mark[k],ms=10,label=label1[k])
        c=c+1
        plt.legend(loc='upper right')
```

# beta distribution

```python
def beta(x,alpha,beeta):
    pdf =[]
    a=math.gamma(alpha)
    b=math.gamma(beeta)
    c=math.gamma(alpha+beeta)
    h=(a*b)/c
    mean = alpha/(alpha+beeta)
    variance =(alpha*beeta)/((alpha+beeta)**2)*(alpha+beeta+1)
    median = (alpha-1/3)/(alpha+beeta-2/3)
    mode = (alpha-1)/(alpha+beeta-2)
    for t in x:
        r=(t**(alpha-1))*((1-t)**(beeta-1))/h
        pdf.append(r)
    cdf = np.cumsum(pdf)
    return pdf,cdf,mean,median,mode,variance
x = np.linspace(-1,1.5,100)
function = [beta(x,1,3),beta(x,5,1),beta(x,2,5)]
mark=['o','^','*','+']
label1 =['mean','median','mode','variance']
z=['pdf','cdf']
g=['beta(1,3)','beta(5,1)','beta(2,5)']
fig,ax=plt.subplots(2,3,figsize=(15,10))
c=1
for i in range(len(function)):
    for j in range(0,2):
        plt.subplot(3,2,c)
        plt.plot(x,function[i][j],'blue')
        plt.gca().set_title(str(z[j])+' for ' + str(g[i]),color='red',size=15)
        for k in range(len(mark)): 
            plt.plot(function[i][k+2],marker=mark[k],ms=10,label=label1[k])
        c=c+1   
        plt.legend(loc='upper right')
```

```python
def bernouli(p):
    pmf=[]
    for i in x:
        if (i==0):
            pmf.append((1-p))
        if i==1:
            pmf.append(p)
    return pmf
x=[0,1]
fig = plt.figure(10,figsize=(15,4))
plt.subplot(1,3,1)
a1=bernouli(0.1)
plt.stem(x,a1,use_line_collection=True)
plt.title('Bernoulli distribution with p = 0.1 & q= 0.9',color='red',size=12)
plt.subplot(1,3,2)
a2=bernouli(0.5)
plt.stem(x,a2,use_line_collection=True)
plt.title('Bernoulli distribution with p = 0.5 & q= 0.5',color='red',size=12)
plt.subplot(1,3,3)
a3=bernouli(0.8)
plt.title('Bernoulli distribution with p = 0.8 & q= 0.2',color='red',size=12)
plt.stem(x,a3,use_line_collection=True)
```

```python
def binomial(n,p):
    b=[]
    for r in range(n+1):
        y=(math.factorial(n)/(math.factorial(n-r)*math.factorial(r)))*(p**r)*((1-p)**(n-r))
        b.append(y)
    return b 
n=20
r=np.arange(0,n+1)
A=binomial(20,0.5)
plt.title('binomial distribution with p = 0.5',color='red',size=20)
plt.stem(r,A,use_line_collection=True)  
```

```python
def Geometric(p):
        b=[]
        x=[]
        for k in range(7):
            if p>=0.01:
                y=((1-p)**(k-1))*p
                b.append(y)
                x.append(k)
        return b,x 
a=Geometric(0.5)
plt.title('geometric distribution with p = 0.5',color='red',size=20)
plt.stem(a[1],a[0],use_line_collection=True)
```

```python
def poisson(lamda):
    PMF=[]
    for k in range(0,25):
        y=((lamda**k)*(np.exp(-lamda)))/math.factorial(k)
        PMF.append(y)
    return PMF
out=poisson(5)
x=np.arange(0,25)
plt.title('poission distribution with lambda = 5',color='red',size=20)
plt.stem(x,out,use_line_collection=True)
```

```python
x1 = [1, 2, 3, 2, 1, 2, 3, 4, 3, 2, 1] 
x2 = [1, 2, 1, 2, 1, 2, 1, 2] 
x3 = [1, 2, 3, 2, 4, 6, 4, 8, 12] 
x4 = [2, 4, 2, 4, 2, 4, 2, 4]
x5 = [1,-2,3,-2,-2,1,-2,3,-4,3,-2,1]
x6=[]
for i in x1:
    x6.append(2*i+1)
print('x6 =',x6)
plt.figure(figsize=(16,14))
c = 1
for i in range(6):
        plt.subplot(3,2,c)
        c = c+1
        plt.stem(x[i],use_line_collection=True)
        plt.gca().set_title('x' + str(i+1),color='red',size=25)
```

```python
def convolution(inputseq,impulseresponse):
        lengthA=np.size(inputseq)
        lengthB=np.size(impulseresponse)
        C = np.zeros(lengthA + lengthB -1)
        for m in np.arange(lengthA):
            for n in np.arange(lengthB):
                C[m+n] = C[m+n] + inputseq[m]*impulseresponse[n]
        return C 
```

```python
x=[x1,x2,x3,x4,x5,x6]
def autocorrelation(q):
    return convolution(np.conj(q),q[::-1])
plt.figure(figsize=(16,20))
c = 1
for i in range(6):
        plt.subplot(3,2,c)
        c = c+1
        print('autocorrelation of_x',[i+1],'=',np.array(autocorrelation(x[i])))
        plt.stem(autocorrelation(x[i]),use_line_collection=True)
        plt.gca().set_title('auto correlation of x' + str(i+1),color='red',size=20)
```

```python
x11=[0,0]+x1
x22=[0,0]+x2
x33=[0,0]+x3
x44=[0,0]+x4
x55=[0,0]+x5
x66=[0,0]+x6
x=[x11,x22,x33,x44,x55,x66]
```

```python
def crosscorrelation(A,B):
    return convolution(np.conj(A),B[::-1])
plt.figure(figsize=(16,60))
c = 1
for i in range(6):
    for j in range(i+1,6):
        plt.subplot(15,2,c)
        c = c+1
        plt.stem(crosscorrelation(x[i],x[j]),use_line_collection=True)
        plt.gca().set_title('cross correlation of x' + str(i+1)+' & ' + 'x' + str(j+1) ,color='red',size=20)
```
