# • Z Transform:

```python
"""
Consider the discrete time sequence x[n] = z n with −∞ < n < ∞ where
z ∈ C.. Let z = 0.95, z = 1.05, z = −1.05, z = −0.97, z = 1, z = −1.
Now plot the sequences for different values of z. For each plot what is the nature of the plot?
"""

# defining discrete time sequence x[n] = z^n

def sequence1(z):
    a = []
    n = []
    for j in range(-30,30):
        n.append(j)
    for i  in n:
        x_n = z**i
        a.append(x_n)
    return(a,n)

#Calling the above defined sequence with different values of z and plotting::::::>


plt.figure(figsize=(15,12))
z = [0.95,1.05,-1.05,-0.97,1,-1]
x = 1
for i in range(len(z)):
    b,n = sequence1(z[i])
    plt.subplot(3,2,x)
    x = x+1
    plt.stem(n,b,use_line_collection=True,)
    style.use("ggplot")
    plt.grid(color="gray")
    title = "The value of z= "+str(z[i])
    plt.title(title)
```

```python
"""
Consider the system y[n] = 2x[n − 3], the input to this system is the 
sequence z^n where z belong to the sequences in the previous section (all complex).
Find the response of the system. Also find the corresponding eigen values of the response.
"""

#For the given system if we will give input z^n then we will get output 2z^(n-2).

#Defining the output for the different value of z:

def sequence2(z):
    a = []
    n = []
    for j in range(-40,40):
        n.append(j)
    for i  in n:
        x_n = 2*z**(i-2)
        a.append(x_n)
    return(a)

#determining the response:

z = [complex(-2,3),complex(2,-3),complex(np.cos(np.pi/10),np.sin(np.pi/10)),complex(0.95*np.cos(np.pi/10),0.95*np.sin(np.pi/10)),complex(1.05*np.cos(np.pi/10),1.05*np.sin(np.pi/10))]

z1 = sequence2(z[0])
print("the responce for value of z =",z[0],"is=\n",z1)

z2 = sequence2(z[1])
print("\n\nthe responce for value of z =",z[1],"is=\n",z2)

z3 = sequence2(z[2])
print("\n\nthe responce for value of z =",z[2],"is=\n",z3)

z4 = sequence2(z[3])
print("\n\nthe responce for value of z =",z[3],"is=\n",z4)

z5 = sequence2(z[4])
print("\n\nthe responce for value of z =",z[4],"is=\n",z5)
```

```python
"""
The eigen values of the response is the Z Transform of the impulse response. This is also known as the transfer function denoted by H(z). Compute the magnitude spectrum |H(z)| and the phase spectrum arg[H(z)]
and plot them in separate plots. Plot for the previous section by considering complex z values from last part.

"""

#from above given system H(z)=2*z^(-3)
#putting z = exp^(jw)

w = np.linspace(-np.pi,np.pi,100)


def mag(w):
    magnitude = []
    for i in w:
        mag = 2
        magnitude.append(mag)
    return(magnitude)


def pha(w):
    phase = []
    for i in w:
        phase1 = complex(np.cos(3*i),-np.sin(3*i)) 
        phase.append(cmath.phase(phase1))
    return(phase)

magnitude= mag(w)
phase = pha(w)

fig = plt.figure(1,figsize=(15,5))
plt.subplot(1,2,1)
plt.plot(w,magnitude)
style.use("ggplot")
plt.grid(color="gray")
plt.xlabel('w')
plt.title("Magnitude spectrum")

plt.subplot(1,2,2)
plt.plot(w,phase)
style.use("ggplot")
plt.grid(color="gray")
plt.xlabel('w')
plt.title("Phase spectrum")
plt.show()
```

# • Discrete Time Fourier Series:

```python
"""
Consider a periodic square wave with fundamental period N = 16 having 
magnitude 1 between n = 0 and n = 7 and zero elsewhere. Make a plot
of this signal over two periods in blue colour.
"""

# Plotting periodic square wave:


a = []
n = np.arange(32)
for i in range(2):
    for j in range(16):
        if (j<=7):
            a.append(1)
        else:
            a.append(0)
fig = plt.figure(1,figsize=(15,5))
plt.stem(n,a,'tab:blue',use_line_collection=True)
style.use("ggplot")
plt.grid(color="gray")
plt.xlabel("n ---->")
plt.title("Periodic square wave")
plt.show()

```

```python

"""
Determine the DTFS coefficients of this signal and generate magnitude
plots of the coefficients.

"""
a = np.zeros(16)
n = np.arange(16)
a[0] = 0.5
print("DTFS coefficients of the signal are=\n")
print(a[0],end=", ")
for k in range(1,16):
  a[k] = 1/16*(np.sin(k*0.5*np.pi))/(np.sin(k*np.pi/16))
  print(abs(round(a[k],2)),end=", ")

plt.figure(figsize=(10,5))
style.use("ggplot")
plt.grid(color="gray")
plt.stem(n,abs(a),use_line_collection=True)
plt.xlabel("k --->")
plt.title("DTFS coefficients of this signal")
plt.show()
```

# • Continuous Time Fourier Transform:

```python
"""
Consider a rectangular signal
with amplitude 5 and existing between t = −1 and t = 1. Compute
the continuous time Fourier transform of this signal and display a nicely Labelled plot.
"""
#plotting the Given rectangular signal:

plt.figure(figsize=(10,3))
T=2
t = np.linspace(-10,10,1000)
rect_t = (t>=(-T/2)) ^ (t>=(T/2))
plt.plot(t,5*rect_t,color='black')
style.use("ggplot")
plt.grid(color="gray")
plt.title("Given rectangular signal")
plt.xlabel('t  ------>')

plt.show()

#The transform of given signal:

plt.figure(figsize=(10,3))
w = np.linspace(-30,30,1000)
sinc_t = 10*np.sin(w/np.pi)/(w/np.pi)
style.use("ggplot")
plt.grid(color="gray")
plt.plot(w,sinc_t,color='black')
plt.title("The transform of given signal")
plt.xlabel('w ------>')
plt.show()

```

```python
"""
A message signal m(t) is used to modulate a carrier signal c(t) = cos2πfct
and the resultant signal is given by y(t) = m(t)c(t). Plot y(t).
"""
# Defining the message signal m(t):

t = np.linspace(-3,3,1000)
fig = plt.figure(1,figsize=(15,9))
T=1
tri_t = (1 - (abs(t)/T)) * (abs(t) < T)
plt.subplot(3,1,1)
plt.plot(t,tri_t,color='black')
plt.grid(color="gray")
plt.xlabel("t ----->")
plt.title("m(t)")
plt.show()

# Defining the carrier signal c(t):
c_t = np.cos(2*np.pi*10*t)
fig = plt.figure(1,figsize=(15,9))
plt.subplot(3,1,2)
plt.plot(t,c_t,color='black')
plt.grid(color="gray")
plt.xlabel("t ----->")
plt.title("c(t)")
plt.show()

# Defining the signal y(t):

y_t = c_t*tri_t
fig = plt.figure(1,figsize=(15,9))
plt.subplot(3,1,3)
plt.plot(y_t,color='black')
plt.grid(color="gray")
plt.xlabel("t ----->")
plt.title("y(t)=c(t)m(t)")
plt.show()
```
