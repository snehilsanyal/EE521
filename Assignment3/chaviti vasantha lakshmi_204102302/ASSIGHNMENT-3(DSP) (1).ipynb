# 2.DISCRETE TIME FOURIER SERIES

Consider a periodic square wave with fundamental period N = 16 having
magnitude 1 between n = 0 and n = 7 and zero elsewhere. Make a plot
of this signal over two periods in blue colour

```python
list = []
n = np.arange(32)
for j in range(2):
    for i in range(16):
        if (i<=7):
            list.append(1)
        else:
            list.append(0)
fig = plt.figure(1,figsize=(10,4))
plt.stem(n,list,use_line_collection=True)
plt.title("Periodic square wave",color='red',size=20)
plt.show()
```

Determine the DTFS coefficients of this signal and generate magnitude
plots of the coefficients

```python
x = np.zeros(16)
n = np.arange(16)
x[0] = 0.5
for i in range(1,16):
    x[i] = 1/16*(np.sin(i*0.5*np.pi))/(np.sin(i*np.pi/16))
plt.figure(figsize=(8,5))
plt.stem(n,abs(x),use_line_collection=True)
plt.title("coefficients of the signal",color='red',size=20)
plt.show()
```

Using the coefficients determined above, make an approximate reconstruction of the square wave and display it on the plot of the original signal in
red colour.

```python
def dtft(N,y):
    a=[]
    for n in range(32):
        total=0
        for i in range(N):
            total=total+y[i]*np.exp((1j*n*2*np.pi*i)/N)
        a.append(round(abs(total),2))
    return a 
n=np.arange(0,32,1)
signal=dtft(16,x)
fig= plt.figure(figsize=(8,4))
plt.stem(list,use_line_collection=True,label='original signal')
plt.stem(n,signal,'r',use_line_collection=True,label='recontructed signal')
plt.legend()
plt.title('reconstruction of square wave',color='red',size=20)
```

# 3.CONTINOUS TIME FOURIER TRANSFORM

---->Consider a rectangular signal
with amplitude 5 and existing between t = −1 and t = 1. Compute
the continuous time Fourier transform of this signal and display a nicely
labelled plot.

```python
def rect(t): 
    rect =[] 
    for k in t :
        if k>=-1 and k <1:
            rect.append(5) 
        else: 
            rect.append(0)
    return rect
def ftofrect(omega):
    a=(10*np.sin(omega))/omega
    return a
plt.figure(figsize=(18,4))
t=np.linspace(-5,5,100)
omega=np.linspace(-30,30,1000)
plt.subplot(1,2,1)
plt.plot(t,rect(t),color='blue')
plt.title('rectangular pulse',color='red',size=20)
plt.grid()
plt.xlim(-5,5)
plt.subplot(1,2,2)
plt.plot(omega,ftofrect(omega),color='blue')
plt.title('fourier transform of rectangular pulse',color='red',size=20)
plt.grid()
```

--->Compute the Fourier transform of m(t) and display the plots.
m(t) is a triangular signal existing between t = −1 and t = 1 with a peak
amplitude of 1

```python
def tri(t):
    return (1 -(abs(t)/1))*(abs(t)<1)
def ftoftri(w1):
    return ((1/w1)*np.sin(w1))**2
plt.figure(figsize=(18,4))
t=np.linspace(-5,5,1000)
w1 = np.linspace(-5*np.pi,5*np.pi,1000)
plt.subplot(1,2,1)
plt.plot(t,tri(t),color='blue')
plt.title('triangular pulse-->m(t)',color='red',size=20)
plt.grid()
plt.xlim(-5,5)
plt.subplot(1,2,2)
plt.plot(omega,ftoftri(omega),color='blue')
plt.title('fourier transform of triangular pulse',color='red',size=20)
plt.grid()
```
