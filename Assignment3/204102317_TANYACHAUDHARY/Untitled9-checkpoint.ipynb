## PLOTTING DISCRETE TIME SEQUENCE = Z^n

```python
import matplotlib.pyplot as plt  
import numpy as np  
import random
import math

def z(n,val):
    ztr=[]
    for t in n:
        res=val**t
        ztr.append(res)
        
    return ztr
```

## To find the response of the given system

###  Given the system   y[n] = 2*x[n-3]  
###   Taking the z transform both sides , Y(z) = 2*X(z)*z^-3  , further  Y(z)/X(z) = H(z) = 2*z^-3

We have a property in the frequency response , which states that if input to a system is an complex exponential then the output is the same exponential , multiplied by a coefficient that is the transfer function H(z) which is evaluated at the input frequency.

So we are given different values of input that is z^n, for each individual input we calculate H(z) at that particular z value 
which is known as the eigen value at the output side.
Here we have directly substituted the z and plotted the magnitude and phase spectrum

```python
import numpy as np
import matplotlib.pyplot as plt
from  matplotlib import patches
from matplotlib.figure import Figure
from matplotlib import rcParams
    
def zplane(b,a,filename=None):
    ax = plt.subplot(111)
    uc = patches.Circle((0,0), radius=1, fill=False,
                        color='black', ls='dashed')
    ax.add_patch(uc)
    if np.max(b) > 1:
        kn = np.max(b)
        b = b/float(kn)
    else:
        kn = 1

    if np.max(a) > 1:
        kd = np.max(a)
        a = a/float(kd)
    else:
        kd = 1
        
    p = np.roots(a)
    z = np.roots(b)
    k = kn/float(kd)
    
    # Plot the zeros and set marker properties    
    t1 = plt.plot(z.real, z.imag, 'go', ms=10)
    plt.setp( t1, markersize=10.0, markeredgewidth=1.0,
              markeredgecolor='k', markerfacecolor='g')

    # Plot the poles and set marker properties
    t2 = plt.plot(p.real, p.imag, 'rx', ms=10)
    plt.setp( t2, markersize=12.0, markeredgewidth=3.0,
              markeredgecolor='r', markerfacecolor='r')

    ax.spines['left'].set_position('center')
    ax.spines['bottom'].set_position('center')
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)

    # set the ticks
    r = 1.5; plt.axis('scaled'); plt.axis([-r, r, -r, r])
    ticks = [-1, -.5, .5, 1]; plt.xticks(ticks); plt.yticks(ticks)

    if filename is None:
        plt.show()
    else:
        plt.savefig(filename)
    

    return z, p, k

b=[1]
a=[1]
res=zplane(b,a,'none')
res
```

# DISCRETE TIME FOURIER SERIES

### QUES - The DTFS coefficients of a periodic discrete time signal with period N = 5
### a0 = 1, a2 = a-2=exp^j*pi/4 ,a4= a-4=2exp^j*pi/3

### USE OF THE SYMMETRY  PROPERTY  IS APPLIED HERE , WHICH IS X[N-K]= X*[K] 
### SO N is 5 here , X[3]=X[5-2]=X*[2] which implies a3=exp^(-j*pi/4)
### X[1] = X[5-4]=X*[4] which implies a1=2*exp^(-j*pi/3)

```python
import numpy as np
import time

def dtin(arr,N):
    
    out1=[]
    for i in range(0,N):
        sum=0
        for l in range(0,N):
            ans=(arr[l]*np.exp(1j*2*i*np.pi*l/N))
            sum = sum + ans
            
        out1.append(sum)
        
    return out1
            

lst=[1,(np.exp(-1j*np.pi*3))*2,np.exp(1j*np.pi/4),np.exp(-1j*np.pi/4),(np.exp(1j*np.pi*3))*2]

n=np.arange(0,5,1)
res=dtin(lst,5)
plt.stem(n,res,use_line_collection=True)
```

# CONTINUOUS TIME FOURIER TRANSFORM

```python
import matplotlib.pyplot as plt  
import numpy as np  
import random
import math
from scipy import integrate


def cft(g, f):    
    result = np.zeros(len(f), dtype=complex)
    for i, val in enumerate(f):
        result[i] = complex_quad(lambda t: g(t)*np.exp(-2j*np.pi*val*t), -5, 5)
    return result

def complex_quad(g, a, b):
    t = np.linspace(a, b, 2501)  
    x = g(t)
    return integrate.simps(x, t)


def rect(t):
    return (abs(t) < 1).astype(float)
```

```python
t = np.linspace(-2,2, 1000)
plt.title('RECTANGULAR SIGNAL')
plt.plot(t, rect(t))
```

```python
f = np.linspace(-5, 5, 1000)
R = cft(rect, f)  # Calculate the numeric Fourier Transform of the rect function
plt.title('RECTANULAR FOURIER TRANSFORM')
plt.plot(f, R, 'r-', lw=2)
```

```python
import matplotlib.pyplot as plt  
import numpy as np  
import random
import math
from scipy import integrate



def tri(t):
    return (1 - abs(t))*1 * (abs(t) < 1)

def cft(g, f):    
    result = np.zeros(len(f), dtype=complex)
    for i, val in enumerate(f):
        result[i] = complex_quad(lambda t: g*np.exp(-2j*np.pi*val*t), -5, 5)
    return result

def complex_quad(g, a, b):
    t = np.linspace(a, b, 300)  
    x = g(t)
    return integrate.simps(x, t)

```

##  BONUS QUESTION

Given two signals , rect and tri and how to obtain one from the other,using some operations.

### So using Convolution , we know that convolution of two rect pulses yield a triangular pulse

```python
import numpy as np
import matplotlib.pyplot as plt



def rect(t):
    return 1 * (abs(t) < 0.5)

t1 = np.linspace(-10, 10, 1999)
t = np.linspace(-4, 4, 1000)
re=rect
conv =np.convolve(re(t),re(t),'full')
plt.plot(t1,conv)
```

## Also convolution in time domain leads to multiplication in frequency domain
## Here rect fourier transform is sinc and multiplying sinc with itself results in tri

```python
import matplotlib.pyplot as plt  
import numpy as np  
import random
import math
from scipy import integrate


def cft(g, f):    
    result = np.zeros(len(f), dtype=complex)
    for i, val in enumerate(f):
        result[i] = complex_quad(lambda t: g(t)*np.exp(-2j*np.pi*val*t), -5, 5)
    return result

def complex_quad(g, a, b):
    t = np.linspace(a, b, 2501)  
    x = g(t)
    return integrate.simps(x, t)


def rect(t):
    return 1* (abs(t) < 0.5)


f = np.linspace(-3, 3, 300)
R = cft(rect, f)  
plt.plot(f, R*R, 'r-', lw=2)# Multiplying two sinc together to find triangular
```
