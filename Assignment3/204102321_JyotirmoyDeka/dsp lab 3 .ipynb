# Discrete Time Fourier Series:

The DTFS coefficients of a periodic discrete time signal with period N = 5
are given below:
a0 = 1, a2 = a∗−2 = ejπ/4, a4 = a∗−4 = 2ejπ/3
Using the properties of DTFS, determine the values of a0 through a4. Using these coefficients, determine one period of the signal and display a
labelled plot.



```python
## using property of DTFS : a1=a4, a2 = a3 since an = a(N-n)

a = np.array([1,2*np.exp(1j*(math.pi/3)),np.exp(1j*(math.pi/4)),np.exp(1j*(math.pi/4)),2*np.exp(1j*(math.pi/3))])
x = np.zeros(5)
n = np.arange(5)
for i in n:
    for k in range(len(a)):
        x[i] += a[k]*np.exp(1j*k*(math.pi/8)*i)
print("The value of x[0],x[1],x[2],x[3],x[4] are : ",x) ## forier series using the above coefficients 
plt.xlabel('real')
plt.ylabel('imag')
plt.scatter(n,x)
plt.title("labeled plot of DTFS")
plt.grid()
plt.show()
```

Consider a periodic square wave with fundamental period N = 16 having magnitude 1 between n = 0 and n = 7 and zero elsewhere. Make a plot of this signal over two periods in blue colour.

```python
plt.figure(figsize=(10,5))
plt.xlabel("Time")
plt.ylabel("Magnitude")
plt.title("Periodic square wave")
n = np.arange(-32,32)
y = np.array([1]*8+[0]*8+[1]*8+[0]*8+[1]*8+[0]*8+[1]*8+[0]*8) ## period of 16 units of time with 1 for a0 to a7 ,and 0 elsewhere
plt.stem(n,y)
```

Determine the DTFS coefficients of this signal and generate magnitude plots of the coefficients.


```python
## Determining the DTFS coefficients and plotting of magnitude plot
def coef(N,y):
    c=[]
    for k in range(16):
        c_k=0
        for i in n:
            c_k=c_k + y[i]*np.exp((-1j*k*2*np.pi*i)/N)
        c.append(round(abs(c_k/N),2))
    return c
n = np.arange(0,16,1)
res = coef(16,y)
print('the DTFS coefficients are :')
print(res)
l = np.asarray(res)
plt.stem(n,abs(l))
plt.title('labelled plot of DTFT coefficients os square wave')
plt.xlabel('n')
plt.ylabel('amplitude')
plt.grid()
plt.show()
```

Using the coefficients determined above, make an approximate reconstruction of the square wave and display it on the plot of the original signal in red colour.


```python
n = np.arange(0,32)
a_k = [0.08, 0.03, 0.01, 0.01, 0.01, 0.01, 0.01, 0.01, 0.01, 0.01, 0.01, 0.01, 0.01, 0.01, 0.01, 0.03]
x = np.zeros(32)
for i in n:
    for k in range(len(a_k)):
        x[i] =x[i]+ a_k[k]*np.exp(1j*k*(np.pi/8)*i)
y = np.array([1]*8+[0]*8+[1]*8+[0]*8)

print("the DTFS is :")
print(x)
plt.grid()
plt.xlabel("n")
plt.ylabel("Magnitude")
plt.title("Reconstruction of square wave from the DTFS coefficients")
plt.stem(n+0.2,x,'r',markerfmt="ro",use_line_collection=True)
plt.stem(n,y,use_line_collection=True)
plt.show()
```

# Continuous Time Fourier Transform:

Consider a rectangular signal with amplitude 5 and existing between t = −1 and t = 1. Compute the continuous time Fourier transform of this signal and display a nicely labelled plot

```python
N = 100                  # Number of samplepoints
T =0.1                     # sample spacing
x = np.linspace(-5.0, N*T, N)
y = np.zeros(x.shape)
for i in range(x.shape[0]):
    if x[i] > -1.0 and x[i] < 1.0: 
        y[i] = 5.0
plt.plot(x,y)
plt.xlim(-5,5)
plt.grid()
plt.title('Rectangular signal')
plt.show()
yf = np.fft.fft(y)
yf = np.fft.fftshift(yf)
xf = np.linspace(-1.0/(2.0*T), 1.0/(2.0*T), N)
fig, ax = plt.subplots()
ax.plot(xf, np.abs(yf) ) 
plt.title('fourier transform of rectangular signal')
plt.grid()
plt.show()
```

# BONUS QUESTION:

Make out a triangular wave from rectangular waves.

We can get a triangular signal by convolution of two rectangular signals.This is an application of its Fourier Transform

```python
n = 1000
t = np.linspace(0, 8, n)
T = t[1] - t[0]  # sampling width

x1 = np.where(t<=4, 1, 0)             # build input functions
x2 = np.where(np.logical_and(t>=4, t<=6), 2, 0)

y = np.convolve(x1, x2) * T       # scaled convolution 
ty = np.linspace(0, 2*8, n*2-1)                # double time interval

    
fg, ax = plt.subplots(1, 1)
ax.plot(t, x1, label="$x_1$")
ax.plot(t, x2, label="$x_2$")
ax.plot(ty, y, label="$x_1\\star x_2$")
ax.legend()
ax.grid()
```
