```python
#------------------------------------------x[n]=z^n-----------------------------------------#
import sympy as sp
from pprint import pprint
import numpy as np
import matplotlib.pyplot as plt
import cmath
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
n=np.arange(-15,14,1)
z=[0.95,1.05,-1.05,-0.97,1,-1]
x={}
col=['r','g','c','k','y','b','m','r','g','c','k','y','b','m','r','g','c','k','y','b','m']
for i in z:
    p="x_"+str(i)
    x[p]=[]
r=0
for key in x:
    for i in n:
        if i<0:
            x[key].append(1/pow(z[r],-i))
        else:
            x[key].append(pow(z[r],i))
    r+=1
fig=1
r=0
for key in x:
    plt.figure(fig)
    labl="z="+str(z[r])
    titl="plot for x[n] for x[n]=z^n where z="+str(z[r])
    plt.scatter(n,x[key],color=col[r],label=labl)
    plt.bar(n,x[key],color=col[r],width=0.1)
    plt.xlabel('n')
    plt.ylabel('x[n]=z^n')
    plt.title(titl)
    plt.grid("True")
    r+=1
    fig+=1
    plt.legend()
```

```python
#----------------------------------Complex Z-------------------------------------#
y={}
col=['r','g','c','k','y','b','m','r','g','c','k','y','b','m','r','g','c','k','y','b','m']
s=complex(np.cos(np.pi/10),np.sin(np.pi/10))
y_re={}
y_im={}
z=[complex(-2,3),complex(2,-3),s,0.95*s,1.05*s]
zlabel=["z=-2+3j","z=2-3j","z=e^sin(pi/10)","z=0.95sin(pi/10)","z=1.05sin(pi/10)"]
for i in z:
    p="y_"+str(i)
    y[p]=[]
    y_re[p]=[]
    y_im[p]=[]
r=0
n=np.arange(0,100,1)
q=np.arange(0,10,1)
for key in y:
    if r<2:
        for i in q:
            y[key].append(pow(z[r],i))
    else:
        for i in n:
            y[key].append(pow(z[r],i))
    r+=1
r=0
for key in y:
    if r<2:
        for i in range(len(q)):
            y_re[key].append(y[key][i].real)
            y_im[key].append(y[key][i].imag)
    else:
        for i in range(len(n)):
            y_re[key].append(y[key][i].real)
            y_im[key].append(y[key][i].imag)
    r+=1
r=0
#--------------------------------------------Printing the complex values-------------------------------------#
for key in y:
    figu = plt.figure(fig,figsize=(10,10))
    ax = plt.axes( projection='3d')
    if r<2:
        ax.plot3D(y_re[key],q, y_im[key], c=col[r], marker='o')
    else:
        ax.plot3D(y_re[key],n,y_im[key],c=col[r],marker='o')
    titl="Plot for "+str(zlabel[r])
    ax.set_xlabel("Real Part")
    ax.set_zlabel("Imaginary part")
    fig+=1
    r+=1
    ax.set_title(titl)
```

```python
#----------------------------------y[n]=x[n-3]----------------------------------------------#


s=complex(np.cos(np.pi/10),np.sin(np.pi/10))
col=['r','g','c','k','y','b','m','r','g','c','k','y','b','m','r','g','c','k','y','b','m']
x={}
y={}
y_re={}
y_im={}
x_re={}
x_im={}
z=[complex(-2,3),complex(2,-3),s,0.95*s,1.05*s]
zlabel=["z=-2+3j","z=2-3j","z=e^sin(pi/10)","z=0.95sin(pi/10)","z=1.05sin(pi/10)"]
for i in z:
    p=str(i)
    x[p]=[]
    y[p]=[]
    y_re[p]=[]
    y_im[p]=[]
    x_re[p]=[]
    x_im[p]=[]
r=0
n=np.arange(0,100,1)
q=np.arange(0,40,1)
for key in x:
    if r<2:
        for i in q:
            x[key].append(pow(z[r],i))
    else:
        for i in n:
            x[key].append(pow(z[r],i))
    r+=1
r=0
for key in x:
    if r<2:
        for i in q:
            if i-3<0:
                y[key].append(0)
            else:
                y[key].append(x[key][i-3])
    else:
        for i in n:
            if i-3<0:
                y[key].append(0)
            else:
                y[key].append(x[key][i-3])
    r+=1
r=0
for key in x:
    if r<2:
        for i in q:
            x_re[key].append(x[key][i].real)
            x_im[key].append(x[key][i].imag)
            y_re[key].append(y[key][i].real)
            y_im[key].append(y[key][i].imag)
    else:
        for i in n:
            x_re[key].append(x[key][i].real)
            x_im[key].append(x[key][i].imag)
            y_re[key].append(y[key][i].real)
            y_im[key].append(y[key][i].imag)
    r+=1
r=0
#--------------------------------------------Printing the complex values-------------------------------------#
for key in y:
    figu = plt.figure(fig,figsize=(10,10))
    titl1="Plot for Real part of "+str(zlabel[r])
    ax = plt.axes( projection='3d')
    if r<2:
        ax.plot3D(y_re[key],q, y_im[key], c=col[r],label="y[n]")
        ax.plot3D(x_re[key],q, x_im[key], c=col[r+1],label="x[n]")
    else:
        ax.plot3D(y_re[key],n, y_im[key], c=col[r], marker='o',label="y[n]t")
        ax.plot3D(x_re[key],n, x_im[key], c=col[r+1], marker='^',label="x[n]t")
    ax.set_xlabel("Real Part")
    ax.set_zlabel("Imaginary part")
    ax.legend()
    ax.set_title(titl1)
    fig+=1
    r+=1
    
```

```python
def convolution(inputSeq, impulseResponse):
    inputSeq = np.array(inputSeq)
    impulseResponse = np.array(impulseResponse)
    n1=len(inputSeq)
    n2=len(impulseResponse)
    outputSeq=[]
    for n in range(n1+n2-1):
        summ=0
        for k in range(n1):
            if n-k>=0 and n-k<=n2-1:
                summ=summ+inputSeq[k]*impulseResponse[n-k]
        outputSeq=np.append(outputSeq,summ)
    return outputSeq
plt.figure()
response=np.array(convolution(z,Impulse_res))
p=np.linspace(0,13,14)
plt.figure(fig,figsize=(15,4))
plt.plot(p,response.real,label='real part')
plt.xlabel("n")
plt.ylabel("Real Part")
plt.grid('True')
plt.legend()
fig+=1
plt.figure(fig)
plt.plot(p,response.imag,label='imag part')
plt.legend()
plt.grid('True')
```
