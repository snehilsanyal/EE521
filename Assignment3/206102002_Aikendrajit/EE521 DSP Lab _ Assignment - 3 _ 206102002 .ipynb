### Continuous Time Fourier Transform
Here, for a rectangular signal with amplitude 5 and, t = -1 to 1.

```python
# A rectangular signal whose magnitude is 5 and tou = 2
# since t = -1 to 1

A = 5
T = 2
n = np.linspace(-15, 16, 1000)
rec_sig = (n>=(-T/2))^(n>=(T/2))
plt.plot(n, A*rec_sig)
plt.title("Rectangular signal with amplitude A = 5 and T = 2")
plt.grid()
plt.show()
```

```python
#Continuous Z transform of the rectangular signal:

A = 5
T = 2
w = np.linspace(-15, 16, 1000)
zt_rec_sig = (A/w)*2*np.sin(w*T/2)
plt.plot(w, zt_rec_sig)
plt.title("Transformed Rectangular Signal")
plt.grid()
plt.show()
```

```python
#Message Signal: Given as Triangular Signal

def mt(x,c,hc): #  with amplitude hc, width c, and slope hc/2c
    if x>=c/2:
        r = 0.0
    elif x<=-c/2:
        r = 0.0
    elif x > -c/2 and x<0:
        r=2*x/c*hc+hc
    else:
        r=-2*x/c*hc+hc
    return r

x=np.linspace(-2.5,2.5,1000)
y=np.array([mt(t,2.0,1.0) for t in x])
plt.ylim(-0.2,1.2)
plt.plot(x,y)
plt.title("Triangular Signal (Message Signal m(t))")
plt.grid()
plt.show()
```

```python
#Carrier Signal:

t = np.linspace(-1, 1, 1000)
f = 20
ct = np.cos(2*np.pi*f*t)
fig = plt.figure(figsize = (8, 6))
plt.plot(t, ct)
plt.title("Carrier Signal c(t)")
plt.grid()
plt.show()
```

```python
#Resultant Signal:

t = np.linspace(-1, 1, 1000)
yt = mt(t,2.0,1.0)*ct
fig = plt.figure(figsize = (8, 6))
plt.plot(t, yt)
plt.title("Resultant Signal c(t)")
plt.grid()
plt.show()
```
