# Discrete time fourier series

```python
'''calculation of  DTFS coefficient a[0] from a[4]'''
# we can calculate missing coefficient using the property of periodicity i. e. C_k = C_N-k \\where N is the period

import matplotlib.pyplot as plt
import numpy as np
import warnings

a = np.array([1,2*np.exp(1j*(np.pi/3)),np.exp(1j*(np.pi/4)),np.exp(1j*(np.pi/4)),2*np.exp(1j*(np.pi/3))])
x = [0,0,0,0,0]
n = np.arange(5)
for i in n:
    for k in range(len(a)):
        x[i] += a[k]*np.exp(1j*k*(np.pi/8)*i)
print("The values of a[0] through a[4] are-")
print(x)
plt.xlabel('real')
plt.ylabel('imag')
plt.scatter(n,x)
plt.grid()
plt.show()
```

```python
'''periodic square wave'''
import numpy as np 
import matplotlib.pyplot as plt
n = np.arange(0,32)
y = np.array([1]*8+[0]*8+[1]*8+[0]*8)
plt.stem(n,y,use_line_collection=True)
plt.title('periodic square wave')
plt.xlabel('n')
plt.ylabel('amplitude')
plt.grid()
plt.show()
```

```python
'''Determining the DTFS coefficients and plotting of magnitude plot '''
import matplotlib.pyplot as plt
import numpy as np 
def coef(N,y):
    c=[]
    for k in range(16):
        c_k=0
        for i in n:
            c_k=c_k + y[i]*np.exp((-1j*k*2*np.pi*i)/N)
        c.append(round(abs(c_k/N),2))
    return c
n = np.arange(0,16,1)
res = coef(16,y)
print('the DTFS coefficients are :')
print(res)
l = np.asarray(res)
plt.stem(n,abs(l))
plt.title('labelled plot of DTFT coefficients os square wave')
plt.xlabel('n')
plt.ylabel('amplitude')
plt.grid()
plt.show()
        

```

```python
'''reconstruction of square wave from the DTFS coefficients'''
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
n = np.arange(0,32)
a_k = [0.5, 0.32, 0.0, 0.11, 0.0, 0.08, 0.0, 0.06, 0.0, 0.06, 0.0, 0.08, 0.0, 0.11, 0.0, 0.32]
x = np.zeros(32)
for i in n:
    for k in range(len(a_k)):
        x[i] =x[i]+ a_k[k]*np.exp(1j*k*(np.pi/16)*i)
y = np.array([1]*8+[0]*8+[1]*8+[0]*8)

print(x)
plt.grid()
plt.xlabel("n")
plt.ylabel("Magnitude")
plt.title("Reconstruction of square wave from the DTFS coefficients")
plt.stem(n+0.2,x,'r',markerfmt="ro",use_line_collection=True)
plt.stem(n,y,use_line_collection=True)
plt.show()
```

# Continuous time fourier transform

```python
'''rectangular signal and its fourier transform'''
import numpy as np
import matplotlib.pyplot as plt

N = 1000 # Number of samplepoints
T =0.1# sample spacing
x = np.linspace(-5.0, N*T, N)
y = np.zeros(x.shape)
for i in range(x.shape[0]):
    if x[i] > -1.0 and x[i] < 1.0: 
        y[i] = 5.0
plt.plot(x,y)
plt.xlim(-5,5)
plt.grid()
plt.title('Rectangular signal')
plt.show()
yf = np.fft.fft(y)
yf = np.fft.fftshift(yf)
xf = np.linspace(-1.0/(2.0*T), 1.0/(2.0*T), N)
fig, ax = plt.subplots()
ax.plot(xf, abs(yf) ) 
plt.title('fourier transform of rectangular signal')
plt.grid()
plt.show()
```
