# Fourier series representation of a Square wave 

Consider a periodic square wave with time period T = 2 seconds having
magnitude 1 between t = 0 s and t = 1 s and zero elsewhere

```python
import numpy as np 
import matplotlib.pyplot as plt 

sqr=[]
n = np.linspace(-4, 4, 1000)
for i in n:
    if i%2<1:
        sqr.append(1)
    else:
        sqr.append(0)
        
plt.plot(n,sqr)
plt.title('Square wave with time period=2')
plt.show()
```

Define a function to plot g(t) that is a fourier series representation of the square wave

```python
def fourier(sig,T,terms):
    a=[]
    b=[]
    s=[]
    inte=(T+2)/len(n)
    a0= 1/T*inte*np.sum(sig)
    for k in range(1,terms):
        coeff_a=1/T*inte*np.sum(sig*np.cos(2*np.pi*k*n/T))
        a.append(coeff_a)
        coeff_b=1/T*inte*np.sum(sig*np.sin(2*np.pi*k*n/T))
        b.append(coeff_b)
        s.append(coeff_a*np.cos(2*np.pi*k*n/T)+ coeff_b*np.sin(2*np.pi*k*n/T) )
    
    overall_sum=a0/2+ sum(s)
    return overall_sum

a=fourier(sqr,2,5)
fig=plt.figure(figsize=(15,4))
plt.plot(n,a,label='Fourier Series')
plt.plot(n,sqr,label='Original square wave')
plt.title('Discrete fourier series')
plt.legend()
```

Gibbs  phenomenon occurs near a jump discontinuity in the signal.Any signal with a sudden discontinuity or step will show the Gibbs phenomenon, which is in form of ripples of increasing frequency.
This phenomenon  says that no matter how many terms you include in your Fourier series as asked in question that if large number of terms are included in the fourier series there will always be an error in the form of an overshoot near the disconti nuity.

# 2. Plot the power spectral density and visualize the spectral leakage

```python
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt

def DFT(arr,N):
    out=[]
    for k in range(N):
        sum=0
        for i in range(N):
            res=(arr[i]*np.exp(-1j*2*i*np.pi*k/N))
            sum = sum + res
            
        out.append(sum)
        
    return out


f = 30 
fs1 = 100 
T1 = 0.99
N1 = int(fs1*T1) # No. of samples of DFT
t1 = np.arange(0,T1,1/fs1)
x1 = np.cos(2*np.pi*f*t1)
n1 = np.arange(0,N1,1)
X1_dft = DFT(x1,N1)
f_n = (n1*fs1)/N1 # since f=k/NTs
plt.plot(t1,x1)
plt.title('Original signal')
```

```python
plt.stem(X1_dft)
plt.title('DFT of signal')
```

```python
plt.psd(X1_dft)
plt.title('PSD in db/Hz')
```

# Using Different  Window functions to minimize the spectral leakage

###  HANNING WINDOW

```python
c = (1-np.cos(2*np.pi*t1))/2 # Hanning window function
han = DFT(c,N1)
res=[]
for i in n1:
    r = X1_dft[i]*han[i]
    res.append(r)
    
d1=DFT(res,N1)
plt.stem(c)
plt.title('HANNING window function')
    
```

```python
plt.psd(d1)
plt.title('PSD after Window applied on the signal in db/Hz')
```

### HAMMING WINDOW

```python
M=1.99

c = (0.54-0.46*(np.cos(2*np.pi*t1/M-1))) # Hamming window
ham = DFT(c,N1)
res=[]
for i in n1:
    r = X1_dft[i]*ham[i]
    res.append(r)
   
 
d2=DFT(res,N1)
plt.stem(c)
plt.title('HAMMING window function')
```

```python
plt.psd(d2)
plt.title('PSD after Window applied on the signal in db/Hz')
```

### BARTLETT WINDOW

```python
c = np.bartlett(99) # Bartlett window function
Bar = DFT(c,N1)
res=[]
for i in n1:
    r = X1_dft[i]*Bar[i]
    res.append(r)
    
d3=DFT(res,N1)
plt.stem(c)
plt.title('Bartlett window function')
    
```

```python
plt.psd(d3)
plt.title('PSD after Window applied on the signal in db/Hz')
```

###  3.  x(t) = cos(2πf1t) + cos(2πf2t) where f1 = 10 Hz and f2 = 30 Hz. The signal y(t) is the noisy version of x(t)

```python
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt

f1=10
f2=30
t=np.arange(0,1.01,0.01)
x= np.cos(2*np.pi*f1*t) +  np.cos(2*np.pi*f2*t)
plt.plot(t,x)
plt.title('Original signal')
```

```python
noise =np.random.normal(0, 1, x.shape) 
y = x + noise
plt.plot(t,y)
plt.title('Signal after gaussian noise is added')
```

```python
noise =np.random.normal(0, 1, x.shape) 
y = x + noise
plt.psd(y)
plt.title('PSD of the noisy signal in db/Hz')
```
