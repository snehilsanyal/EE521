```python
import speech_recognition as sr 
import pyttsx3
from datetime import date
import numpy as np
import webbrowser
from googlesearch import search
from os import startfile
from os import system
import matplotlib.pyplot as plt
r = sr.Recognizer()
def SpeechtoText():
    MyText="Hello"
    try:
        # use the microphone as source for input. 
        with sr.Microphone() as source2: 
            r.adjust_for_ambient_noise(source2, duration=0.2)#listens for the user's input
            text="Please Speak"            
            print(text)
            SpeakText(text)
            audio2 = r.listen(source2) 	# Using google to recognize audio 
            MyText = r.recognize_google(audio2) 
            MyText = MyText.lower() 
    except sr.RequestError as e: 
        print("Could not request results; {0}".format(e))
    except sr.UnknownValueError: 
            print("Speak Again\n")
    return MyText
def PlotSin(fig):
    f=int(input("Enter the frequency\n"))
    A=float(input("Enter Amplitude\n"))
    phi=int(input("Enter Phase in Degree"))
    t=np.linspace(0,8/f,1000)
    x=A*np.sin(2*np.pi*f*t+phi)
    plt.figure(fig,figsize=(15,4))
    plt.plot(t,x,color='b')
    fig+=1
    plt.grid('true')
    plt.title("Sine wave x(t)="+str(A)+"sin(2πft+phi) where f="+str(f)+"Hz and phi="+str(phi))
    plt.xlabel('t→')
    plt.ylabel('Amp→')
    fig+=1
    return fig
def SpeakText(command): 
#engine initialization 
    engine = pyttsx3.init() 
    engine.say(command) 
    engine.runAndWait()
    engine.runAndWait()
def nishan():
    # recognizer initialization
    # convert text to speech 
    fig=1
    q=0
    r=0
    mt=""
    a=int(input("Please press\n1 for recognition of speech through microphone.\n2 for recognition of speech through file.\n"))
    close=0
    if a==1:
        while(close!=1):
            text="Hello, please tell what you want to do? Open a website or know time or plot a sinusoidal signal. Say close the session to close. Please speak only when i say."
            print(text)
            SpeakText(text)
            mt=SpeechtoText()
            print("Did you say "+mt) 
            SpeakText(mt)
            if mt=="open a website" or mt=="open website":
                while (r!=1):
                    print("Which website you want me to open? Like google or news or Facebook or Twitter or Youtube or Corona cases?")
                    SpeakText("Which website you want me to open? Like google or news or Facebook or Twitter or Youtube or Corona cases?")
                    mt=SpeechtoText()
                    if mt=="youtube" or mt=="you tube":
                        url="https://www.youtube.com"
                        r=1
                        close=1
                    elif mt=="news" :
                        url="https://www.timesofindia.indiatimes.com"
                        r=1
                        close=1
                    elif mt=="google"  :
                        url="https://www.google.com"
                        r=1
                        close=1
                    elif mt=="facebook" :
                        url="https://www.facebook.com"
                        r=1
                        close=1
                    elif mt=="twitter":
                        url="https://twitter.com"
                        r=1
                        close=1
                    elif mt=="corona cases" or mt=="corona" or mt=="covid":
                        url="https://www.covid19india.org"
                        r=1
                        close=1
                    else:
                        print("your choice not found.")
                        SpeakText("Your command not found.")
                        continue
                    SpeakText("Opening "+mt)
                    webbrowser.register('chrome',None,webbrowser.BackgroundBrowser("C:\Program Files\Google\Chrome\Application\chrome.exe"))
                    webbrowser.get('chrome').open(url)
            elif mt=="know time" or mt=="know date" or mt=="know time and date" or mt=="no time":
                today = str(date.today())
                SpeakText(today)
            elif mt=="plot a sine signal" or mt=="plot sign signal"  or mt=="plot a sinusoidal signal" or mt=="plot a sinusoid":
                while(q!=1):
                    print("Please enter the frequency, amplitude and phase in the box")
                    SpeakText("Please enter the frequency, amplitude and phase in the box")
                    fig=PlotSin(fig)
                    q=1
                    close=1
                    break
            elif mt=="close the session" or mt=="quit" or mt=="exit" or mt=="close session" or mt=="close":
                close=1
            else:
                print("Your choice not found.")
    elif a==2:
        filename=input("Enter file name\n")
        AUDIO_FILE = (filename)   # use the audio file as the audio source 
        r = sr.Recognizer() 
        with sr.AudioFile(AUDIO_FILE) as source: 
            audio = r.record(source)   
        try: 
            print("The Text from audio file are:\n" + r.recognize_google(audio)) 
            SpeakText(r.recognize_google(audio))
        except sr.UnknownValueError: 
            print("Audio is not Clear") 
        except sr.RequestError as e: 
            print("Try Again; {0}".format(e))
    else:
        print("Wrong choice selected\n")
        nishan()
    print("Thankyou for using the software created by Squadron Leader Nishan Tiwari. Good Bye")
    SpeakText("Thankyou for using the software created by Squadron Leader knee shawn Tea War e. Good Bye.")
nishan()
```
