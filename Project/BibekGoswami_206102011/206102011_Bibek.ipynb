```python
# Importing libraries
import numpy as np
from scipy import signal
from matplotlib import pyplot as plt
```

```python
# Definning the time frame
t = np.arange(0,5,0.025)

desired = 5*np.sin(2*np.pi*t) # Desired Output Signal
noise = np.random.normal(0,1,len(t)) # Gaussian Noise
refer = np.random.normal(0,2,len(t)) # Reference Signal correlated to noise
primary = desired + noise # Corrupted Signal

order = 2 # Order of the filter
mu = 0.01 # Step Size
n = len(primary) # Signal Length
delayed = adap = np.zeros(order) # Definning the weights and delayed signal
output = np.zeros(n) # Definning the output signal

for i in range(n):
    delayed[0] = refer[i] # Setting the delayed signal
    y = np.dot(delayed,adap) # Computing the linear operation
    output[i] = primary[i] - y # Computing the output
    adap = adap + (2*mu*(primary[i] - y)*delayed) # Updating the weights
    delayed[1:order] = delayed[0:order-1] # Updating the delayed signal
    

# Plotting the signals
fig = plt.figure(figsize=(10,25))

plt.subplot(4,1,1)
plt.plot(t,desired)
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Desired Signal')

plt.subplot(4,1,2)
plt.plot(t,refer)
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Noise Signal')

plt.subplot(4,1,3)
plt.plot(t,primary)
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Corrupted Signal')

plt.subplot(4,1,4)
plt.plot(t,output)
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Noise Cancelled Signal')
```
