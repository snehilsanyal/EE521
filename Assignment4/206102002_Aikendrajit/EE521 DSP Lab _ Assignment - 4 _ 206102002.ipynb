### Discrete Fourier Transform


```python
def DFT(array):
    array = np.asarray(array, dtype = complex)
    N = array.shape[0]
    k = np.arange(N)
    n = k.reshape((N, 1))
    M = np.exp(-2j * np.pi * k * n / N)
    return np.dot(M, array)
    

def IDFT(array):
    array = np.asarray(array, dtype = complex)
    N = array.shape[0]
    n = np.arange(N)
    k = n.reshape((N, 1))
    M = np.exp(2j * np.pi * k * n / N)
    return 1 / N * np.dot(M, array)
```

```python
def circular_convo(x, h):
    n = max(len(x), len(h))
    output = np.zeros(n)
    for i in range(len(x)):
        for j in range(len(h)):
            k = (i + j) % n
            output[k] += x[i] * h[j]
    return output

x = [1, 2, 3, 0, 0, 0, 0]
h = [1, 1, 1, 1, 1, 1, 1]

print("Required circulation convolution of the given 2 sequences is:", circular_convo(x, h))
```

```python
def DFT_(x, N):
    X = np.zeros(N, dtype = complex)
    for k in range(N):
        for n in range(N):
            X[k] += x[n] * np.exp(-1 * (2j * np.pi / N) * k * n)
    return X
```

### Discrete Cosine Transform:

```python
def DCT(x, N):
    X = np.zeros(N)
    for m in range(N):
        for n in range(N):
            if m == 0:
                cm = 1/(np.sqrt(2))
            else:
                cm = 1
            X[m] += x[n] * cm * np.cos(((np.pi * m) / (2 * N)) * (2 * n + 1))
    X = X * np.sqrt(2/N)
    return X
```

```python
def IDCT(x, N):
    X = np.zeros(N)
    for n in range(N):
        for m in range(N):
            if m == 0:
                cm = 1/(np.sqrt(2))
            else:
                cm = 1
            X[n] += x[m] * cm * np.cos(((np.pi * m) / (2 * N)) * (2 * n + 1))
    X = X * np.sqrt(2/N)
    return X
```

```python
# Given

x = [8, 16, 24, 32, 40, 48, 56, 64]
# and
N = 8

# Now

output_DCT = DCT(x, N)
print("Output DCT of x is:", output_DCT)
```
